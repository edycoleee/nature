```python
#SCRIPT34
import tkinter as tk
import math

# ----------------------------
# Kelas Vektor 2D
# ----------------------------
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self, v):
        self.x += v.x
        self.y += v.y

    def limit(self, max_val):
        mag = math.sqrt(self.x**2 + self.y**2)  # panjang vektor
        if mag > max_val:
            self.x = (self.x / mag) * max_val
            self.y = (self.y / mag) * max_val

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2)  # panjang vektor

# ----------------------------
# Kelas Mobil / Pelari
# ----------------------------
class Car:
    def __init__(self, canvas, width, height):
        self.canvas = canvas
        self.width = width
        self.height = height
        self.radius = 15

        self.location = Vector(width // 2, height // 2)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.topspeed = 7  # batas kecepatan maksimum

        # Gambar mobil/pelari
        self.id = canvas.create_oval(
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius,
            fill="blue"
        )

        # Tambahan: Tampilkan kecepatan
        self.speed_text = canvas.create_text(10, 10, anchor="nw", text="Speed: 0.00", font=("Arial", 12), fill="black")

    def update(self):
        self.velocity.add(self.acceleration)       # tambahkan percepatan ke kecepatan
        self.velocity.limit(self.topspeed)         # batasi kecepatan maksimum
        self.location.add(self.velocity)           # posisi berubah
        self.check_edges()                         # cek jika keluar layar

        # Gambar ulang posisi mobil
        self.canvas.coords(self.id,
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius)

        # Tampilkan kecepatan saat ini
        speed = self.velocity.magnitude()
        self.canvas.itemconfig(self.speed_text, text=f"Speed: {speed:.2f}")

    def check_edges(self):
        if self.location.x > self.width:
            self.location.x = 0
        elif self.location.x < 0:
            self.location.x = self.width
        if self.location.y > self.height:
            self.location.y = 0
        elif self.location.y < 0:
            self.location.y = self.height

    def accelerate(self):
        self.acceleration = Vector(0.05, 0)  # dorong ke kanan

    def brake(self):
        self.acceleration = Vector(-0.05, 0)  # rem ke kiri

    def stop(self):
        self.acceleration = Vector(0, 0)

# ----------------------------
# Fungsi utama
# ----------------------------
def main():
    WIDTH, HEIGHT = 600, 400
    root = tk.Tk()
    root.title("Exercise 1.5 - Mobil Dorong Rem")

    canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
    canvas.pack()

    car = Car(canvas, WIDTH, HEIGHT)

    # Fungsi animasi berulang
    def animate():
        car.update()
        root.after(30, animate)

    # Tangani tombol keyboard
    def on_key_press(event):
        if event.keysym == "Up":
            car.accelerate()
        elif event.keysym == "DownS":
            car.brake()

    def on_key_release(event):
        car.stop()

    root.bind("<KeyPress>", on_key_press)
    root.bind("<KeyRelease>", on_key_release)

    animate()
    root.mainloop()

main()
```

```python
#SCRIPT35
import tkinter as tk
import math
import random

# -----------------------------
# Kelas Vektor: menyimpan x dan y
# -----------------------------
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self, v):  # Menambahkan vektor lain
        self.x += v.x
        self.y += v.y

    def mult(self, scalar):  # Mengalikan dengan angka
        self.x *= scalar
        self.y *= scalar

    def limit(self, max_val):  # Membatasi panjang vektor
        mag = math.sqrt(self.x**2 + self.y**2)
        if mag > max_val:
            self.x = (self.x / mag) * max_val
            self.y = (self.y / mag) * max_val

    @staticmethod
    def random2D():  # Membuat vektor acak 2D
        angle = random.uniform(0, 2 * math.pi)
        return Vector(math.cos(angle), math.sin(angle))

    def magnitude(self):  # Menghitung panjang (besar) vektor
        return math.sqrt(self.x**2 + self.y**2)

# -----------------------------
# Kelas Mover: Bola yang bisa bergerak
# -----------------------------
class Mover:
    def __init__(self, canvas, width, height):
        self.canvas = canvas
        self.width = width
        self.height = height
        self.radius = 15

        self.location = Vector(width // 2, height // 2)  # posisi awal
        self.velocity = Vector(0, 0)                     # kecepatan awal nol
        self.acceleration = Vector(0, 0)                 # percepatan awal nol
        self.topspeed = 5                                # kecepatan maksimum

        # Gambar bola
        self.id = canvas.create_oval(
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius,
            fill="green"
        )

        # Teks perhitungan di layar
        self.text_acc = canvas.create_text(10, 10, anchor="nw", text="", font=("Arial", 10), fill="black")
        self.text_vel = canvas.create_text(10, 30, anchor="nw", text="", font=("Arial", 10), fill="blue")
        self.text_pos = canvas.create_text(10, 50, anchor="nw", text="", font=("Arial", 10), fill="red")

    def update(self):
        # 1. Buat percepatan acak
        self.acceleration = Vector.random2D()
        self.acceleration.mult(0.5)  # percepatan acak tapi kecil

        # 2. Tambahkan percepatan ke kecepatan
        self.velocity.add(self.acceleration)

        # 3. Batasi kecepatan agar tidak terlalu cepat
        self.velocity.limit(self.topspeed)

        # 4. Tambahkan kecepatan ke posisi (gerakkan bola)
        self.location.add(self.velocity)

        # 5. Jika keluar layar, pindah ke sisi sebaliknya
        self.check_edges()

        # 6. Gambar ulang posisi bola
        self.canvas.coords(self.id,
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius)

        # 7. Tampilkan perhitungan di layar
        acc_mag = self.acceleration.magnitude()
        vel_mag = self.velocity.magnitude()
        pos_x, pos_y = self.location.x, self.location.y

        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: {acc_mag:.2f}")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: {vel_mag:.2f}")
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({pos_x:.1f}, {pos_y:.1f})")

    def check_edges(self):
        # Jika bola keluar layar, masuk dari sisi sebaliknya
        if self.location.x > self.width:
            self.location.x = 0
        elif self.location.x < 0:
            self.location.x = self.width
        if self.location.y > self.height:
            self.location.y = 0
        elif self.location.y < 0:
            self.location.y = self.height

# -----------------------------
# Fungsi utama untuk menjalankan animasi
# -----------------------------
def main():
    WIDTH, HEIGHT = 600, 400
    root = tk.Tk()
    root.title("Simulasi Bola Acak dengan Percepatan")
    canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
    canvas.pack()

    mover = Mover(canvas, WIDTH, HEIGHT)

    def animate():
        mover.update()          # update posisi dan kecepatan
        root.after(30, animate) # ulangi tiap 30 ms

    animate()
    root.mainloop()

main()
```

```python
#SCRIPT41
import tkinter as tk
import math

class Bola:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.location = [x, y]          # Posisi bola (x, y)
        self.velocity = [0, 0]          # Kecepatan awal (vx, vy)
        self.acceleration = [0, 0]      # Percepatan awal (ax, ay)
        self.radius = 10
        self.topspeed = 10              # Kecepatan maksimum

        # Gambar bola di layar
        self.id = canvas.create_oval(x - self.radius, y - self.radius,
                                     x + self.radius, y + self.radius,
                                     fill='blue')

        # Tambahkan teks untuk menampilkan hasil perhitungan
        self.text_acc = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_pos = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')

    def update(self, mouse_x, mouse_y):
        # ðŸ§­ Step 1: Hitung arah dari bola ke mouse
        dx = mouse_x - self.location[0]
        dy = mouse_y - self.location[1]

        # ðŸ“ Step 2: Hitung jarak dan normalisasi arah
        dist = math.sqrt(dx**2 + dy**2)
        if dist != 0:
            dx /= dist
            dy /= dist

        # ðŸŒ€ Step 3: Skala percepatan (agar kecil)
        dx *= 0.5
        dy *= 0.5

        # ðŸ’¨ Step 4: Simpan sebagai percepatan
        self.acceleration = [dx, dy]

        # âž• Tambahkan percepatan ke kecepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        # ðŸš« Batasi kecepatan maksimum
        speed = math.sqrt(self.velocity[0]**2 + self.velocity[1]**2)
        if speed > self.topspeed:
            self.velocity[0] = (self.velocity[0] / speed) * self.topspeed
            self.velocity[1] = (self.velocity[1] / speed) * self.topspeed

        # ðŸ“ Update posisi
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        # ðŸ” Gambar ulang posisi bola
        x, y = self.location
        self.canvas.coords(self.id, x - self.radius, y - self.radius, x + self.radius, y + self.radius)

        # ðŸ§¾ Tampilkan perhitungan di layar
        ax, ay = self.acceleration
        vx, vy = self.velocity
        px, py = self.location

        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: ({ax:.2f}, {ay:.2f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: ({vx:.2f}, {vy:.2f})")
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({px:.1f}, {py:.1f})")

# ðŸŒŸ Setup Tkinter
root = tk.Tk()
root.title("Simulasi Bola Mengejar Mouse")

canvas = tk.Canvas(root, width=600, height=400, bg="white")
canvas.pack()

bola = Bola(canvas, 300, 200)

# ðŸ” Fungsi animasi: jalan terus setiap 16 milidetik
def animasi():
    # Dapatkan posisi mouse relatif terhadap canvas
    x = canvas.winfo_pointerx() - canvas.winfo_rootx()
    y = canvas.winfo_pointery() - canvas.winfo_rooty()
    bola.update(x, y)
    root.after(16, animasi)  # sekitar 60 frame per detik

animasi()
root.mainloop()
```

```python
#SCRIPT43
import tkinter as tk
import math

class Bola:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.location = [x, y]
        self.velocity = [0, 0]
        self.acceleration = [0, 0]
        self.radius = 10
        self.topspeed = 10
        self.id = canvas.create_oval(x - self.radius, y - self.radius,
                                     x + self.radius, y + self.radius,
                                     fill='green')

        # ðŸŒ€ Trail effect
        self.trails = []  # menyimpan jejak
        self.trail_ids = []  # menyimpan id jejak yang digambar

        # Info teks
        self.text_acc = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_pos = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')
        self.text_dist = canvas.create_text(10, 70, anchor='nw', text="", font=('Arial', 10), fill='purple')

    def update(self, mouse_x, mouse_y):
        # Hitung arah
        dx = mouse_x - self.location[0]
        dy = mouse_y - self.location[1]

        # Hitung jarak
        distance = math.sqrt(dx**2 + dy**2)
        if distance == 0:
            return

        # Normalisasi
        dx /= distance
        dy /= distance

        # Percepatan bervariasi
        magnitude = distance * 0.01
        self.acceleration = [dx * magnitude, dy * magnitude]

        # Update kecepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        # Batasi kecepatan
        speed = math.sqrt(self.velocity[0]**2 + self.velocity[1]**2)
        if speed > self.topspeed:
            self.velocity[0] = (self.velocity[0] / speed) * self.topspeed
            self.velocity[1] = (self.velocity[1] / speed) * self.topspeed

        # Update posisi
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        x, y = self.location
        self.canvas.coords(self.id, x - self.radius, y - self.radius,
                                      x + self.radius, y + self.radius)

        # âœ¨ Tambahkan titik ke trail
        self.trails.append((x, y))
        if len(self.trails) > 30:
            self.trails.pop(0)  # hapus jejak paling lama

        # ðŸ§¹ Hapus jejak lama yang digambar
        for trail_id in self.trail_ids:
            self.canvas.delete(trail_id)
        self.trail_ids.clear()

        # ðŸŽ¨ Gambar ulang trail
        for i, (tx, ty) in enumerate(self.trails):
            size = max(2, 6 - (30 - len(self.trails) + i) * 0.1)  # makin lama makin kecil
            color = f'#00aa00'  # warna hijau
            trail = self.canvas.create_oval(tx - size, ty - size, tx + size, ty + size, fill=color, outline="")
            self.trail_ids.append(trail)

        # Tampilkan info
        ax, ay = self.acceleration
        vx, vy = self.velocity
        px, py = self.location

        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: ({ax:.2f}, {ay:.2f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: ({vx:.2f}, {vy:.2f})")
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({px:.1f}, {py:.1f})")
        self.canvas.itemconfig(self.text_dist, text=f"Jarak ke mouse: {distance:.1f}")

# Setup tkinter
root = tk.Tk()
root.title("Bola dengan Percepatan Variabel + Trail")

canvas = tk.Canvas(root, width=600, height=400, bg="white")
canvas.pack()

bola = Bola(canvas, 300, 200)

def animasi():
    x = canvas.winfo_pointerx() - canvas.winfo_rootx()
    y = canvas.winfo_pointery() - canvas.winfo_rooty()
    bola.update(x, y)
    root.after(16, animasi)

animasi()
root.mainloop()
```

```python
#SCRIPT44
import tkinter as tk       # Modul untuk membuat GUI jendela dan gambar
import math                # Untuk menghitung jarak dan kecepatan
import random              # Untuk memilih posisi awal secara acak

# ðŸŽ‚ Cetakan untuk setiap bola
class Mover:
    def __init__(self, canvas, width, height):
        self.canvas = canvas             # Tempat menggambar (kanvas)
        self.width = width              # Lebar area layar
        self.height = height            # Tinggi area layar
        # ðŸŽ¯ Posisi awal acak di layar
        self.location = [random.randint(0, width), random.randint(0, height)]
        self.velocity = [0, 0]          # Kecepatan awal nol
        self.acceleration = [0, 0]      # Percepatan awal nol
        self.radius = 8                 # Ukuran bola
        self.topspeed = 4               # Batas kecepatan maksimum
        # ðŸ”µ Gambar bola di canvas (awal kosong, nanti di-update)
        self.id = canvas.create_oval(0, 0, 0, 0, fill='skyblue', outline='black')

    # ðŸš€ Fungsi untuk memperbarui posisi bola agar mendekati mouse
    def update(self, mouse_x, mouse_y):
        dx = mouse_x - self.location[0]     # Selisih posisi horizontal ke mouse
        dy = mouse_y - self.location[1]     # Selisih posisi vertikal ke mouse

        dist = math.sqrt(dx**2 + dy**2)     # Jarak ke mouse (pakai rumus Pythagoras)
        if dist == 0:
            return                          # Kalau sudah tepat di mouse, tidak perlu gerak

        dx /= dist                          # Normalisasi arah agar hanya menunjukkan arah
        dy /= dist

        magnitude = 0.5                     # Besarnya percepatan (selalu tetap)
        self.acceleration = [dx * magnitude, dy * magnitude]  # ðŸš— Arah percepatan ke mouse

        # ðŸ’¨ Update kecepatan berdasarkan percepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        speed = math.sqrt(self.velocity[0]**2 + self.velocity[1]**2)  # Hitung total kecepatan

        # ðŸ›‘ Batasi kecepatan agar tidak terlalu cepat
        if speed > self.topspeed:
            self.velocity[0] = (self.velocity[0] / speed) * self.topspeed
            self.velocity[1] = (self.velocity[1] / speed) * self.topspeed

        # ðŸ§­ Update posisi berdasarkan kecepatan
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        self.check_edges()  # ðŸŒ Jika keluar layar, muncul dari sisi lain

        # ðŸŽ¨ Gambar ulang bola di posisi barunya
        x, y = self.location
        self.canvas.coords(self.id, x - self.radius, y - self.radius,
                                      x + self.radius, y + self.radius)

    # ðŸ”„ Fungsi jika bola keluar dari layar, muncul kembali dari sisi berlawanan
    def check_edges(self):
        if self.location[0] > self.width:
            self.location[0] = 0
        elif self.location[0] < 0:
            self.location[0] = self.width

        if self.location[1] > self.height:
            self.location[1] = 0
        elif self.location[1] < 0:
            self.location[1] = self.height

# ==== Bagian Utama Program ====
root = tk.Tk()  # Buat jendela utama
root.title("Array of Movers - Menuju Mouse")

canvas_width = 640
canvas_height = 360
canvas = tk.Canvas(root, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# ðŸŸ¢ Buat 20 bola yang disimpan dalam daftar movers
movers = [Mover(canvas, canvas_width, canvas_height) for _ in range(20)]

# ðŸŽ¬ Fungsi animasi berulang (frame per frame)
def animate():
    # Ambil posisi mouse saat ini
    x = canvas.winfo_pointerx() - canvas.winfo_rootx()
    y = canvas.winfo_pointery() - canvas.winfo_rooty()

    # Update semua bola untuk bergerak mendekati mouse
    for mover in movers:
        mover.update(x, y)

    root.after(16, animate)  # Panggil fungsi ini lagi setelah 16ms (~60fps)

animate()  # Jalankan animasi pertama
root.mainloop()  # Tampilkan jendela GUI
```

```python
#SCRIPT45
import tkinter as tk
import random
import math

# -----------------------------
# Bagian 1: Membuat Kelas Ikan
# -----------------------------
class Fish:
    def __init__(self, canvas, lebar, tinggi):
        # Set awal posisi dan kecepatan ikan
        self.canvas = canvas  # Tempat gambar ikan
        self.lebar = lebar    # Lebar layar
        self.tinggi = tinggi  # Tinggi layar
        
        # Posisi awal acak
        self.posisi = [random.randint(0, lebar), random.randint(0, tinggi)]
        
        # Kecepatan awal acak antara -2 sampai 2
        self.kecepatan = [random.uniform(-2, 2), random.uniform(-2, 2)]
        
        # Awalnya tidak ada percepatan
        self.percepatan = [0, 0]  
        
        # Kecepatan maksimum ikan
        self.kecepatan_max = 3  
        
        # Bentuk ikan (lingkaran orange)
        self.bentuk = canvas.create_oval(0, 0, 0, 0, fill='orange')
        self.ukuran = 12  # Ukuran ikan
# -----------------------------
# Bagian 2: Cara Ikan Bergerak
# -----------------------------
    def update(self):
        # 1. Kadang ikan ganti arah tiba-tiba (5% kemungkinan)
        if random.random() < 0.05:
            sudut = random.uniform(0, 2 * math.pi)  # Arah acak
            kekuatan = 0.5
            self.percepatan = [math.cos(sudut) * kekuatan, math.sin(sudut) * kekuatan]
        
        # 2. Hitung kecepatan baru
        self.kecepatan[0] += self.percepatan[0]  # X
        self.kecepatan[1] += self.percepatan[1]  # Y
        
        # 3. Pastikan tidak terlalu cepat
        total_cepat = math.sqrt(self.kecepatan[0]**2 + self.kecepatan[1]**2)
        if total_cepat > self.kecepatan_max:
            self.kecepatan[0] = (self.kecepatan[0] / total_cepat) * self.kecepatan_max
            self.kecepatan[1] = (self.kecepatan[1] / total_cepat) * self.kecepatan_max
        
        # 4. Pindahkan ikan
        self.posisi[0] += self.kecepatan[0]
        self.posisi[1] += self.kecepatan[1]
        
        # 5. Cek tepi layar
        self.cek_tepi()
        
        # 6. Gambar ikan di posisi baru
        x, y = self.posisi
        self.canvas.coords(self.bentuk, 
                          x - self.ukuran, y - self.ukuran,
                          x + self.ukuran, y + self.ukuran)
# -----------------------------
# Bagian 3: Ikan Takut Tepi Layar
# -----------------------------
    def cek_tepi(self):
        jarak_aman = 20  # Jarak dari tepi saat mulai takut
        kekuatan_balik = 1.5  # Seberapa kuat menghindar
        
        # Jika terlalu kiri
        if self.posisi[0] < jarak_aman:
            self.percepatan[0] = kekuatan_balik  # Dorong ke kanan
            
        # Jika terlalu kanan
        elif self.posisi[0] > self.lebar - jarak_aman:
            self.percepatan[0] = -kekuatan_balik  # Dorong ke kiri
            
        # Jika terlalu atas
        if self.posisi[1] < jarak_aman:
            self.percepatan[1] = kekuatan_balik  # Dorong ke bawah
            
        # Jika terlalu bawah
        elif self.posisi[1] > self.tinggi - jarak_aman:
            self.percepatan[1] = -kekuatan_balik  # Dorong ke atas
# -----------------------------
# Bagian 4: Membuat Layar dan Animasi
# -----------------------------
# Buat jendela utama
root = tk.Tk()
root.title("Ikan Gugup")

# Ukuran kolam (canvas)
lebar_kolam = 640
tinggi_kolam = 360
kolam = tk.Canvas(root, width=lebar_kolam, height=tinggi_kolam, bg="lightblue")
kolam.pack()

# Buat satu ikan
ikan = Fish(kolam, lebar_kolam, tinggi_kolam)

# Fungsi animasi
def animasi():
    ikan.update()  # Update posisi ikan
    root.after(20, animasi)  # Ulangi setiap 20ms

# Mulai animasi
animasi()
root.mainloop()
```

```python
#SCRIPT46
# -----------------------------
# ðŸ‡ Bagian 1: Membuat Kelas Kelinci
# -----------------------------
import tkinter as tk
import random
import math

class Kelinci:
    def __init__(self, kanvas, lebar, tinggi):
        self.kanvas = kanvas  # Tempat menggambar
        self.lebar = lebar    # Lebar layar
        self.tinggi = tinggi  # Tinggi layar
        
        # Posisi awal acak
        self.posisi = [random.randint(0, lebar), random.randint(0, tinggi)]
        
        # Kecepatan awal (diam)
        self.kecepatan = [0, 0]
        
        # Percepatan awal (tidak ada)
        self.percepatan = [0, 0]
        
        # Ukuran kelinci
        self.ukuran = 14
        
        # Kecepatan maksimum
        self.kecepatan_max = 7
        
        # Waktu tunggu sebelum lompat
        self.waktu_tunggu = 0
        
        # Bentuk kelinci (lingkaran pink)
        self.bentuk = kanvas.create_oval(0, 0, 0, 0, fill='pink')

# -----------------------------
# ðŸ“Š Bagian 2: Logika Melompat (Update)
# -----------------------------
    def update(self):
        # Tampilkan informasi perhitungan
        print(f"\n--- Perhitungan Frame Baru ---")
        print(f"Posisi awal: ({self.posisi[0]:.1f}, {self.posisi[1]:.1f})")
        print(f"Kecepatan awal: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 1. Cek apakah sudah waktunya lompat
        if self.waktu_tunggu <= 0:
            print("Waktunya lompat!")
            # Buat lompatan acak
            sudut = random.uniform(0, 2 * math.pi)  # Arah acak
            kekuatan = random.uniform(3, 6)         # Kekuatan acak
            self.percepatan = [
                math.cos(sudut) * kekuatan, 
                math.sin(sudut) * kekuatan
            ]
            print(f"Percepatan baru: ({self.percepatan[0]:.1f}, {self.percepatan[1]:.1f})")
            
            # Set waktu tunggu acak (20-40 frame)
            self.waktu_tunggu = random.randint(20, 40)
            print(f"Set waktu tunggu: {self.waktu_tunggu} frame")
        else:
            # Tidak lompat, kurangi waktu tunggu
            self.waktu_tunggu -= 1
            self.percepatan = [0, 0]
            print(f"Menunggu... sisa {self.waktu_tunggu} frame")
        
        # 2. Update kecepatan
        self.kecepatan[0] += self.percepatan[0]
        self.kecepatan[1] += self.percepatan[1]
        print(f"Kecepatan setelah percepatan: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 3. Perlambatan (efek gesekan)
        self.kecepatan[0] *= 0.85
        self.kecepatan[1] *= 0.85
        print(f"Kecepatan setelah perlambatan: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 4. Batasi kecepatan maksimum
        total_cepat = math.sqrt(self.kecepatan[0]**2 + self.kecepatan[1]**2)
        if total_cepat > self.kecepatan_max:
            self.kecepatan[0] = (self.kecepatan[0]/total_cepat) * self.kecepatan_max
            self.kecepatan[1] = (self.kecepatan[1]/total_cepat) * self.kecepatan_max
            print(f"Kecepatan dibatasi: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 5. Update posisi
        self.posisi[0] += self.kecepatan[0]
        self.posisi[1] += self.kecepatan[1]
        print(f"Posisi baru: ({self.posisi[0]:.1f}, {self.posisi[1]:.1f})")
        
        # 6. Cek tepi layar
        self.cek_tepi()
        
        # 7. Gambar ulang kelinci
        x, y = self.posisi
        self.kanvas.coords(self.bentuk, 
                          x - self.ukuran, y - self.ukuran,
                          x + self.ukuran, y + self.ukuran)

# -----------------------------
# ðŸŒ Bagian 3: Cek Tepi Layar
# -----------------------------
    def cek_tepi(self):
        # Jika kelinci keluar layar, muncul di sisi seberang
        if self.posisi[0] < 0:
            self.posisi[0] = self.lebar
            print("Kelinci keluar kiri, muncul di kanan")
        elif self.posisi[0] > self.lebar:
            self.posisi[0] = 0
            print("Kelinci keluar kanan, muncul di kiri")
            
        if self.posisi[1] < 0:
            self.posisi[1] = self.tinggi
            print("Kelinci keluar atas, muncul di bawah")
        elif self.posisi[1] > self.tinggi:
            self.posisi[1] = 0
            print("Kelinci keluar bawah, muncul di atas")

# -----------------------------
# ðŸ–¥ï¸ Bagian 4: Program Utama
# -----------------------------
# Buat jendela utama
root = tk.Tk()
root.title("Kelinci Melompat")

# Ukuran layar
lebar_layar = 640
tinggi_layar = 360

# Buat kanvas (area gambar)
kanvas = tk.Canvas(root, width=lebar_layar, height=tinggi_layar, bg="lightgreen")
kanvas.pack()

# Buat satu kelinci
kelinci = Kelinci(kanvas, lebar_layar, tinggi_layar)

# Fungsi animasi
def animasi():
    kelinci.update()
    root.after(30, animasi)  # Update setiap 30ms

# Mulai animasi
print("=== Program Dimulai ===")
animasi()
root.mainloop()
```

```python
#SCRIPT48
import tkinter as tk
import random

class NewtonFirstLawDemo:
    def __init__(self, root):
        self.root = root
        self.root.title("Demonstrasi Hukum Pertama Newton")
        
        # Buat canvas untuk menggambar
        self.width = 800
        self.height = 400
        self.canvas = tk.Canvas(root, width=self.width, height=self.height, bg='white')
        self.canvas.pack(pady=20)
        
        # Teks penjelasan
        explanation = """HUKUM PERTAMA NEWTON (HUKUM INERSIA):
Sebuah benda akan mempertahankan keadaan geraknya 
(tetap diam atau bergerak lurus beraturan) 
kecuali ada gaya eksternal yang bekerja padanya"""
        self.canvas.create_text(self.width//2, 30, text=explanation, font=('Arial', 12), justify='center')
        
        # Buat tombol
        self.start_btn = tk.Button(root, text="Mulai Demo", command=self.start_demo)
        self.start_btn.pack()
        self.reset_btn = tk.Button(root, text="Reset", command=self.reset)
        self.reset_btn.pack()
        
        # Variabel untuk objek
        self.ball = None
        self.ball_moving = False
        self.velocity = 5  # Kecepatan awal
        self.friction = 0  # Gaya gesek (awal tidak ada)
        
        # Tambah slider untuk gesekan
        self.friction_label = tk.Label(root, text="Gaya Gesek: 0")
        self.friction_label.pack()
        self.friction_slider = tk.Scale(root, from_=0, to=0.5, resolution=0.01, orient='horizontal', command=self.set_friction)
        self.friction_slider.pack(fill='x', padx=20)
        
        # Tambah slider untuk kecepatan awal
        self.velocity_label = tk.Label(root, text="Kecepatan Awal: 5")
        self.velocity_label.pack()
        self.velocity_slider = tk.Scale(root, from_=1, to=20, orient='horizontal', command=self.set_velocity)
        self.velocity_slider.pack(fill='x', padx=20)
    
    # Fungsi untuk mengatur gesekan
    def set_friction(self, value):
        self.friction = float(value)
        self.friction_label.config(text=f"Gaya Gesek: {self.friction:.2f}")
    
    # Fungsi untuk mengatur kecepatan awal
    def set_velocity(self, value):
        self.velocity = int(value)
        self.velocity_label.config(text=f"Kecepatan Awal: {self.velocity}")
    
    # Fungsi untuk memulai demo
    def start_demo(self):
        if not self.ball_moving:
            self.create_ball()
            self.ball_moving = True
            self.move_ball()
    
    # Fungsi untuk membuat bola
    def create_ball(self):
        if self.ball:
            self.canvas.delete(self.ball)
        # Buat bola di posisi acak di kiri canvas
        start_y = random.randint(50, self.height-50)
        self.ball_pos = [50, start_y]
        self.ball = self.canvas.create_oval(50-20, start_y-20, 50+20, start_y+20, fill='blue')
        self.current_velocity = self.velocity  # Set kecepatan awal
    
    # Fungsi untuk menggerakkan bola
    def move_ball(self):
        if self.ball_moving:
            # Update posisi bola berdasarkan kecepatan
            self.ball_pos[0] += self.current_velocity
            self.canvas.coords(self.ball, 
                             self.ball_pos[0]-20, self.ball_pos[1]-20,
                             self.ball_pos[0]+20, self.ball_pos[1]+20)
            
            # Terapkan gesekan (jika ada)
            if self.friction > 0:
                self.current_velocity -= self.friction
                if self.current_velocity < 0:
                    self.current_velocity = 0
            
            # Cek jika bola mencapai tepi
            if self.ball_pos[0] > self.width + 20:
                self.ball_pos[0] = -20  # Kembali ke kiri
            elif self.ball_pos[0] < -20:
                self.ball_pos[0] = self.width + 20  # Kembali ke kanan
            
            # Panggil fungsi ini lagi setelah 50ms
            self.root.after(50, self.move_ball)
    
    # Fungsi untuk reset demo
    def reset(self):
        self.ball_moving = False
        if self.ball:
            self.canvas.delete(self.ball)
            self.ball = None

# Jalankan program
if __name__ == "__main__":
    root = tk.Tk()
    app = NewtonFirstLawDemo(root)
    root.mainloop()
```

```python
#COPY PASTE
#SCRIPT49
import tkinter as tk
import math

class NewtonThirdLawDemo:
    def __init__(self, root):
        self.root = root
        self.root.title("Demonstrasi Hukum Ketiga Newton")
        
        # Setup canvas
        self.width = 800
        self.height = 400
        self.canvas = tk.Canvas(root, width=self.width, height=self.height, bg='white')
        self.canvas.pack(pady=20)
        
        # Explanation text
        explanation = """HUKUM KETIGA NEWTON (AKSI-REAKSI):
Untuk setiap gaya aksi, selalu ada gaya reaksi 
yang sama besar tetapi berlawanan arah"""
        self.canvas.create_text(self.width//2, 30, text=explanation, font=('Arial', 12), justify='center')
        
        # Create objects
        self.rocket = self.canvas.create_rectangle(100, 200, 200, 300, fill='red', tags='rocket')
        self.exhaust = self.canvas.create_oval(90, 300, 110, 310, fill='yellow', state='hidden', tags='exhaust')
        self.earth = self.canvas.create_oval(400, 200, 600, 400, fill='blue', tags='earth')
        
        # Variables
        self.rocket_pos = [150, 250]  # Center of rocket
        self.rocket_vel = [0, 0]
        self.earth_pos = [500, 300]   # Center of earth
        self.thrust = 0.5
        self.gravity = 0.1
        self.fuel = 100
        
        # Controls
        self.start_btn = tk.Button(root, text="Mulai Mesin", command=self.start_engine)
        self.start_btn.pack(side='left', padx=10)
        self.reset_btn = tk.Button(root, text="Reset", command=self.reset)
        self.reset_btn.pack(side='left', padx=10)
        
        # Fuel gauge
        self.fuel_label = tk.Label(root, text="Bahan Bakar: 100%")
        self.fuel_label.pack(side='left', padx=10)
        
        # Start simulation
        self.simulating = False
        self.root.after(100, self.update)
    
    def start_engine(self):
        if self.fuel > 0:
            self.simulating = True
            self.canvas.itemconfig('exhaust', state='normal')
    
    def reset(self):
        self.simulating = False
        self.canvas.coords('rocket', 100, 200, 200, 300)
        self.canvas.itemconfig('exhaust', state='hidden')
        self.rocket_pos = [150, 250]
        self.rocket_vel = [0, 0]
        self.fuel = 100
        self.fuel_label.config(text="Bahan Bakar: 100%")
    
    def update(self):
        if self.simulating and self.fuel > 0:
            # AKSI: Rocket mendorong gas ke bawah (gaya pada gas)
            force_down = -self.thrust
            
            # REAKSI: Gas mendorong rocket ke atas (gaya pada rocket)
            force_up = -force_down  # Sama besar, berlawanan arah
            
            # Terapkan gaya pada rocket
            self.rocket_vel[1] -= force_up * 0.1
            
            # Kurangi bahan bakar
            self.fuel -= 0.5
            self.fuel_label.config(text=f"Bahan Bakar: {max(0, int(self.fuel))}%")
            
            # Hitung gravitasi bumi
            dx = self.earth_pos[0] - self.rocket_pos[0]
            dy = self.earth_pos[1] - self.rocket_pos[1]
            dist = max(50, math.sqrt(dx*dx + dy*dy))
            
            # Gaya gravitasi (aksi-reaksi juga)
            gravity_force = self.gravity * 1000 / (dist*dist)
            self.rocket_vel[0] += dx/dist * gravity_force * 0.1
            self.rocket_vel[1] += dy/dist * gravity_force * 0.1
            
            # Update posisi
            self.rocket_pos[0] += self.rocket_vel[0]
            self.rocket_pos[1] += self.rocket_vel[1]
            
            # Update gambar
            self.canvas.coords('rocket', 
                             self.rocket_pos[0]-50, self.rocket_pos[1]-50,
                             self.rocket_pos[0]+50, self.rocket_pos[1]+50)
            self.canvas.coords('exhaust',
                             self.rocket_pos[0]-10, self.rocket_pos[1]+50,
                             self.rocket_pos[0]+10, self.rocket_pos[1]+60)
            
            if self.fuel <= 0:
                self.canvas.itemconfig('exhaust', state='hidden')
        
        # Terus update
        self.root.after(50, self.update)

# Run the program
if __name__ == "__main__":
    root = tk.Tk()
    app = NewtonThirdLawDemo(root)
    root.mainloop()
```

```python
#SCRIPT50
import tkinter as tk
import math

# === CLASS VECTOR UNTUK POSISI, KECEPATAN, GAYA ===
class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, other):
        self.x += other.x
        self.y += other.y

    def sub(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def mult(self, scalar):
        self.x *= scalar
        self.y *= scalar

    def div(self, scalar):
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar

    def mag(self):  # Menghitung panjang vektor (besar gaya/kecepatan)
        return math.sqrt(self.x**2 + self.y**2)

    def normalize(self):  # Ubah jadi unit vector (arah saja)
        m = self.mag()
        if m != 0:
            self.div(m)

    def copy(self):
        return Vector(self.x, self.y)

# === CLASS UNTUK BENDA BERGERAK (MOVER) ===
class Mover:
    def __init__(self, mass, x, y):
        self.mass = mass  # massa benda
        self.location = Vector(x, y)  # posisi awal
        self.velocity = Vector(0, 0)  # kecepatan awal = 0
        self.acceleration = Vector(0, 0)  # percepatan awal = 0
        self.radius = mass * 10  # semakin besar massa, lingkaran lebih besar

    def applyForce(self, force):
        # HUKUM NEWTON II: a = F / m
        f = force.copy()
        f.div(self.mass)
        self.acceleration.add(f)

    def update(self):
        # Tambahkan percepatan ke kecepatan
        self.velocity.add(self.acceleration)
        # Tambahkan kecepatan ke posisi
        self.location.add(self.velocity)
        # Reset percepatan setelah setiap update
        self.acceleration.mult(0)

# === CLASS SIMULASI UTAMA ===
class Simulation:
    def __init__(self, root):
        self.root = root
        self.width = 600
        self.height = 400
        self.canvas = tk.Canvas(root, width=self.width, height=self.height, bg="white")
        self.canvas.pack()

        # Buat 2 mover dengan massa berbeda
        self.mover1 = Mover(mass=1, x=100, y=200)   # massa kecil
        self.mover2 = Mover(mass=4, x=100, y=300)   # massa besar

        # Gaya angin ke kanan
        self.wind = Vector(0.1, 0)

        # Label teks untuk menampilkan perhitungan
        self.info_text = tk.Label(root, font=("Courier", 12), justify="left", bg="white", anchor="w")
        self.info_text.pack(fill="both")

        self.update()

    def update(self):
        # === Terapkan gaya angin ke dua benda ===
        self.mover1.applyForce(self.wind)
        self.mover2.applyForce(self.wind)

        # === Update posisi mereka ===
        self.mover1.update()
        self.mover2.update()

        # === Bersihkan dan gambar ulang layar ===
        self.canvas.delete("all")

        # Gambar bola biru (massa kecil)
        x1, y1, r1 = self.mover1.location.x, self.mover1.location.y, self.mover1.radius
        self.canvas.create_oval(x1 - r1, y1 - r1, x1 + r1, y1 + r1, fill="blue")

        # Gambar bola merah (massa besar)
        x2, y2, r2 = self.mover2.location.x, self.mover2.location.y, self.mover2.radius
        self.canvas.create_oval(x2 - r2, y2 - r2, x2 + r2, y2 + r2, fill="red")

        # === Tampilkan perhitungan di bawah canvas ===
        acc1 = self.wind.copy()
        acc1.div(self.mover1.mass)
        acc2 = self.wind.copy()
        acc2.div(self.mover2.mass)

        info = (
            f"Gaya angin (F): ({self.wind.x:.2f}, {self.wind.y:.2f})\n"
            f"Mover Biru (massa={self.mover1.mass}): a = F/m = {acc1.x:.2f}\n"
            f"Mover Merah (massa={self.mover2.mass}): a = F/m = {acc2.x:.2f}\n\n"
            f"Kecepatan Biru: v = ({self.mover1.velocity.x:.2f}, {self.mover1.velocity.y:.2f})\n"
            f"Kecepatan Merah: v = ({self.mover2.velocity.x:.2f}, {self.mover2.velocity.y:.2f})"
        )
        self.info_text.config(text=info)

        # Lanjutkan ke frame berikutnya
        self.root.after(20, self.update)

# === JALANKAN PROGRAM ===
if __name__ == "__main__":
    root = tk.Tk()
    root.title("Simulasi Gaya & Massa - Newton II")
    app = Simulation(root)
    root.mainloop()
```

```python
#SCRIPT51
import tkinter as tk
import math

# Kelas Mover untuk objek yang bisa bergerak karena gaya
class Mover:
    def __init__(self, canvas, x, y, mass):
        self.canvas = canvas
        self.location = [x, y]         # Posisi awal (x, y)
        self.velocity = [0, 0]         # Kecepatan awal (vx, vy)
        self.acceleration = [0, 0]     # Percepatan awal (ax, ay)
        self.mass = mass               # Massa benda
        self.radius = mass * 2         # Ukuran bola proporsional dengan massa (biar kelihatan)
        self.id = canvas.create_oval(  # Gambar lingkaran pertama kali
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius,
            fill="blue"
        )
        self.text_id = canvas.create_text(  # Tambah teks info gaya dll
            x, y + self.radius + 20,
            text="",
            font=("Arial", 10)
        )
        self.force_text_id = canvas.create_text(
            x, y + self.radius + 40,
            text="",
            font=("Arial", 10)
        )

        # Simpan gaya terakhir
        self.last_force = [0, 0]

    # Terapkan gaya â†’ Rumus: a = F / m
    def apply_force(self, force):
        # Gaya (force) = [Fx, Fy]
        # Percepatan = gaya / massa
        self.acceleration[0] += force[0] / self.mass
        self.acceleration[1] += force[1] / self.mass
        self.last_force = force  # Simpan gaya terakhir

    # Update posisi berdasarkan hukum gerak:
    def update(self):
        # Kecepatan baru = kecepatan + percepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        # Posisi baru = posisi + kecepatan
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        # Reset percepatan (karena gaya baru akan diterapkan di frame selanjutnya)
        self.acceleration = [0, 0]

        # Gambar ulang lingkaran
        self.canvas.coords(
            self.id,
            self.location[0] - self.radius,
            self.location[1] - self.radius,
            self.location[0] + self.radius,
            self.location[1] + self.radius,
        )

        # Tampilkan informasi gaya, massa, percepatan, dan kecepatan
        info = f"""
Massa (m) = {self.mass}
Kecepatan (vx, vy) = ({self.velocity[0]:.2f}, {self.velocity[1]:.2f})
"""
        self.canvas.coords(self.text_id, self.location[0], self.location[1] + self.radius + 20)
        self.canvas.itemconfig(self.text_id, text=info)

        # Info gaya: gaya terakhir yang diterapkan
        force_info = f"Gaya terakhir (Fx, Fy) = ({self.last_force[0]:.2f}, {self.last_force[1]:.2f})"
        self.canvas.coords(self.force_text_id, self.location[0], self.location[1] + self.radius + 40)
        self.canvas.itemconfig(self.force_text_id, text=force_info)

    # Batas layar supaya bola mantul jika kena tepi
    def check_edges(self, width, height):
        if self.location[0] > width - self.radius:
            self.location[0] = width - self.radius
            self.velocity[0] *= -0.8  # Pantulan kiri-kanan (hilang energi)
        elif self.location[0] < self.radius:
            self.location[0] = self.radius
            self.velocity[0] *= -0.8

        if self.location[1] > height - self.radius:
            self.location[1] = height - self.radius
            self.velocity[1] *= -0.8  # Pantulan atas-bawah
        elif self.location[1] < self.radius:
            self.location[1] = self.radius
            self.velocity[1] *= -0.8

# Kelas utama untuk tampilan simulasi
class NewtonSecondLawDemo:
    def __init__(self, root):
        self.root = root
        self.root.title("Hukum Kedua Newton - F = mÃ—a")

        # Ukuran canvas
        self.width = 600
        self.height = 400
        self.canvas = tk.Canvas(root, width=self.width, height=self.height, bg="white")
        self.canvas.pack()

        # Judul dan penjelasan hukum Newton
        explanation = """HUKUM KEDUA NEWTON (F = mÃ—a):
Gaya (F) menyebabkan percepatan (a), 
tapi massa (m) membuat percepatan jadi lebih kecil."""
        self.canvas.create_text(self.width // 2, 20, text=explanation, font=("Arial", 12))

        # Buat satu objek Mover dengan massa 10
        self.mover = Mover(self.canvas, 100, 100, 10)

        # Tombol kontrol
        self.wind_btn = tk.Button(root, text="Angin (â†’)", command=self.apply_wind)
        self.wind_btn.pack(side="left")
        self.gravity_btn = tk.Button(root, text="Gravitasi (â†“)", command=self.apply_gravity)
        self.gravity_btn.pack(side="left")
        self.reset_btn = tk.Button(root, text="Reset", command=self.reset)
        self.reset_btn.pack(side="left")

        # Jalankan loop animasi
        self.root.after(50, self.update)

    # Fungsi untuk menerapkan gaya angin ke kanan
    def apply_wind(self):
        force = [0.5, 0]  # Gaya angin arah kanan
        self.mover.apply_force(force)

    # Fungsi untuk menerapkan gaya gravitasi ke bawah
    def apply_gravity(self):
        force = [0, 0.3]  # Gaya gravitasi ke bawah
        self.mover.apply_force(force)

    # Reset posisi dan kecepatan benda
    def reset(self):
        self.mover.location = [100, 100]
        self.mover.velocity = [0, 0]
        self.mover.acceleration = [0, 0]
        self.mover.last_force = [0, 0]  # Reset gaya juga

    # Fungsi update animasi
    def update(self):
        self.mover.update()
        self.mover.check_edges(self.width, self.height)
        self.root.after(50, self.update)

# Jalankan program
if __name__ == "__main__":
    root = tk.Tk()
    app = NewtonSecondLawDemo(root)
    root.mainloop()
```

```python
#SCRIPT60
import tkinter as tk
import math

# ======== Class Vector 2D untuk Representasi Gaya & Posisi ========
class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, v):
        # Menambahkan vektor lain ke vektor ini
        self.x += v.x
        self.y += v.y

    def sub(self, v):
        # Mengurangi vektor lain dari vektor ini
        self.x -= v.x
        self.y -= v.y

    def mult(self, n):
        # Mengalikan komponen vektor dengan skalar
        self.x *= n
        self.y *= n

    def div(self, n):
        # Membagi komponen vektor dengan skalar (hindari pembagi nol)
        if n != 0:
            self.x /= n
            self.y /= n

    def mag(self):
        # Mengembalikan panjang (magnitude) dari vektor
        return math.sqrt(self.x ** 2 + self.y ** 2)

    def normalize(self):
        # Menjadikan vektor ini sebagai unit vector (panjang = 1)
        m = self.mag()
        if m != 0:
            self.div(m)

    def copy(self):
        # Mengembalikan salinan vektor
        return Vector(self.x, self.y)

    @staticmethod
    def sub_static(v1, v2):
        # Mengembalikan hasil pengurangan vektor secara static
        return Vector(v1.x - v2.x, v1.y - v2.y)

    @staticmethod
    def mult_static(v, n):
        # Mengembalikan hasil perkalian skalar dengan vektor
        return Vector(v.x * n, v.y * n)

    @staticmethod
    def div_static(v, n):
        # Mengembalikan hasil pembagian skalar dengan vektor
        if n != 0:
            return Vector(v.x / n, v.y / n)
        return Vector()

    def __str__(self):
        # Untuk menampilkan vektor dalam format string
        return f"({self.x:.2f}, {self.y:.2f})"

# ======== Class Mover: Benda yang Diberi Gaya ========
class Mover:
    def __init__(self, x, y, mass):
        self.mass = mass                  # Massa benda
        self.radius = mass * 8            # Radius bola disesuaikan massa
        self.position = Vector(x, y)      # Posisi awal
        self.velocity = Vector(0, 0)      # Kecepatan awal
        self.acceleration = Vector(0, 0)  # Percepatan awal

    def apply_force(self, force):
        # F = m * a  â†’  a = F / m
        f = Vector.div_static(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        # Update kecepatan dan posisi
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration = Vector(0, 0)  # Reset percepatan setelah tiap frame

    def contact_edge(self, height):
        # Apakah bola menyentuh lantai? (cek berdasarkan radius)
        return self.position.y + self.radius >= height

    def bounce_edges(self, width, height):
        # Memantulkan bola dari pinggir layar (seperti tumbukan tidak elastis)
        bounce = -0.9
        if self.position.x > width - self.radius:
            self.position.x = width - self.radius
            self.velocity.x *= bounce
        elif self.position.x < self.radius:
            self.position.x = self.radius
            self.velocity.x *= bounce

        if self.position.y > height - self.radius:
            self.position.y = height - self.radius
            self.velocity.y *= bounce

    def show(self, canvas):
        # Gambar bola di canvas
        canvas.create_oval(
            self.position.x - self.radius,
            self.position.y - self.radius,
            self.position.x + self.radius,
            self.position.y + self.radius,
            fill="skyblue", outline="black", width=2
        )

# ======== Setup Simulasi Tkinter ========
WIDTH, HEIGHT = 640, 400
root = tk.Tk()
root.title("Simulasi Gaya Gesek - The Nature of Code")
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()
text = tk.Label(root, text="", font=("Courier", 10), justify="left", anchor="nw")
text.pack()

mover = Mover(WIDTH // 2, 30, 5)  # Bola dengan massa 5
mouse_pressed = [False]  # Flag mouse ditekan

# ======== Fungsi Simulasi Frame-by-Frame ========
def draw():
    canvas.delete("all")  # Hapus frame sebelumnya

    # 1. Buat gaya gravitasi
    gravity = Vector(0, 1)
    mover.apply_force(gravity)

    result = f"Gaya:\n  Gravitasi : {gravity}"

    # 2. Tambahkan gaya angin jika mouse ditekan
    if mouse_pressed[0]:
        wind = Vector(0.2, 0)
        mover.apply_force(wind)
        result += f"\n  Angin     : {wind}"
    else:
        wind = Vector(0, 0)

    # 3. Jika bola menyentuh lantai, terapkan gaya gesek
    if mover.contact_edge(HEIGHT):
        c = 0.1  # Koefisien gesek
        friction = mover.velocity.copy()  # Ambil arah gerak
        friction.mult(-1)                 # Balik arah vektor
        friction.normalize()              # Jadikan unit vector
        friction.mult(c)                  # Skala dengan koefisien
        mover.apply_force(friction)
        result += f"\n  Gesekan   : {friction}"

    # 4. Update gerakan dan tampilkan
    mover.update()
    mover.bounce_edges(WIDTH, HEIGHT)
    mover.show(canvas)

    # 5. Tampilkan informasi di layar
    result += f"\n\nPercepatan : {mover.acceleration}"
    result += f"\nKecepatan  : {mover.velocity}"
    result += f"\nPosisi     : {mover.position}"
    text.config(text=result)

    root.after(30, draw)  # Ulangi setiap 30ms

# ======== Fungsi Interaksi Mouse ========
def on_press(event): mouse_pressed[0] = True
def on_release(event): mouse_pressed[0] = False
canvas.bind("<ButtonPress-1>", on_press)
canvas.bind("<ButtonRelease-1>", on_release)

# ======== Jalankan Simulasi ========
draw()
root.mainloop()
```

```python
#SCRIPT61
import tkinter as tk       # Untuk membuat tampilan grafis
import random              # Untuk membuat angka acak
import math                # Untuk perhitungan matematika

# === Kelas Vector (Mewakili Arah dan Besaran) ===
class Vector:
    def __init__(self, x, y):
        self.x = x  # Posisi horizontal
        self.y = y  # Posisi vertikal
    
    # Fungsi untuk menambahkan vektor lain
    def add(self, v):
        self.x += v.x
        self.y += v.y
    
    # Fungsi untuk mengalikan vektor dengan angka
    def multiply(self, n):
        self.x *= n
        self.y *= n
    
    # Fungsi untuk membuat salinan vektor
    def copy(self):
        return Vector(self.x, self.y)
    
    # Fungsi untuk membuat vektor menjadi panjang 1 (normalisasi)
    def normalize(self):
        panjang = self.magnitude()
        if panjang != 0:
            self.x /= panjang
            self.y /= panjang
    
    # Fungsi untuk menghitung panjang vektor
    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2)

# === Kelas Mover (Benda yang Bergerak) ===
class Mover:
    def __init__(self, massa, x, y):
        self.mass = massa              # Massa benda
        self.position = Vector(x, y)   # Posisi awal
        self.velocity = Vector(0, 0)   # Kecepatan awal (diam)
        self.acceleration = Vector(0, 0) # Percepatan awal
    
    # Memberikan gaya ke benda (F = m Ã— a)
    def apply_force(self, force):
        f = force.copy()       # Salin gaya agar tidak merusak aslinya
        f.multiply(1/self.mass) # Hitung percepatan (a = F/m)
        self.acceleration.add(f)
    
    # Memperbarui posisi benda
    def update(self):
        self.velocity.add(self.acceleration)  # Update kecepatan
        self.position.add(self.velocity)      # Update posisi
        self.acceleration = Vector(0, 0)      # Reset percepatan
    
    # Menggambar benda di layar
    def display(self, canvas):
        radius = self.mass * 8  # Ukuran bola berdasarkan massa
        canvas.create_oval(
            self.position.x - radius, self.position.y - radius,
            self.position.x + radius, self.position.y + radius,
            fill="red"
        )
    
    # Mengecek jika benda menabrak tepi layar
    def check_edges(self, width, height):
        if self.position.x > width:    # Tabrak kanan
            self.position.x = width
            self.velocity.x *= -1      # Memantul
        elif self.position.x < 0:      # Tabrak kiri
            self.position.x = 0
            self.velocity.x *= -1
            
        if self.position.y > height:   # Tabrak bawah
            self.position.y = height
            self.velocity.y *= -1
        elif self.position.y < 0:      # Tabrak atas
            self.position.y = 0
            self.velocity.y *= -1

# === Kelas Area Gesekan ===
class FrictionArea:
    def __init__(self, x, y, w, h, mu, warna):
        self.x = x      # Posisi x area
        self.y = y      # Posisi y area
        self.w = w      # Lebar area
        self.h = h      # Tinggi area
        self.mu = mu    # Koefisien gesekan
        self.color = warna  # Warna area
    
    # Menggambar area di layar
    def draw(self, canvas):
        canvas.create_rectangle(
            self.x, self.y, 
            self.x + self.w, self.y + self.h,
            fill=self.color, outline=""
        )
    
    # Mengecek apakah benda ada di area ini
    def contains(self, mover):
        return (self.x <= mover.position.x <= self.x + self.w and
                self.y <= mover.position.y <= self.y + self.h)

# === Persiapan Program Utama ===
# Ukuran layar
WIDTH, HEIGHT = 800, 400

# Membuat jendela
root = tk.Tk()
root.title("Simulasi Fisika - Area Gesekan")
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

# Membuat 5 benda dengan posisi acak
movers = [Mover(random.uniform(1, 3), 50, 50 + i*40) for i in range(5)]

# Membuat 3 area gesekan dengan sifat berbeda
pockets = [
    FrictionArea(100, 200, 200, 100, 0.05, "lightblue"),  # Gesekan kecil
    FrictionArea(350, 250, 150, 100, 0.1, "blue"),        # Gesekan besar
    FrictionArea(550, 150, 150, 150, -0.02, "lightgreen") # Mempercepat
]

# === Fungsi Utama untuk Animasi ===
def draw():
    canvas.delete("all")  # Bersihkan layar setiap frame
    
    # Gambar semua area gesekan
    for pocket in pockets:
        pocket.draw(canvas)
    
    # Proses setiap benda
    for mover in movers:
        # Buat gaya gravitasi (ke bawah) dan angin (ke kanan)
        gravity = Vector(0, 0.1 * mover.mass)
        wind = Vector(0.01, 0)
        
        # Terapkan gaya ke benda
        mover.apply_force(gravity)
        mover.apply_force(wind)
        
        # Cek jika benda ada di area gesekan
        for pocket in pockets:
            if pocket.contains(mover):
                friction = mover.velocity.copy()  # Salin kecepatan
                friction.normalize()              # Buat panjangnya 1
                friction.multiply(-1)             # Balik arah
                friction.multiply(pocket.mu)      # Kalikan dengan koefisien
                mover.apply_force(friction)       # Terapkan gaya gesek
        
        # Update posisi benda dan cek tabrakan
        mover.update()
        mover.check_edges(WIDTH, HEIGHT)
        mover.display(canvas)
    
    # Ulangi setiap 16 milidetik (~60 frame per detik)
    root.after(16, draw)

# === Menjalankan Program ===
draw()              # Mulai animasi
root.mainloop()     # Jalankan program
```
