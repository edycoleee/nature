```python
import tkinter as tk
import math
from vector import Vector 

WIDTH, HEIGHT = 600, 400

class Vehicle:
    def __init__(self, x, y, canvas):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)  # awal ke kanan
        self.acceleration = Vector(0, 0)
        self.maxspeed = 2
        self.maxforce = 0.05
        self.r = 8
        self.trail = []

        # Segitiga vehicle
        self.shape = self.canvas.create_polygon(0, 0, 0, 0, 0, 0, fill='gray', outline='black')

    def seek(self, target):
        desired = target.subbed(self.position)
        distance = desired.magnitude()
        desired_unit = desired.normalized()
        desired_velocity = desired_unit.multed(self.maxspeed)

        # Steering = desired - velocity
        steering = desired_velocity.subbed(self.velocity)
        steering.limit(self.maxforce)

        # Apply steering force
        self.acceleration = steering

        angle = math.atan2(desired.y, desired.x)
        return desired_velocity, distance, angle, desired_unit

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)

        self.trail.append(self.position.copy())
        if len(self.trail) > 100:
            self.trail.pop(0)

    def show(self):
        # Draw trail
        if len(self.trail) > 1:
            coords = []
            for vec in self.trail:
                coords.extend([vec.x, vec.y])
            self.canvas.create_line(*coords, fill="#888", width=1, smooth=True)

        # Hitung orientasi segitiga berdasarkan arah kecepatan
        angle = self.velocity.heading()
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r * 2)
        left = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r * 2)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r * 2)

        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        self.canvas.coords(self.shape, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)

class App:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.vehicle = Vehicle(100, 100, self.canvas)

        # Waypoints
        self.waypoints = [Vector(250, 300), Vector(250, 100), Vector(500, 150), Vector(100, 350), Vector(400, 350)]
        self.waypoint_index = 0
        self.target = self.waypoints[self.waypoint_index]
        self.target_shape = self.canvas.create_oval(self.target.x - 10, self.target.y - 10,
                                                    self.target.x + 10, self.target.y + 10, fill="green")

        self.text = self.canvas.create_text(10, 10, anchor="nw", font=("Courier", 10), fill="black")

        self.animate()

    def animate(self):
        v = self.vehicle
        desired_velocity, distance, angle, direction_unit = v.seek(self.target)

        if distance > 5:
            v.update()
        else:
            # Pindah ke waypoint berikutnya
            self.waypoint_index = (self.waypoint_index + 1) % len(self.waypoints)
            self.target = self.waypoints[self.waypoint_index]
            self.canvas.coords(self.target_shape, self.target.x - 10, self.target.y - 10,
                                                self.target.x + 10, self.target.y + 10)

        # Tampilkan vehicle
        v.show()

        # Info debug
        p = v.position
        dx = self.target.x - p.x
        dy = self.target.y - p.y
        info = f"""ðŸŽ¯ Vehicle Steering Seek

Waypoint ke-{self.waypoint_index + 1}: ({self.target.x}, {self.target.y})
Posisi: ({p.x:.1f}, {p.y:.1f})

Î”x = {dx:.1f}
Î”y = {dy:.1f}

Sudut ke target (Â°) = {math.degrees(angle):.1f}Â°
Velocity    = {v.velocity}
Acceleration = {v.acceleration}
"""
        self.canvas.itemconfig(self.text, text=info)

        self.canvas.after(30, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("ðŸŽ¯ Vehicle Steering Seek dengan Belok Halus")
    app = App(root)
    root.mainloop()
```

```python
# NOC 5.3 Pursue - The Coding Train
import tkinter as tk
import random
import math
from vector import Vector

WIDTH, HEIGHT = 640, 480

class Vehicle:
    def __init__(self, canvas, x, y, color="gray"):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(random.uniform(-1, 1), random.uniform(-1, 1))
        self.acceleration = Vector(0, 0)
        self.maxspeed = 4
        self.maxforce = 0.1
        self.r = 12
        self.color = color

        # Segitiga vehicle
        self.shape = self.canvas.create_polygon(0, 0, 0, 0, 0, 0,
                                                fill=self.color, outline="black")

    def apply_force(self, force):
        self.acceleration.add(force)

    def pursue(self, target):
        target_pos = target.position.copy()
        prediction = target.velocity.copy().multed(10)  # prediksi posisi masa depan
        future_pos = target_pos.added(prediction)

        # Visual titik prediksi
        r = 5
        if hasattr(self, "future_dot"):
            self.canvas.coords(self.future_dot, future_pos.x - r, future_pos.y - r,
                               future_pos.x + r, future_pos.y + r)
        else:
            self.future_dot = self.canvas.create_oval(future_pos.x - r, future_pos.y - r,
                                                      future_pos.x + r, future_pos.y + r,
                                                      fill="green")

        return self.seek(future_pos)

    def seek(self, target_pos):
        desired = target_pos.subbed(self.position)
        desired.set_mag(self.maxspeed)
        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)
        return steer

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)
        self.edges()

    def edges(self):
        if self.position.x > WIDTH + self.r:
            self.position.x = -self.r
        elif self.position.x < -self.r:
            self.position.x = WIDTH + self.r

        if self.position.y > HEIGHT + self.r:
            self.position.y = -self.r
        elif self.position.y < -self.r:
            self.position.y = HEIGHT + self.r

    def show(self):
        angle = self.velocity.heading()
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r)
        left = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r)

        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        self.canvas.coords(self.shape, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)

class Target(Vehicle):
    def __init__(self, canvas, x, y):
        super().__init__(canvas, x, y, color="#F063A4")
        self.velocity = Vector(random.uniform(-2, 2), random.uniform(-2, 2))
        # Ganti menjadi lingkaran
        self.canvas.delete(self.shape)
        self.shape = self.canvas.create_oval(x - self.r, y - self.r, x + self.r, y + self.r,
                                             fill="#F063A4", outline="black")

    def show(self):
        p = self.position
        self.canvas.coords(self.shape, p.x - self.r, p.y - self.r, p.x + self.r, p.y + self.r)

class App:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.pursuer = Vehicle(self.canvas, 100, 100)
        self.target = Target(self.canvas, 300, 200)

        self.text = self.canvas.create_text(10, 10, anchor="nw", font=("Courier", 10), fill="black")

        self.animate()

    def animate(self):
        steer = self.pursuer.pursue(self.target)
        self.pursuer.apply_force(steer)

        d = self.pursuer.position.dist(self.target.position)
        if d < self.pursuer.r + self.target.r:
            # Target tertangkap, buat target baru dan reset posisi pursuer
            self.canvas.delete(self.target.shape)
            self.target = Target(self.canvas, random.randint(50, WIDTH - 50), random.randint(50, HEIGHT - 50))
            self.pursuer.position = Vector(WIDTH / 2, HEIGHT / 2)
            self.pursuer.velocity = Vector(random.uniform(-1, 1), random.uniform(-1, 1))

        self.pursuer.update()
        self.target.update()

        self.pursuer.show()
        self.target.show()

        # Info debug
        info = f"""ðŸŽ¯ Pursue Target

Pursuer Pos: ({self.pursuer.position.x:.1f}, {self.pursuer.position.y:.1f})
Target Pos : ({self.target.position.x:.1f}, {self.target.position.y:.1f})
Jarak      : {d:.1f}
"""
        self.canvas.itemconfig(self.text, text=info)

        self.canvas.after(30, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("ðŸŽ¯ NOC 5.3 Pursue - The Coding Train")
    app = App(root)
    root.mainloop()
```

```python
# NOC 5.02 - Arrive Behavior
import tkinter as tk
import math
from vector import Vector

WIDTH, HEIGHT = 640, 240

class Vehicle:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(0, -2)
        self.acceleration = Vector(0, 0)
        self.r = 6
        self.maxspeed = 4
        self.maxforce = 0.1
        self.id = canvas.create_polygon(0, 0, 0, 0, 0, 0, fill='gray', outline='black')

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def apply_force(self, force):
        self.acceleration.add(force)

    def arrive(self, target):
        desired = target.subbed(self.position)
        d = desired.mag()
        if d < 100:
            m = self._map(d, 0, 100, 0, self.maxspeed)
            desired.set_mag(m)
        else:
            desired.set_mag(self.maxspeed)

        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)
        self.apply_force(steer)

    def show(self):
        angle = math.atan2(self.velocity.y, self.velocity.x)
        x, y = self.position.x, self.position.y
        def rotate_point(px, py, angle):
            cos_a = math.cos(angle)
            sin_a = math.sin(angle)
            return (px * cos_a - py * sin_a, px * sin_a + py * cos_a)

        p1 = rotate_point(self.r * 2, 0, angle)
        p2 = rotate_point(-self.r * 2, -self.r, angle)
        p3 = rotate_point(-self.r * 2, self.r, angle)

        coords = [x + p1[0], y + p1[1], x + p2[0], y + p2[1], x + p3[0], y + p3[1]]
        self.canvas.coords(self.id, *coords)

    def _map(self, value, start1, stop1, start2, stop2):
        return start2 + (stop2 - start2) * ((value - start1) / (stop1 - start1))

def update():
    canvas.delete("info")
    canvas.delete("mouse")

    mouse_x = app.winfo_pointerx() - app.winfo_rootx()
    mouse_y = app.winfo_pointery() - app.winfo_rooty()
    mouse = Vector(mouse_x, mouse_y)

    # Gambar lingkaran di mouse pointer
    canvas.create_oval(mouse_x - 20, mouse_y - 20, mouse_x + 20, mouse_y + 20, outline="black", tags="mouse")

    vehicle.arrive(mouse)
    vehicle.update()
    vehicle.show()

    info_text = (
        f"pos: ({vehicle.position.x:.1f}, {vehicle.position.y:.1f})\n"
        f"vel: ({vehicle.velocity.x:.1f}, {vehicle.velocity.y:.1f})\n"
        f"acc: ({vehicle.acceleration.x:.2f}, {vehicle.acceleration.y:.2f})\n"
        f"maxspeed: {vehicle.maxspeed:.2f}  maxforce: {vehicle.maxforce:.2f}"
    )
    canvas.create_text(10, 10, anchor='nw', text=info_text, fill="black", font=("Consolas", 10), tags="info")
    app.after(16, update)  # sekitar 60fps

app = tk.Tk()
app.title("NOC 5.02 - Arrive Behavior")
canvas = tk.Canvas(app, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

vehicle = Vehicle(canvas, WIDTH / 2, HEIGHT / 2)

update()
app.mainloop()
```

```python
import tkinter as tk
import math
import random
from typing import Optional

class Vector:
    def __init__(self, x: float = 0, y: float = 0):
        self.x = x
        self.y = y
    
    def add(self, other: 'Vector') -> 'Vector':
        self.x += other.x
        self.y += other.y
        return self
    
    def sub(self, other: 'Vector') -> 'Vector':
        self.x -= other.x
        self.y -= other.y
        return self
    
    def mult(self, scalar: float) -> 'Vector':
        self.x *= scalar
        self.y *= scalar
        return self
    
    def normalize(self) -> 'Vector':
        mag = self.magnitude()
        if mag != 0:
            self.mult(1/mag)
        return self
    
    def limit(self, max_value: float) -> 'Vector':
        mag = self.magnitude()
        if mag > max_value:
            self.mult(max_value / mag)
        return self
    
    def magnitude(self) -> float:
        return math.sqrt(self.x**2 + self.y**2)
    
    def heading(self) -> float:
        return math.atan2(self.y, self.x)
    
    def copy(self) -> 'Vector':
        return Vector(self.x, self.y)
    
    @staticmethod
    def add_vectors(v1: 'Vector', v2: 'Vector') -> 'Vector':
        return Vector(v1.x + v2.x, v1.y + v2.y)
    
    @staticmethod
    def sub_vectors(v1: 'Vector', v2: 'Vector') -> 'Vector':
        return Vector(v1.x - v2.x, v1.y - v2.y)

class Vehicle:
    def __init__(self, x: float, y: float):
        self.acceleration = Vector(0, 0)
        self.velocity = Vector(0, -2)  # Initial slight upward velocity
        self.position = Vector(x, y)
        self.r = 6
        self.wander_theta = 0.0
        self.max_speed = 2
        self.max_force = 0.05
        self.triangle = None
        self.debug_lines = []
        
    def run(self, canvas: tk.Canvas, debug: bool) -> None:
        self.update()
        self.borders(canvas)
        self.show(canvas)
        
    def update(self) -> None:
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.max_speed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)
        
    def wander(self, canvas: tk.Canvas, debug: bool) -> None:
        wander_radius = 25
        wander_distance = 80
        change = 0.3
        self.wander_theta += random.uniform(-change, change)
        
        circle_pos = self.velocity.copy()
        circle_pos.normalize()
        circle_pos.mult(wander_distance)
        circle_pos.add(self.position)
        
        h = self.velocity.heading()
        
        circle_offset = Vector(
            wander_radius * math.cos(self.wander_theta + h),
            wander_radius * math.sin(self.wander_theta + h)
        )
        target = Vector.add_vectors(circle_pos, circle_offset)
        self.seek(target)
        
        if debug:
            self.draw_wander_stuff(canvas, self.position, circle_pos, target, wander_radius)
    
    def apply_force(self, force: Vector) -> None:
        self.acceleration.add(force)
    
    def seek(self, target: Vector) -> None:
        desired = Vector.sub_vectors(target, self.position)
        desired.normalize()
        desired.mult(self.max_speed)
        
        steer = Vector.sub_vectors(desired, self.velocity)
        steer.limit(self.max_force)
        self.apply_force(steer)
    
    def show(self, canvas: tk.Canvas) -> None:
        theta = self.velocity.heading() + math.pi / 2
        x1 = self.position.x + math.cos(theta) * 0 * self.r
        y1 = self.position.y + math.sin(theta) * -2 * self.r
        x2 = self.position.x + math.cos(theta + 2*math.pi/3) * self.r
        y2 = self.position.y + math.sin(theta + 2*math.pi/3) * 2 * self.r
        x3 = self.position.x + math.cos(theta + 4*math.pi/3) * self.r
        y3 = self.position.y + math.sin(theta + 4*math.pi/3) * 2 * self.r
        
        if self.triangle is None:
            self.triangle = canvas.create_polygon(
                [x1, y1, x2, y2, x3, y3], 
                fill='gray', 
                outline='black', 
                width=2
            )
        else:
            canvas.coords(self.triangle, x1, y1, x2, y2, x3, y3)
    
    def borders(self, canvas: tk.Canvas) -> None:
        width = canvas.winfo_width()
        height = canvas.winfo_height()
        
        if self.position.x < -self.r:
            self.position.x = width + self.r
        if self.position.y < -self.r:
            self.position.y = height + self.r
        if self.position.x > width + self.r:
            self.position.x = -self.r
        if self.position.y > height + self.r:
            self.position.y = -self.r
    
    def draw_wander_stuff(self, canvas: tk.Canvas, location: Vector, circle_pos: Vector, 
                          target: Vector, rad: float) -> None:
        # Clear previous debug lines
        for line in self.debug_lines:
            canvas.delete(line)
        self.debug_lines = []
        
        # Draw wander circle
        self.debug_lines.append(canvas.create_oval(
            circle_pos.x - rad, circle_pos.y - rad,
            circle_pos.x + rad, circle_pos.y + rad,
            outline='black', width=1
        ))
        
        # Draw target
        self.debug_lines.append(canvas.create_oval(
            target.x - 2, target.y - 2,
            target.x + 2, target.y + 2,
            outline='black', fill='black'
        ))
        
        # Draw line from vehicle to circle
        self.debug_lines.append(canvas.create_line(
            location.x, location.y,
            circle_pos.x, circle_pos.y,
            fill='black', width=1
        ))
        
        # Draw line from circle to target
        self.debug_lines.append(canvas.create_line(
            circle_pos.x, circle_pos.y,
            target.x, target.y,
            fill='black', width=1
        ))

class App:
    def __init__(self, root: tk.Tk):
        self.root = root
        self.root.title("Wander Behavior")
        
        self.width = 640
        self.height = 240
        
        self.canvas = tk.Canvas(root, width=self.width, height=self.height, bg='white')
        self.canvas.pack()
        
        self.debug = True
        self.vehicle = Vehicle(self.width / 2, self.height / 2)
        
        self.canvas.bind("<Button-1>", self.toggle_debug)
        self.root.after(0, self.update)
    
    def toggle_debug(self, event: tk.Event) -> None:
        self.debug = not self.debug
    
    def update(self) -> None:
        self.canvas.delete("all")
        self.vehicle.wander(self.canvas, self.debug)
        self.vehicle.run(self.canvas, self.debug)
        self.root.after(30, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
# Stay Within Walls Steering Behavior
import tkinter as tk
import math
import random
from vector import Vector

WIDTH, HEIGHT = 400, 300

class Vehicle:
    def __init__(self, x, y, canvas):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(random.uniform(-2, 2), random.uniform(-2, 2))
        self.acceleration = Vector(0, 0)
        self.r = 12
        self.max_speed = 8
        self.max_force = 0.2

        self.triangle = None
        self.info_text = self.canvas.create_text(10, 10, anchor="nw", font=("Courier", 12), fill="black")
        self.canvas.create_rectangle(50, 50, 350, 250, outline="blue")
    def apply_force(self, force):
        self.acceleration.add(force)

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.max_speed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def edges_wrap(self):
        """Jika ingin wrapping layar."""
        if self.position.x < -self.r:
            self.position.x = WIDTH + self.r
        if self.position.y < -self.r:
            self.position.y = HEIGHT + self.r
        if self.position.x > WIDTH + self.r:
            self.position.x = -self.r
        if self.position.y > HEIGHT + self.r:
            self.position.y = -self.r

    def stay_within_walls(self, offset=25):
        desired = None

        if self.position.x < offset:
            desired = Vector(self.max_speed, self.velocity.y)
        elif self.position.x > WIDTH - offset:
            desired = Vector(-self.max_speed, self.velocity.y)

        if self.position.y < offset:
            desired = Vector(self.velocity.x, self.max_speed)
        elif self.position.y > HEIGHT - offset:
            desired = Vector(self.velocity.x, -self.max_speed)

        if desired:
            desired.normalize()
            desired.mult(self.max_speed)
            steer = desired.subbed(self.velocity)
            steer.limit(self.max_force)
            self.apply_force(steer)

    def show(self):
        angle = self.velocity.heading()
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r)
        left = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r)

        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        if self.triangle is None:
            self.triangle = self.canvas.create_polygon(
                p1.x, p1.y, p2.x, p2.y, p3.x, p3.y,
                fill="gray", outline="black", width=2
            )
        else:
            self.canvas.coords(self.triangle, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)

        info = (
            f"Posisi: ({self.position.x:.1f}, {self.position.y:.1f})\n"
            f"Kecepatan: ({self.velocity.x:.1f}, {self.velocity.y:.1f})\n"
            f"Heading: {math.degrees(angle):.1f}Â°"
        )
        self.canvas.itemconfig(self.info_text, text=info)

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Stay Within Walls Steering Behavior")
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.vehicle = Vehicle(WIDTH / 2, HEIGHT / 2, self.canvas)
        
        self.animate()

    def animate(self):
        self.vehicle.stay_within_walls(offset=50)
        self.vehicle.update()
        self.vehicle.show()
        self.vehicle.edges_wrap()

        self.root.after(30, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
# Angle Between Two Vectors
import tkinter as tk
import math
from vector import Vector

WIDTH, HEIGHT = 640, 480

class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()
        self.update()
    
    def draw_arrow(self, start, vec, color="black"):
        end = start.added(vec)
        self.canvas.create_line(start.x, start.y, end.x, end.y, arrow=tk.LAST, width=2, fill=color)
    
    def update(self):
        self.canvas.delete("all")
        
        # Vector dari tengah ke mouse
        mouse_pos = Vector(self.canvas.winfo_pointerx() - self.canvas.winfo_rootx(),
                           self.canvas.winfo_pointery() - self.canvas.winfo_rooty())
        center_pos = Vector(WIDTH/2, HEIGHT/2)
        
        v = mouse_pos.subbed(center_pos)
        if v.mag() == 0:
            v = Vector(1, 0)
        v_norm = v.copy().normalized()
        v_draw = v_norm.copy().mult(100)

        # Vektor sumbu X
        xaxis = Vector(100, 0)

        # Hitung Dot Product, Magnitude, dan Theta
        dot = v_norm.x * 1 + v_norm.y * 0  # karena xaxis adalah (1,0)
        mag_v = v_norm.mag()
        mag_xaxis = 1  # karena (1,0)
        cos_theta = dot / (mag_v * mag_xaxis)
        theta = math.acos(cos_theta)
        deg_theta = math.degrees(theta)

        # Gambar vektor
        self.draw_arrow(center_pos, v_draw, color="red")
        self.draw_arrow(center_pos, xaxis, color="blue")

        # Tampilkan rumus dan perhitungan
        text = (
            f"Dot Product: {dot:.2f} = {v_norm.x:.2f} * 1 + {v_norm.y:.2f} * 0 \n"
            f"x = (1,0) y={v_norm} > ||v||: {mag_v:.2f}, ||xaxis||: {mag_xaxis:.2f}\n"
            f"cos(theta) = dot / (||v|| * ||xaxis||) = {cos_theta:.2f}\n"
            f"theta = arccos({cos_theta:.2f}) = {theta:.2f} rad = {deg_theta:.2f} deg"
        )
        self.canvas.create_text(10, 10, anchor="nw", text=text, font=("Courier", 14), fill="black")

        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Angle Between Two Vectors - Nature of Code")
    app = App(root)
    root.mainloop()
```

```python
#Vehicle Seeks Moving Target
import tkinter as tk
from vector import Vector

WIDTH, HEIGHT = 640, 240

class Path:
    def __init__(self):
        self.radius = 20
        self.start = Vector(0, HEIGHT / 3)
        self.end = Vector(WIDTH, 2 * HEIGHT / 3)

    def show(self, canvas):
        # Toleransi radius path
        canvas.create_line(
            self.start.x, self.start.y,
            self.end.x, self.end.y,
            width=self.radius * 2, fill="gray90", tags="path"
        )
        # Garis tengah path
        canvas.create_line(
            self.start.x, self.start.y,
            self.end.x, self.end.y,
            width=2, fill="black", tags="path"
        )

class Target:
    def __init__(self, path, speed=2):
        self.path = path
        self.position = path.start.copy()
        self.speed = speed
        self.direction = Vector.sub_vectors(path.end, path.start)
        self.direction.normalize()
        self.direction.mult(self.speed)

    def update(self):
        self.position.add(self.direction)
        if self.position.x > self.path.end.x:
            self.position = self.path.start.copy()

    def show(self, canvas):
        r = 6
        x, y = self.position.x, self.position.y
        canvas.create_oval(
            x - r, y - r, x + r, y + r,
            fill="red", outline="black", tags="target"
        )

class Vehicle:
    def __init__(self, canvas, x, y, maxspeed=4, maxforce=0.1):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.r = 8
        self.maxspeed = maxspeed
        self.maxforce = maxforce

        points = self.calculate_points()
        self.id = self.canvas.create_polygon(points, fill="blue", outline="black", tags="vehicle")

    def calculate_points(self):
        theta = self.velocity.heading()
        cos_t = math.cos(theta)
        sin_t = math.sin(theta)
        points = []
        local_points = [(self.r * 2, 0), (-self.r * 2, -self.r), (-self.r * 2, self.r)]
        for px, py in local_points:
            x = self.position.x + px * cos_t - py * sin_t
            y = self.position.y + px * sin_t + py * cos_t
            points.extend([x, y])
        return points

    def apply_force(self, force):
        self.acceleration.add(force)

    def seek(self, target):
        desired = Vector.sub_vectors(target, self.position)
        distance = desired.mag()
        if distance == 0:
            return

        desired.normalize()
        if distance < 50:
            # Slow down near target
            desired.mult(self.maxspeed * (distance / 50))
        else:
            desired.mult(self.maxspeed)

        steer = Vector.sub_vectors(desired, self.velocity)
        steer.limit(self.maxforce)

        self.apply_force(steer)

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def show(self):
        points = self.calculate_points()
        self.canvas.coords(self.id, *points)

    def run(self, target):
        self.seek(target)
        self.update()
        self.show()

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Path Following Step 3 - Vehicle Seeks Moving Target")

        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.path = Path()
        self.target = Target(self.path, speed=2)
        self.vehicle = Vehicle(self.canvas, 30, HEIGHT / 2, maxspeed=4, maxforce=0.1)

        self.update()

    def update(self):
        self.canvas.delete("path")
        self.canvas.delete("target")
        self.path.show(self.canvas)
        self.canvas.tag_lower("path")  # â¬…ï¸ Tambahkan baris ini
        self.target.update()
        self.target.show(self.canvas)

        self.vehicle.run(self.target.position)

        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
# Path Following
import tkinter as tk
import math
import random
from vector import Vector

WIDTH, HEIGHT = 640, 240

class Path:
    def __init__(self):
        self.radius = 20
        self.start = Vector(0, HEIGHT / 3)
        self.end = Vector(WIDTH, 2 * HEIGHT / 3)
    
    def show(self, canvas):
        # Draw path thickness
        canvas.create_line(
            self.start.x, self.start.y,
            self.end.x, self.end.y,
            width=self.radius*2, fill="gray90", tags="path"
        )
        # Draw center line
        canvas.create_line(
            self.start.x, self.start.y,
            self.end.x, self.end.y,
            width=1, fill="black", tags="path"
        )

class Vehicle:
    def __init__(self, canvas, x, y, maxspeed, maxforce):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)
        self.acceleration = Vector(0, 0)
        self.r = 6
        self.maxspeed = maxspeed
        self.maxforce = maxforce
        self.color = random.choice(["red", "blue", "green", "orange"])
        self.debug = True
        self.path = None

        points = self.calculate_points()
        self.id = canvas.create_polygon(points, fill=self.color, outline="black", tags="vehicle")

    def calculate_points(self):
        theta = self.velocity.heading()
        cos_t = math.cos(theta)
        sin_t = math.sin(theta)
        points = []
        local_points = [(self.r * 2, 0), (-self.r * 2, -self.r), (-self.r * 2, self.r)]
        for px, py in local_points:
            x = self.position.x + px * cos_t - py * sin_t
            y = self.position.y + px * sin_t + py * cos_t
            points.extend([x, y])
        return points

    def follow(self, path):
        if not path:
            return

        # Step 1: Predict future position
        future = self.velocity.copy().setted_mag(25).added(self.position)

        # Step 2: Find normal point
        normal_point = self.get_normal_point(future, path.start, path.end)

        # Step 3: Target point 25 ahead on the path
        b = Vector.subbed(path.end, path.start).setted_mag(25)
        target = Vector.added(normal_point, b)

        # Step 4: Steer if off path
        distance = future.dist(normal_point)
        if distance > path.radius:
            self.seek(target)

        # Debug drawing
        if self.debug:
            self.canvas.create_line(self.position.x, self.position.y, future.x, future.y, fill="gray", tags="debug")
            self.canvas.create_oval(future.x-2, future.y-2, future.x+2, future.y+2, fill="gray", outline="", tags="debug")
            self.canvas.create_line(future.x, future.y, normal_point.x, normal_point.y, fill="gray", tags="debug")
            self.canvas.create_oval(normal_point.x-2, normal_point.y-2, normal_point.x+2, normal_point.y+2, fill="gray", outline="", tags="debug")
            #if distance > path.radius:
            #    self.canvas.create_oval(target.x-4, target.y-4, target.x+4, target.y+4, fill="red", outline="", tags="debug")
            color = "red" if distance > path.radius else "green"
            self.canvas.create_oval(target.x - 4, target.y - 4, target.x + 4, target.y + 4,fill=color, outline="", tags="debug")

    def get_normal_point(self, position, a, b):
        ap = Vector.subbed(position, a)
        ab = Vector.subbed(b, a).normalized()
        ab.mult(ap.dot(ab))
        normal_point = Vector.added(a, ab)
        return normal_point

    def seek(self, target):
        desired = Vector.subbed(target, self.position)
        if desired.mag() == 0:
            return
        desired = desired.normalized().multed(self.maxspeed)
        steer = Vector.subbed(desired, self.velocity).limited(self.maxforce)
        self.apply_force(steer)

    def apply_force(self, force):
        self.acceleration = self.acceleration.added(force)

    def update(self):
        self.velocity = self.velocity.added(self.acceleration).limited(self.maxspeed)
        self.position = self.position.added(self.velocity)
        self.acceleration = self.acceleration.multed(0)

    def borders(self, path):
        if self.position.x > path.end.x + self.r:
            self.position.x = path.start.x - self.r
            self.position.y = path.start.y + (self.position.y - path.end.y)

    def show(self):
        points = self.calculate_points()
        self.canvas.coords(self.id, *points)

    def run(self):
        self.update()
        if self.path:
            self.borders(self.path)
        self.show()

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Path Following - Tkinter")

        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.debug = True
        self.path = Path()

        self.vehicle1 = Vehicle(self.canvas, 30, HEIGHT/2, 2, 0.05)
        self.vehicle2 = Vehicle(self.canvas, 30, HEIGHT/2, 3, 0.1)

        self.vehicle1.path = self.path
        self.vehicle2.path = self.path
        self.vehicle1.debug = self.debug
        self.vehicle2.debug = self.debug

        self.root.bind("<space>", self.toggle_debug)
        self.canvas.bind("<Button-1>", self.reset_vehicles)

        self.update()

    def toggle_debug(self, event):
        self.debug = not self.debug
        self.vehicle1.debug = self.debug
        self.vehicle2.debug = self.debug
        if not self.debug:
            self.canvas.delete("debug")

    def reset_vehicles(self, event):
        for v in [self.vehicle1, self.vehicle2]:
            v.position = Vector(0, HEIGHT/2)
            v.velocity = Vector(2, 0)
            v.acceleration = Vector(0, 0)

    def update(self):
        self.canvas.delete("path")
        self.canvas.delete("debug")

        self.path.show(self.canvas)
        self.canvas.tag_lower("path")  # â¬…ï¸ Tambahkan baris ini

        self.vehicle1.follow(self.path)
        self.vehicle2.follow(self.path)

        self.vehicle1.run()
        self.vehicle2.run()

        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import math
from vector import Vector

WIDTH, HEIGHT = 640, 480

class Path:
    def __init__(self):
        self.radius = 20
        # Path kontrol titik (Bezier-like polyline)
        self.points = [
            Vector(0, HEIGHT/2),
            Vector(WIDTH/3, HEIGHT/3),
            Vector(2*WIDTH/3, 2*HEIGHT/3),
            Vector(WIDTH, HEIGHT/2)
        ]

    def show(self, canvas):
        # Radius tolerance area
        for i in range(len(self.points) - 1):
            canvas.create_line(
                self.points[i].x, self.points[i].y,
                self.points[i+1].x, self.points[i+1].y,
                fill="gray90", width=self.radius * 2,
                tags="path"
            )
        # Center line
        for i in range(len(self.points) - 1):
            canvas.create_line(
                self.points[i].x, self.points[i].y,
                self.points[i+1].x, self.points[i+1].y,
                fill="black", width=2,
                tags="path"
            )

    def get_closest_point(self, predict):
        record = float('inf')
        closest_point = None

        for i in range(len(self.points) - 1):
            a = self.points[i]
            b = self.points[i+1]
            normal_point = get_normal_point(predict, a, b)

            if (normal_point.x < min(a.x, b.x) - 0.1 or
                normal_point.x > max(a.x, b.x) + 0.1 or
                normal_point.y < min(a.y, b.y) - 0.1 or
                normal_point.y > max(a.y, b.y) + 0.1):
                normal_point = b.copy()

            d = predict.dist(normal_point)
            if d < record:
                record = d
                closest_point = normal_point

        return closest_point, record

class Vehicle:
    def __init__(self, canvas, x, y, maxspeed=3, maxforce=0.1):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)
        self.acceleration = Vector(0, 0)
        self.r = 8
        self.maxspeed = maxspeed
        self.maxforce = maxforce
        self.path = None

        points = self.calculate_points()
        self.id = self.canvas.create_polygon(points, fill="blue", outline="black", tags="vehicle")

    def calculate_points(self):
        theta = self.velocity.heading()
        cos_t = math.cos(theta)
        sin_t = math.sin(theta)
        points = []
        local_points = [(self.r * 2, 0), (-self.r * 2, -self.r), (-self.r * 2, self.r)]
        for px, py in local_points:
            x = self.position.x + px * cos_t - py * sin_t
            y = self.position.y + px * sin_t + py * cos_t
            points.extend([x, y])
        return points

    def apply_force(self, force):
        self.acceleration.add(force)

    def seek(self, target):
        desired = Vector.sub_vectors(target, self.position)
        d = desired.mag()
        if d == 0:
            return
        desired.normalize()
        desired.mult(self.maxspeed)
        steer = Vector.sub_vectors(desired, self.velocity)
        steer.limit(self.maxforce)
        self.apply_force(steer)

    def follow(self, path):
        predict = self.velocity.copy()
        predict.set_mag(25)
        predict_point = Vector.add_vectors(self.position, predict)

        normal_point, distance = path.get_closest_point(predict_point)

        dir = Vector.sub_vectors(predict_point, normal_point)
        dir.set_mag(10)
        target = Vector.add_vectors(normal_point, dir)

        # Debug Visuals
        self.canvas.create_line(
            self.position.x, self.position.y,
            predict_point.x, predict_point.y,
            fill="gray", tags="debug"
        )
        self.canvas.create_oval(
            predict_point.x - 2, predict_point.y - 2,
            predict_point.x + 2, predict_point.y + 2,
            fill="gray", outline="", tags="debug"
        )
        self.canvas.create_line(
            predict_point.x, predict_point.y,
            normal_point.x, normal_point.y,
            fill="gray", tags="debug"
        )
        self.canvas.create_oval(
            normal_point.x - 2, normal_point.y - 2,
            normal_point.x + 2, normal_point.y + 2,
            fill="gray", outline="", tags="debug"
        )
        if distance > path.radius:
            self.canvas.create_oval(
                target.x - 4, target.y - 4,
                target.x + 4, target.y + 4,
                fill="red", outline="", tags="debug"
            )
            self.seek(target)

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def show(self):
        points = self.calculate_points()
        self.canvas.coords(self.id, *points)

    def run(self):
        if self.path:
            self.follow(self.path)
        self.update()
        self.show()

def get_normal_point(p, a, b):
    ap = Vector.sub_vectors(p, a)
    ab = Vector.sub_vectors(b, a)
    ab_normalized = ab.normalized()
    ab_normalized.mult(ap.dot(ab_normalized))
    normal_point = Vector.add_vectors(a, ab_normalized)
    return normal_point

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Path Following Bezier")

        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.path = Path()
        self.vehicle = Vehicle(self.canvas, 0, HEIGHT / 2, maxspeed=3, maxforce=0.05)
        self.vehicle.path = self.path

        self.update()

    def update(self):
        self.canvas.delete("path")
        self.canvas.delete("debug")

        self.path.show(self.canvas)
        self.canvas.tag_lower("path")  # â¬…ï¸ Tambahkan baris ini
        self.vehicle.run()

        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import math
import random
from vector import Vector

WIDTH, HEIGHT = 800, 600

class Path:
    def __init__(self):
        self.radius = 20
        self.segments = [
            [Vector(0, 100), Vector(200, 300)],  # Segmen atas (horizontal)
            [Vector(200, 300), Vector(600, 200)],  # Segmen diagonal (turun)
            [Vector(600, 200), Vector(800, 300)]   # Segmen bawah (horizontal)
        ]
    
    def show(self, canvas):
        for segment in self.segments:
            # Draw path thickness
            canvas.create_line(
                segment[0].x, segment[0].y,
                segment[1].x, segment[1].y,
                width=self.radius*2, fill="gray90", tags="path"
            )
            # Draw center line
            canvas.create_line(
                segment[0].x, segment[0].y,
                segment[1].x, segment[1].y,
                width=1, fill="black", tags="path"
            )

class Vehicle:
    def __init__(self, canvas, x, y, maxspeed, maxforce):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(random.uniform(-1, 1), random.uniform(-1, 1))
        self.velocity.set_mag(maxspeed/2)
        self.acceleration = Vector(0, 0)
        self.r = 8
        self.maxspeed = maxspeed
        self.maxforce = maxforce
        self.color = random.choice(["red", "blue", "green", "orange"])
        self.debug = True
        self.path = None
        self.current_segment = 0

        points = self.calculate_points()
        self.id = canvas.create_polygon(points, fill=self.color, outline="black", tags="vehicle")

    def calculate_points(self):
        theta = self.velocity.heading()
        cos_t = math.cos(theta)
        sin_t = math.sin(theta)
        points = []
        local_points = [(self.r * 2, 0), (-self.r * 2, -self.r), (-self.r * 2, self.r)]
        for px, py in local_points:
            x = self.position.x + px * cos_t - py * sin_t
            y = self.position.y + px * sin_t + py * cos_t
            points.extend([x, y])
        return points

    def follow(self, path):
        if not path:
            return

        # Step 1: Predict future position
        predict_dist = 50
        future = self.velocity.copy().setted_mag(predict_dist).added(self.position)

        # Step 2: Find nearest segment and normal point
        world_record = float('inf')
        target = None
        normal_point = None
        
        for i in range(len(path.segments)):
            seg_start = path.segments[i][0]
            seg_end = path.segments[i][1]
            
            np = self.get_normal_point(future, seg_start, seg_end)
            
            # Check if normal point is within segment
            seg_vec = Vector.sub_vectors(seg_end, seg_start)
            seg_len = seg_vec.mag()
            seg_vec.normalize()
            
            sp_to_np = Vector.sub_vectors(np, seg_start)
            proj = seg_vec.dot(sp_to_np)
            
            if proj < 0:
                np = seg_start.copy()
            elif proj > seg_len:
                np = seg_end.copy()
            
            dist = future.dist(np)
            
            if dist < world_record:
                world_record = dist
                self.current_segment = i
                normal_point = np.copy()
                
                # Step 3: Target point ahead on the path
                seg_dir = Vector.sub_vectors(seg_end, seg_start).normalized()
                target = normal_point.added(seg_dir.multed(predict_dist))
                
                # Adjust target if beyond segment end
                seg_to_target = Vector.sub_vectors(target, seg_start)
                seg_proj = seg_vec.dot(seg_to_target)
                
                if seg_proj > seg_len:
                    remaining = seg_proj - seg_len
                    next_seg = (i + 1) % len(path.segments)
                    next_seg_vec = Vector.sub_vectors(path.segments[next_seg][1], path.segments[next_seg][0]).normalized()
                    target = path.segments[next_seg][0].added(next_seg_vec.multed(remaining))

        # Step 4: Steer if off path
        if world_record > path.radius:
            self.seek(target)
        else:
            # Continue to follow path direction
            seg_dir = Vector.sub_vectors(path.segments[self.current_segment][1], 
                                       path.segments[self.current_segment][0]).normalized()
            target = self.position.added(seg_dir.multed(predict_dist))
            self.seek(target)

        # Debug drawing
        if self.debug:
            self.canvas.create_line(self.position.x, self.position.y, future.x, future.y, fill="gray", tags="debug")
            self.canvas.create_oval(future.x-3, future.y-3, future.x+3, future.y+3, fill="gray", outline="", tags="debug")
            self.canvas.create_line(future.x, future.y, normal_point.x, normal_point.y, fill="gray", tags="debug")
            self.canvas.create_oval(normal_point.x-3, normal_point.y-3, normal_point.x+3, normal_point.y+3, fill="gray", outline="", tags="debug")
            
            color = "red" if world_record > path.radius else "green"
            self.canvas.create_oval(target.x - 5, target.y - 5, target.x + 5, target.y + 5, fill=color, outline="", tags="debug")

    def get_normal_point(self, position, a, b):
        ap = Vector.sub_vectors(position, a)
        ab = Vector.sub_vectors(b, a)
        ab_normalized = ab.normalized()
        ab_normalized.mult(ap.dot(ab_normalized))
        normal_point = a.added(ab_normalized)
        return normal_point

    def seek(self, target):
        desired = Vector.sub_vectors(target, self.position)
        if desired.mag() == 0:
            return
        desired = desired.normalized().multed(self.maxspeed)
        steer = Vector.sub_vectors(desired, self.velocity).limited(self.maxforce)
        self.apply_force(steer)

    def apply_force(self, force):
        self.acceleration = self.acceleration.added(force)

    def update(self):
        self.velocity = self.velocity.added(self.acceleration).limited(self.maxspeed)
        self.position = self.position.added(self.velocity)
        self.acceleration = self.acceleration.multed(0)

    def borders(self, path):
        if self.position.x < 0:
            self.position.x = WIDTH
        elif self.position.x > WIDTH:
            self.position.x = 0
        if self.position.y < 0:
            self.position.y = HEIGHT
        elif self.position.y > HEIGHT:
            self.position.y = 0

    def show(self):
        points = self.calculate_points()
        self.canvas.coords(self.id, *points)

    def run(self):
        self.update()
        if self.path:
            self.follow(self.path)
            self.borders(self.path)
        self.show()

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Multi-Segment Path Following - Z Shape")

        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.debug = True
        self.path = Path()

        # Create multiple vehicles
        self.vehicles = []
        for i in range(2):
            x = random.randint(50, 150)
            y = random.randint(50, 150)
            maxspeed = random.uniform(2, 4)
            maxforce = random.uniform(0.05, 0.2)
            vehicle = Vehicle(self.canvas, x, y, maxspeed, maxforce)
            vehicle.path = self.path
            vehicle.debug = self.debug
            self.vehicles.append(vehicle)

        self.root.bind("<space>", self.toggle_debug)
        self.canvas.bind("<Button-1>", self.reset_vehicles)

        self.update()

    def toggle_debug(self, event):
        self.debug = not self.debug
        for vehicle in self.vehicles:
            vehicle.debug = self.debug
        if not self.debug:
            self.canvas.delete("debug")

    def reset_vehicles(self, event):
        for vehicle in self.vehicles:
            vehicle.position = Vector(random.randint(50, 150), random.randint(50, 150))
            vehicle.velocity = Vector(random.uniform(-1, 1), random.uniform(-1, 1))
            vehicle.velocity.set_mag(vehicle.maxspeed/2)
            vehicle.acceleration = Vector(0, 0)
            vehicle.current_segment = 0

    def update(self):
        self.canvas.delete("path")
        self.canvas.delete("debug")

        self.path.show(self.canvas)
        self.canvas.tag_lower("path")

        for vehicle in self.vehicles:
            vehicle.run()

        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import math
from vector import Vector

WIDTH, HEIGHT = 640, 480

class BezierPath:
    def __init__(self, p0, p1, p2, p3, radius=20):
        self.p0 = p0
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.radius = radius

    def bezier_point(self, t):
        """Evaluasi titik pada cubic Bezier dengan t [0..1]"""
        u = 1 - t
        p = self.p0.multed(u ** 3)
        p = p.added(self.p1.multed(3 * u ** 2 * t))
        p = p.added(self.p2.multed(3 * u * t ** 2))
        p = p.added(self.p3.multed(t ** 3))
        return p

    def show(self, canvas):
        prev = self.p0
        steps = 100
        for i in range(1, steps + 1):
            t = i / steps
            point = self.bezier_point(t)
            canvas.create_line(prev.x, prev.y, point.x, point.y, fill="gray90", width=self.radius*2, tags="path")
            prev = point

        prev = self.p0
        for i in range(1, steps + 1):
            t = i / steps
            point = self.bezier_point(t)
            canvas.create_line(prev.x, prev.y, point.x, point.y, fill="black", width=2, tags="path")
            prev = point

    def get_closest_point(self, predict, samples=100):
        record = float('inf')
        closest_point = None

        for i in range(samples + 1):
            t = i / samples
            point = self.bezier_point(t)
            d = predict.dist(point)
            if d < record:
                record = d
                closest_point = point

        return closest_point, record

class Vehicle:
    def __init__(self, canvas, x, y, maxspeed=3, maxforce=0.1):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)
        self.acceleration = Vector(0, 0)
        self.r = 8
        self.maxspeed = maxspeed
        self.maxforce = maxforce
        self.path = None

        points = self.calculate_points()
        self.id = self.canvas.create_polygon(points, fill="blue", outline="black", tags="vehicle")

    def calculate_points(self):
        theta = self.velocity.heading()
        cos_t = math.cos(theta)
        sin_t = math.sin(theta)
        points = []
        local_points = [(self.r * 2, 0), (-self.r * 2, -self.r), (-self.r * 2, self.r)]
        for px, py in local_points:
            x = self.position.x + px * cos_t - py * sin_t
            y = self.position.y + px * sin_t + py * cos_t
            points.extend([x, y])
        return points

    def apply_force(self, force):
        self.acceleration.add(force)

    def seek(self, target):
        desired = Vector.sub_vectors(target, self.position)
        d = desired.mag()
        if d == 0:
            return
        desired.normalize()
        desired.mult(self.maxspeed)
        steer = Vector.sub_vectors(desired, self.velocity)
        steer.limit(self.maxforce)
        self.apply_force(steer)

    def follow(self, path):
        predict = self.velocity.copy()
        predict.set_mag(25)
        predict_point = Vector.add_vectors(self.position, predict)

        normal_point, distance = path.get_closest_point(predict_point)

        dir = Vector.sub_vectors(predict_point, normal_point)
        dir.set_mag(10)
        target = Vector.add_vectors(normal_point, dir)

        # Debug Visuals
        self.canvas.create_line(
            self.position.x, self.position.y,
            predict_point.x, predict_point.y,
            fill="gray", tags="debug"
        )
        self.canvas.create_oval(
            predict_point.x - 2, predict_point.y - 2,
            predict_point.x + 2, predict_point.y + 2,
            fill="gray", outline="", tags="debug"
        )
        self.canvas.create_line(
            predict_point.x, predict_point.y,
            normal_point.x, normal_point.y,
            fill="gray", tags="debug"
        )
        self.canvas.create_oval(
            normal_point.x - 2, normal_point.y - 2,
            normal_point.x + 2, normal_point.y + 2,
            fill="gray", outline="", tags="debug"
        )
        if distance > path.radius:
            self.canvas.create_oval(
                target.x - 4, target.y - 4,
                target.x + 4, target.y + 4,
                fill="red", outline="", tags="debug"
            )
            self.seek(target)

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def show(self):
        points = self.calculate_points()
        self.canvas.coords(self.id, *points)

    def run(self):
        if self.path:
            self.follow(self.path)
        self.update()
        self.show()

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Path Following with Bezier Evaluation")

        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        # Buat Bezier Path
        p0 = Vector(0, HEIGHT * 0.5)
        p1 = Vector(WIDTH * 0.3, HEIGHT * 0.1)
        p2 = Vector(WIDTH * 0.7, HEIGHT * 0.9)
        p3 = Vector(WIDTH, HEIGHT * 0.5)
        self.path = BezierPath(p0, p1, p2, p3, radius=20)

        # Vehicle
        self.vehicle = Vehicle(self.canvas, 0, HEIGHT / 2, maxspeed=3, maxforce=0.05)
        self.vehicle.path = self.path

        self.update()

    def update(self):
        self.canvas.delete("path")
        self.canvas.delete("debug")

        self.path.show(self.canvas)
        self.vehicle.run()

        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import random
import math
from vector import Vector

WIDTH, HEIGHT = 640, 360

# =============== PATH CLASS ===============
class Path:
    def __init__(self):
        self.radius = 20
        self.points = []

    def add_point(self, x, y):
        self.points.append(Vector(x, y))

    def show(self, canvas):
        # draw thick path
        for i in range(len(self.points)):
            a = self.points[i]
            b = self.points[(i + 1) % len(self.points)]
            canvas.create_line(a.x, a.y, b.x, b.y, fill="lightgray", width=self.radius * 2, tags="path")
        # draw center line
        for i in range(len(self.points)):
            a = self.points[i]
            b = self.points[(i + 1) % len(self.points)]
            canvas.create_line(a.x, a.y, b.x, b.y, fill="black", width=1, tags="path")


# =============== VEHICLE CLASS ===============
class Vehicle:
    def __init__(self, canvas, x, y, maxspeed, maxforce):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(random.uniform(-1, 1), random.uniform(-1, 1))
        self.acceleration = Vector(0, 0)
        self.r = 6
        self.maxspeed = maxspeed
        self.maxforce = maxforce

        self.id = canvas.create_oval(
            self.position.x - self.r, self.position.y - self.r,
            self.position.x + self.r, self.position.y + self.r,
            fill="skyblue", outline="black", tags="vehicle"
        )

    def apply_force(self, force):
        self.acceleration.add(force)

    def apply_behaviors(self, vehicles, path):
        follow_force = self.follow(path).multed(3)
        separate_force = self.separate(vehicles).multed(1)
        self.apply_force(follow_force)
        self.apply_force(separate_force)

    def follow(self, path):
        predict = self.velocity.copy()
        predict.set_mag(25)
        predictpos = self.position.added(predict)

        normal = None
        target = None
        world_record = float('inf')

        for i in range(len(path.points)):
            a = path.points[i]
            b = path.points[(i + 1) % len(path.points)]
            normal_point = self.get_normal_point(predictpos, a, b)

            dir = b.subbed(a)
            if not self.is_point_on_line(normal_point, a, b):
                normal_point = b.copy()
                a = path.points[(i + 1) % len(path.points)]
                b = path.points[(i + 2) % len(path.points)]
                dir = b.subbed(a)

            d = predictpos.dist(normal_point)
            if d < world_record:
                world_record = d
                dir.normalize()
                dir.mult(10)
                target = normal_point.added(dir)

        if world_record > path.radius:
            return self.seek(target)
        else:
            return Vector(0, 0)

    def get_normal_point(self, p, a, b):
        ap = p.subbed(a)
        ab = b.subbed(a)
        ab_norm = ab.normalized()
        ab_norm.mult(ab_norm.dot(ap))
        normal_point = a.added(ab_norm)
        return normal_point

    def is_point_on_line(self, p, a, b):
        return (min(a.x, b.x) <= p.x <= max(a.x, b.x)) and (min(a.y, b.y) <= p.y <= max(a.y, b.y))

    def separate(self, vehicles):
        desired_separation = self.r * 2
        steer = Vector(0, 0)
        count = 0
        for other in vehicles:
            d = self.position.dist(other.position)
            if 0 < d < desired_separation:
                diff = self.position.subbed(other.position)
                diff.normalize()
                diff.div(d)
                steer.add(diff)
                count += 1
        if count > 0:
            steer.div(count)
            steer.set_mag(self.maxspeed)
            steer.sub(self.velocity)
            steer.limit(self.maxforce)
        return steer

    def seek(self, target):
        desired = target.subbed(self.position)
        desired.set_mag(self.maxspeed)
        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)
        return steer

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def borders(self):
        if self.position.x < -self.r: self.position.x = WIDTH + self.r
        if self.position.y < -self.r: self.position.y = HEIGHT + self.r
        if self.position.x > WIDTH + self.r: self.position.x = -self.r
        if self.position.y > HEIGHT + self.r: self.position.y = -self.r

    def show(self):
        self.canvas.coords(
            self.id,
            self.position.x - self.r, self.position.y - self.r,
            self.position.x + self.r, self.position.y + self.r
        )

    def run(self):
        self.update()
        self.borders()
        self.show()

# =============== APP CLASS ===============
class App:
    def __init__(self, root):
        self.root = root
        self.root.title("NOC_5_13_CrowdPathFollowing - Tkinter")

        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.path = Path()
        offset = 30
        self.path.add_point(offset, offset)
        self.path.add_point(WIDTH - offset, offset)
        self.path.add_point(WIDTH - offset, HEIGHT - offset)
        self.path.add_point(WIDTH / 2, HEIGHT - offset * 3)
        self.path.add_point(offset, HEIGHT - offset)

        self.vehicles = []
        for _ in range(50):
            maxspeed = random.uniform(2, 4)
            maxforce = 0.3
            v = Vehicle(self.canvas, random.uniform(0, WIDTH), random.uniform(0, HEIGHT), maxspeed, maxforce)
            self.vehicles.append(v)

        self.canvas.bind("<Button-1>", self.add_vehicle_mouse)
        self.update()

    def add_vehicle_mouse(self, event):
        maxspeed = random.uniform(2, 4)
        maxforce = 0.3
        v = Vehicle(self.canvas, event.x, event.y, maxspeed, maxforce)
        self.vehicles.append(v)

    def update(self):
        self.canvas.delete("path")
        self.path.show(self.canvas)
        self.canvas.tag_lower("path")  # â¬…ï¸ Tambahkan baris ini
        for v in self.vehicles:
            v.apply_behaviors(self.vehicles, self.path)
            v.run()

        self.root.after(30, self.update)

# =============== RUN ===============
if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import random
from vector import Vector

WIDTH, HEIGHT = 640, 360

class Boid:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(random.uniform(-1, 1), random.uniform(-1, 1))
        self.acceleration = Vector(0, 0)
        self.r = 6
        self.maxspeed = 3
        self.maxforce = 0.05

        # Draw as a triangle
        points = self.calculate_points()
        self.id = canvas.create_polygon(points, fill="gray", outline="black", tags="boid")

    def calculate_points(self):
        theta = self.velocity.heading()
        cos_t = math.cos(theta)
        sin_t = math.sin(theta)
        points = []
        local_points = [(self.r * 2, 0), (-self.r * 2, -self.r), (-self.r * 2, self.r)]
        for px, py in local_points:
            x = self.position.x + px * cos_t - py * sin_t
            y = self.position.y + px * sin_t + py * cos_t
            points.extend([x, y])
        return points

    def apply_force(self, force):
        self.acceleration.added(force)

    def seek(self, target):
        desired = Vector.subbed(target, self.position)
        desired.normalize()
        desired.multed(self.maxspeed)
        steer = Vector.subbed(desired, self.velocity)
        steer.limit(self.maxforce)
        return steer

    def separate(self, boids):
        desired_separation = 25
        steer = Vector(0, 0)
        count = 0
        for other in boids:
            d = self.position.dist(other.position)
            if 0 < d < desired_separation:
                diff = Vector.subbed(self.position, other.position)
                diff.normalize()
                diff.dived(d)
                steer.added(diff)
                count += 1
        if count > 0:
            steer.dived(count)
        if steer.mag() > 0:
            steer.normalize()
            steer.multed(self.maxspeed)
            steer.subbed(self.velocity)
            steer.limit(self.maxforce)
        return steer

    def align(self, boids):
        neighbor_dist = 50
        sum_vel = Vector(0, 0)
        count = 0
        for other in boids:
            d = self.position.dist(other.position)
            if 0 < d < neighbor_dist:
                sum_vel.added(other.velocity)
                count += 1
        if count > 0:
            sum_vel.dived(count)
            sum_vel.normalize()
            sum_vel.multed(self.maxspeed)
            steer = Vector.subbed(sum_vel, self.velocity)
            steer.limit(self.maxforce)
            return steer
        return Vector(0, 0)

    def cohere(self, boids):
        neighbor_dist = 50
        sum_pos = Vector(0, 0)
        count = 0
        for other in boids:
            d = self.position.dist(other.position)
            if 0 < d < neighbor_dist:
                sum_pos.added(other.position)
                count += 1
        if count > 0:
            sum_pos.dived(count)
            return self.seek(sum_pos)
        return Vector(0, 0)

    def flock(self, boids):
        sep = self.separate(boids)
        ali = self.align(boids)
        coh = self.cohere(boids)

        sep.multed(1.5)
        ali.multed(1.0)
        coh.multed(1.0)

        self.apply_force(sep)
        self.apply_force(ali)
        self.apply_force(coh)

    def update(self):
        self.velocity.added(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.added(self.velocity)
        self.acceleration.multed(0)

    def borders(self):
        if self.position.x < -self.r:
            self.position.x = WIDTH + self.r
        if self.position.y < -self.r:
            self.position.y = HEIGHT + self.r
        if self.position.x > WIDTH + self.r:
            self.position.x = -self.r
        if self.position.y > HEIGHT + self.r:
            self.position.y = -self.r

    def show(self):
        points = self.calculate_points()
        self.canvas.coords(self.id, *points)

    def run(self, boids):
        self.flock(boids)
        self.update()
        self.borders()
        self.show()



class App:
    def __init__(self, root):
        self.root = root
        self.root.title("Flocking Simulation - The Nature of Code")
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.boids = [Boid(self.canvas, WIDTH/2, HEIGHT/2) for _ in range(100)]

        self.canvas.bind("<B1-Motion>", self.add_boid)
        self.update()

    def add_boid(self, event):
        self.boids.append(Boid(self.canvas, event.x, event.y))

    def update(self):
        for boid in self.boids:
            boid.run(self.boids)
        self.root.after(33, self.update)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import math
from vector import Vector

WIDTH, HEIGHT = 640, 360

class Vehicle:
    def __init__(self, x, y, canvas, color='gray'):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(1, 0)
        self.acceleration = Vector()
        self.r = 6
        self.maxspeed = 4
        self.maxforce = 0.2
        self.color = color

        self.shape = self.canvas.create_polygon(0, 0, 0, 0, 0, 0, fill=self.color, outline='black')

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)
        self.wraparound()

    def apply_force(self, force):
        self.acceleration.add(force)

    def wraparound(self):
        if self.position.x < 0: self.position.x = WIDTH
        elif self.position.x > WIDTH: self.position.x = 0
        if self.position.y < 0: self.position.y = HEIGHT
        elif self.position.y > HEIGHT: self.position.y = 0

    def seek(self, target):
        desired = target.subbed(self.position)
        desired.set_mag(self.maxspeed)
        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)
        self.apply_force(steer)

    def pursue(self, target_vehicle):
        prediction = 25  # time units ahead
        future_position = target_vehicle.position.added(
            target_vehicle.velocity.multed(prediction)
        )
        self.seek(future_position)

    def show(self):
        angle = self.velocity.heading()
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r * 2)
        left = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r * 2)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r * 2)

        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        self.canvas.coords(self.shape, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)


class App:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg='white')
        self.canvas.pack()

        self.target = Vehicle(WIDTH // 4, HEIGHT // 2, self.canvas, color='red')
        self.pursuer = Vehicle(WIDTH // 2, HEIGHT // 2, self.canvas, color='blue')

        self.animate()

    def animate(self):
        # Gerakkan target dengan arah acak sedikit
        angle_change = (math.sin(self.target.position.x / 50)) * 0.05
        angle = self.target.velocity.heading() + angle_change
        self.target.velocity = Vector(math.cos(angle), math.sin(angle)).setted_mag(self.target.maxspeed)

        self.target.update()
        self.pursuer.pursue(self.target)
        self.pursuer.update()

        self.target.show()
        self.pursuer.show()

        self.canvas.after(16, self.animate)


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Exercise 5.3 â€“ Pursue Behavior")
    app = App(root)
    root.mainloop()
```
