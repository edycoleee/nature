```python
# Vehicle Bergerak dan Berputar Pelan *8
import tkinter as tk
import math
from vector import Vector

class Vehicle:
    def __init__(self, x, y, canvas):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(1, 0.5)  # bergerak pelan ke kanan dan bawah
        self.r = 6  # ukuran
        self.shape = self.canvas.create_polygon(0, 0, 0, 0, 0, 0, fill='lightblue', outline='black', width=2)
        self.update()

    def update(self):
        # Update posisi vehicle
        self.position.add(self.velocity)
        self.show()
        self.canvas.after(30, self.update)  # panggil update setiap 30 ms

    def show(self):
        angle = self.velocity.heading()  # arah segitiga mengikuti arah velocity

        # Titik depan, kiri, kanan
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r * 8)
        left  = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r * 8)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r * 8)

        # Tambahkan posisi pusat vehicle
        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        # Gambar segitiga
        self.canvas.coords(self.shape, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)

        # Hapus label sebelumnya
        self.canvas.delete('info')

        # Label titik-titik untuk belajar
        for p, label in zip([p1, p2, p3], ['P1', 'P2', 'P3']):
            self.canvas.create_oval(p.x-3, p.y-3, p.x+3, p.y+3, fill='red', tags='info')
            self.canvas.create_text(p.x+10, p.y, text=f"{label} ({int(p.x)}, {int(p.y)})", anchor='w', font=("Arial", 9), tags='info')

        # Titik pusat
        cx, cy = self.position.x, self.position.y
        self.canvas.create_oval(cx-3, cy-3, cx+3, cy+3, fill='green', tags='info')
        self.canvas.create_text(cx+10, cy, text=f"Center ({int(cx)}, {int(cy)})", anchor='w', font=("Arial", 9), tags='info')

# Membuat window
root = tk.Tk()
root.title("Vehicle Bergerak dan Berputar Pelan")

# Membuat canvas
canvas = tk.Canvas(root, width=500, height=400, bg='white')
canvas.pack()

# Membuat Vehicle pada posisi awal (100, 100)
v = Vehicle(100, 100, canvas)

root.mainloop()
```

```python
import tkinter as tk
import math
from vector import Vector 

WIDTH, HEIGHT = 600, 400

class Vehicle:
    def __init__(self, x, y, canvas):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)  # awal ke kanan
        self.acceleration = Vector(0, 0)
        self.maxspeed = 2
        self.maxforce = 0.05
        self.r = 8
        self.trail = []

        # Segitiga vehicle
        self.shape = self.canvas.create_polygon(0, 0, 0, 0, 0, 0, fill='gray', outline='black')

    def seek(self, target):
        desired = target.subbed(self.position)
        distance = desired.magnitude()
        desired_unit = desired.normalized()
        desired_velocity = desired_unit.multed(self.maxspeed)

        # Steering = desired - velocity
        steering = desired_velocity.subbed(self.velocity)
        steering.limit(self.maxforce)

        # Apply steering force
        self.acceleration = steering

        angle = math.atan2(desired.y, desired.x)
        return desired_velocity, distance, angle, desired_unit

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)

        self.trail.append(self.position.copy())
        if len(self.trail) > 100:
            self.trail.pop(0)

    def show(self):
        # Draw trail
        if len(self.trail) > 1:
            coords = []
            for vec in self.trail:
                coords.extend([vec.x, vec.y])
            self.canvas.create_line(*coords, fill="#888", width=1, smooth=True)

        # Hitung orientasi segitiga berdasarkan arah kecepatan
        angle = self.velocity.heading()
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r * 2)
        left = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r * 2)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r * 2)

        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        self.canvas.coords(self.shape, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)

class App:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.vehicle = Vehicle(100, 100, self.canvas)

        # Waypoints
        self.waypoints = [Vector(250, 300), Vector(250, 100), Vector(500, 150), Vector(100, 350), Vector(400, 350)]
        self.waypoint_index = 0
        self.target = self.waypoints[self.waypoint_index]
        self.target_shape = self.canvas.create_oval(self.target.x - 10, self.target.y - 10,
                                                    self.target.x + 10, self.target.y + 10, fill="green")

        self.text = self.canvas.create_text(10, 10, anchor="nw", font=("Courier", 10), fill="black")

        self.animate()

    def animate(self):
        v = self.vehicle
        desired_velocity, distance, angle, direction_unit = v.seek(self.target)

        if distance > 5:
            v.update()
        else:
            # Pindah ke waypoint berikutnya
            self.waypoint_index = (self.waypoint_index + 1) % len(self.waypoints)
            self.target = self.waypoints[self.waypoint_index]
            self.canvas.coords(self.target_shape, self.target.x - 10, self.target.y - 10,
                                                self.target.x + 10, self.target.y + 10)

        # Tampilkan vehicle
        v.show()

        # Info debug
        p = v.position
        dx = self.target.x - p.x
        dy = self.target.y - p.y
        info = f"""ðŸŽ¯ Vehicle Steering Seek

Waypoint ke-{self.waypoint_index + 1}: ({self.target.x}, {self.target.y})
Posisi: ({p.x:.1f}, {p.y:.1f})

Î”x = {dx:.1f}
Î”y = {dy:.1f}

Sudut ke target (Â°) = {math.degrees(angle):.1f}Â°
Velocity    = {v.velocity}
Acceleration = {v.acceleration}
"""
        self.canvas.itemconfig(self.text, text=info)

        self.canvas.after(30, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("ðŸŽ¯ Vehicle Steering Seek dengan Belok Halus")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import math
import random
from vector import Vector

# --- Parameter Debug ---
debug = True

class Vehicle:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)
        self.acceleration = Vector(0, 0)
        self.r = 6
        self.maxspeed = 2
        self.maxforce = 0.05
        self.wandertheta = 0

        self.heading_target = 0  # untuk heading segitiga agar sama dengan line1_id

        self.id = canvas.create_polygon(0, 0, 0, 0, 0, 0, fill="gray", outline="black")
        self.circle_id = canvas.create_oval(0, 0, 0, 0, outline="blue")
        self.line1_id = canvas.create_line(0, 0, 0, 0, fill="blue")
        self.line2_id = canvas.create_line(0, 0, 0, 0, fill="blue")
        self.target_id = canvas.create_oval(0, 0, 0, 0, outline="red", fill="red")

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)
        self.borders()

    def apply_force(self, force):
        self.acceleration.add(force)

    def seek(self, target):
        desired = Vector.sub_vectors(target, self.position)
        desired.set_mag(self.maxspeed)
        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)
        self.apply_force(steer)

    def wander(self):
        wander_r = 25
        wander_d = 80
        change = 0.3
        self.wandertheta += random.uniform(-change, change)

        if self.velocity.mag() == 0:
            heading = 0
            forward = Vector(0, -1)
        else:
            forward = self.velocity.copy().normalized()
            heading = self.velocity.heading()

        circle_pos = forward.copy().multed(wander_d).added(self.position)

        # Ambil heading segitiga dari position ke circle_pos agar sejajar line1_id
        dir_to_circle = circle_pos.subbed(self.position)
        self.heading_target = dir_to_circle.heading()

        circle_offset = Vector(
            wander_r * math.cos(self.wandertheta + heading),
            wander_r * math.sin(self.wandertheta + heading)
        )

        target = circle_pos.added(circle_offset)
        self.seek(target)

        if debug:
            x, y = circle_pos.x, circle_pos.y
            self.canvas.coords(self.circle_id, x - wander_r, y - wander_r, x + wander_r, y + wander_r)
            self.canvas.coords(self.line1_id, self.position.x, self.position.y, x, y)
            self.canvas.coords(self.line2_id, x, y, target.x, target.y)
            r = 4
            self.canvas.coords(self.target_id, target.x - r, target.y - r, target.x + r, target.y + r)
        else:
            self.canvas.coords(self.circle_id, 0, 0, 0, 0)
            self.canvas.coords(self.line1_id, 0, 0, 0, 0)
            self.canvas.coords(self.line2_id, 0, 0, 0, 0)
            self.canvas.coords(self.target_id, 0, 0, 0, 0)

    def show(self):
        angle = self.heading_target  # gunakan heading sejajar line1_id
        x = self.position.x
        y = self.position.y
        r = self.r
        p1 = Vector(0, -r * 2).rotate(angle)
        p2 = Vector(-r, r * 2).rotate(angle)
        p3 = Vector(r, r * 2).rotate(angle)
        points = [
            x + p1.x, y + p1.y,
            x + p2.x, y + p2.y,
            x + p3.x, y + p3.y
        ]
        self.canvas.coords(self.id, *points)

    def borders(self):
        width = int(self.canvas["width"])
        height = int(self.canvas["height"])
        if self.position.x < -self.r:
            self.position.x = width + self.r
        if self.position.y < -self.r:
            self.position.y = height + self.r
        if self.position.x > width + self.r:
            self.position.x = -self.r
        if self.position.y > height + self.r:
            self.position.y = -self.r

# --- Setup Tkinter ---
root = tk.Tk()
WIDTH, HEIGHT = 640, 240
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

vehicle = Vehicle(canvas, WIDTH / 2, HEIGHT / 2)

def toggle_debug(event):
    global debug
    debug = not debug

def update():
    canvas.delete("text")
    vehicle.wander()
    vehicle.update()
    vehicle.show()

    canvas.create_text(10, 10, anchor="nw", text=f"velocity: {vehicle.velocity}", font=("Courier", 10), tags="text")
    canvas.create_text(10, 25, anchor="nw", text=f"acceleration: {vehicle.acceleration}", font=("Courier", 10), tags="text")

    root.after(33, update)

canvas.bind("<Button-1>", toggle_debug)
update()
root.mainloop()
```

```python
import tkinter as tk
import math
from vector import Vector

WIDTH, HEIGHT = 640, 360

class Vehicle:
    def __init__(self, x, y, canvas, color='gray'):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(1, 0)
        self.acceleration = Vector()
        self.r = 6
        self.maxspeed = 4
        self.maxforce = 0.2
        self.color = color

        self.shape = self.canvas.create_polygon(0, 0, 0, 0, 0, 0, fill=self.color, outline='black')

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)
        self.wraparound()

    def apply_force(self, force):
        self.acceleration.add(force)

    def wraparound(self):
        if self.position.x < 0: self.position.x = WIDTH
        elif self.position.x > WIDTH: self.position.x = 0
        if self.position.y < 0: self.position.y = HEIGHT
        elif self.position.y > HEIGHT: self.position.y = 0

    def seek(self, target):
        desired = target.subbed(self.position)
        desired.set_mag(self.maxspeed)
        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)
        self.apply_force(steer)

    def pursue(self, target_vehicle):
        prediction = 25  # time units ahead
        future_position = target_vehicle.position.added(
            target_vehicle.velocity.multed(prediction)
        )
        self.seek(future_position)

    def show(self):
        angle = self.velocity.heading()
        front = Vector(math.cos(angle), math.sin(angle)).multed(self.r * 2)
        left = Vector(math.cos(angle + 2.5), math.sin(angle + 2.5)).multed(self.r * 2)
        right = Vector(math.cos(angle - 2.5), math.sin(angle - 2.5)).multed(self.r * 2)

        p1 = self.position.added(front)
        p2 = self.position.added(left)
        p3 = self.position.added(right)

        self.canvas.coords(self.shape, p1.x, p1.y, p2.x, p2.y, p3.x, p3.y)


class App:
    def __init__(self, root):
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg='white')
        self.canvas.pack()

        self.target = Vehicle(WIDTH // 4, HEIGHT // 2, self.canvas, color='red')
        self.pursuer = Vehicle(WIDTH // 2, HEIGHT // 2, self.canvas, color='blue')

        self.animate()

    def animate(self):
        # Gerakkan target dengan arah acak sedikit
        angle_change = (math.sin(self.target.position.x / 50)) * 0.05
        angle = self.target.velocity.heading() + angle_change
        self.target.velocity = Vector(math.cos(angle), math.sin(angle)).setted_mag(self.target.maxspeed)

        self.target.update()
        self.pursuer.pursue(self.target)
        self.pursuer.update()

        self.target.show()
        self.pursuer.show()

        self.canvas.after(16, self.animate)


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Exercise 5.3 â€“ Pursue Behavior")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector
import math

WIDTH, HEIGHT = 800, 600

class Vehicle:
    def __init__(self, x, y):
        self.position = Vector(x, y)
        self.velocity = Vector(2, 0)
        self.acceleration = Vector(0, 0)
        self.maxspeed = 4
        self.maxforce = 0.1

    def apply_force(self, force):
        self.acceleration.add(force)

    def seek(self, target):
        # desired = target - position
        desired = target.subbed(self.position)
        desired.set_mag(self.maxspeed)

        # steer = desired - velocity
        steer = desired.subbed(self.velocity)
        steer.limit(self.maxforce)

        self.apply_force(steer)

    def update(self):
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.maxspeed)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def draw(self, canvas):
        r = 10
        angle = self.velocity.heading()
        x, y = self.position.x, self.position.y

        p1 = (x + r * math.cos(angle), y + r * math.sin(angle))
        p2 = (x + r * math.cos(angle + 2.5), y + r * math.sin(angle + 2.5))
        p3 = (x + r * math.cos(angle - 2.5), y + r * math.sin(angle - 2.5))
        canvas.coords(self.shape, p1[0], p1[1], p2[0], p2[1], p3[0], p3[1])

    def init_draw(self, canvas):
        self.shape = canvas.create_polygon(0, 0, 0, 0, 0, 0, fill='blue')


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg='white')
        self.canvas.pack()

        self.vehicle = Vehicle(WIDTH // 2, HEIGHT // 2)
        self.vehicle.init_draw(self.canvas)

        self.target = Vector(WIDTH // 2, HEIGHT // 2)
        self.target_dot = self.canvas.create_oval(0, 0, 0, 0, fill='red')

        self.canvas.bind("<Motion>", self.update_target)
        self.animate()

    def update_target(self, event):
        self.target.set(event.x, event.y)

    def animate(self):
        self.vehicle.seek(self.target)
        self.vehicle.update()
        self.vehicle.draw(self.canvas)
        self.canvas.coords(
            self.target_dot,
            self.target.x - 5, self.target.y - 5,
            self.target.x + 5, self.target.y + 5
        )
        self.root.after(16, self.animate)


if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```
