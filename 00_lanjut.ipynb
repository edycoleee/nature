````python
# BOLA STATIS DENGAN TEXT
import tkinter as tk
from vector import Vector

WIDTH = 400
HEIGHT = 300

#CLASS, FUNGSI 
x,y,radius = 100,100,10

class Ball:
    #initialisasi parameter
    def __init__(self, canvas,x,y,radius):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.radius = radius
        #buat id class
        self.id = canvas.create_oval(
            self.x - self.radius, self.y - self.radius,
            self.x + self.radius, self.y + self.radius,
            fill="gray", outline="black", width=2
        )   
        # Info teks kosong
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')

    def display(self):
        x, y = self.x, self.y
        #display ulang Ball >> id
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )

        # Update teks kosong dengan info posisi (x,y)
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f})")


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        #MAIN PROGRAM
        # bola >> create class Ball
        self.bola = Ball(self.canvas,x,y,radius)
        self.animate()

    def animate(self):
        # update display setiap frame berganti
        self.bola.display()
        self.root.after(30, self.animate)



if __name__ == "__main__":
    root = tk.Tk()
    root.title("BOLA STATIS DENGAN TEXT")
    app = App(root)
    root.mainloop()
````

````python
# BOLA BERGERAK DENGAN TEXT DAN BOLAK BALIK SUMBU X
import tkinter as tk
from vector import Vector

WIDTH = 400
HEIGHT = 300

#CLASS, FUNGSI 
x,y,radius = 100,100,10
dx = 5

class Ball:
    def __init__(self, canvas,x,y,radius):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.radius = radius
        self.id = canvas.create_oval(
            self.x - self.radius, self.y - self.radius,
            self.x + self.radius, self.y + self.radius,
            fill="gray", outline="black", width=2
        )    

        # Info teks
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')

    def move(self):
        # Pantulan dari tepi
        global dx
        if self.x >= WIDTH - self.radius or self.x <= self.radius:
            dx *= -1

        self.x += dx

    def display(self):
        x, y = self.x, self.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )

         # Tampilkan info
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f}) , dx: {dx}")


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        #MAIN PROGRAM
        self.bola = Ball(self.canvas,x,y,radius)
        self.animate()

    def animate(self):
        self.bola.display()
        self.bola.move()
        root.after(30, self.animate)



if __name__ == "__main__":
    root = tk.Tk()
    root.title("BOLA BERGERAK DENGAN TEXT DAN BOLAK BALIK SUMBU X")
    app = App(root)
    root.mainloop()
````

PAHAMI VECTOR (X,Y), PENJUMLAHAN, PENGURANGAN, PERKALIAN, PEMBAGIAN, COPY

````python
# PENJUMLAHAN VECTOR (add)

class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    # ---------- Penambahan ----------
    def add(self, other):
        # Menambahkan nilai vektor lain ke vektor ini (ubah self)
        self.x += other.x
        self.y += other.y
        return self

    def added(self, other):
        # Menghasilkan vektor baru hasil penjumlahan tanpa mengubah self
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# Penjumlahan vector va + vb hasilnya disimpan di va    
va = Vector(2, 3)
vb = Vector(3, 7)
va.add(vb)
print("Hasil penjumlahan (disimpan di va):", va) #(5.00, 10.00)

# Penjumlahan vector va + vb hasilnya disimpan di vc
va = Vector(2, 3)
vb = Vector(3, 7)
vc = va.added(vb)
print("Vektor va tetap:", va) #(2.00, 3.00)
print("Hasil penjumlahan (disimpan di vc):", vc) #(5.00, 10.00)
````

````python
# PENGURANGAN VECTOR (sub)

class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    # ---------- Pengurangan ----------
    def sub(self, other):
        # Mengurangi nilai vektor lain dari vektor ini (ubah self)
        self.x -= other.x
        self.y -= other.y
        return self

    def subbed(self, other):
        # Menghasilkan vektor baru hasil pengurangan tanpa mengubah self
        return Vector(self.x - other.x, self.y - other.y)
    
    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# pengurangan vector va - vb hasilnya disimpan di va    
va = Vector(2, 3)
vb = Vector(3, 7)
va.sub(vb)
print("Hasil pengurangan (disimpan di va):", va) #(-1.00, -4.00)

# pengurangan vector va - vb hasilnya disimpan di vc
va = Vector(2, 3)
vb = Vector(3, 7)
vc = va.subbed(vb)
print("Vektor va tetap:", va) #(2.00, 3.00)
print("Hasil pengurangan (disimpan di vc):", vc) #(-1.00, -4.00)
````

````python
# PERKALIAN VECTOR (mult)
class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    # ---------- Perkalian dengan skalar ----------
    def mult(self, scalar):
        # Mengalikan nilai vektor dengan skalar (ubah self)
        self.x *= scalar
        self.y *= scalar
        return self

    def multed(self, scalar):
        # Menghasilkan vektor baru hasil perkalian dengan skalar
        return Vector(self.x * scalar, self.y * scalar)
    
    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# perkalian vector va * 2 hasilnya disimpan di va    
va = Vector(2, 3)
f = 2
va.mult(f)
print("Hasil perkalian (disimpan di va):", va)

# perkalian vector va * vb hasilnya disimpan di vc
va = Vector(2, 3)
f = 2
vc = va.multed(f)
print("Vektor va tetap:", va)
print("Hasil perkalian (disimpan di vc):", vc)
````

````python
# PEMBAGIAN VECTOR (div)
class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    # ---------- Pembagian dengan skalar ----------
    def div(self, scalar):
        # Membagi nilai vektor dengan skalar (ubah self)
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar
            return self

    def dived(self, scalar):
        # Menghasilkan vektor baru hasil pembagian dengan skalar
        if scalar != 0:
            return Vector(self.x / scalar, self.y / scalar)
        return Vector(0, 0)
    
    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# perkalian vector va * 2 hasilnya disimpan di va    
va = Vector(4, 10)
f = 2
va.div(f)
print("Hasil perkalian (disimpan di va):", va) #(2.00, 5.00)

# perkalian vector va * vb hasilnya disimpan di vc
va = Vector(4, 10)
f = 2
vc = va.dived(f)
print("Vektor va tetap:", va) # (4.00, 10.00)
print("Hasil perkalian (disimpan di vc):", vc) #(2.00, 5.00)
````

````python
# COPY VECTOR (copy)
class Vector:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    # ---------- Copy / clone ----------
    def copy(self):
        # Menghasilkan salinan dari vektor ini
        return Vector(self.x, self.y)
    
    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# Membuat vector vb
vb = Vector(3, 7)

# Menyalin vb ke va menggunakan method copy()
va = vb.copy()

# Menampilkan hasil salinan
print("Vektor vb:", vb)
print("Hasil copy (disimpan di va):", va)
````

KOORDINAT X DAN Y , MENJADI SATU KESATUAN PADA VECTOR (X,Y) PENJUMLAHAN VECTOR DENGAN VELOCITY (5,0)

````python
# BOLA BERGERAK DENGAN TEXT DAN BOLAK BALIK SUMBU X >> VECTOR
import tkinter as tk
from vector import Vector

WIDTH = 400
HEIGHT = 300

x, y, radius = 100, 100, 10

class Ball:
    def __init__(self, canvas, x, y, radius):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(5, 0)  # Hanya bergerak pada sumbu X
        self.radius = radius

        self.id = canvas.create_oval(
            self.position.x - self.radius, self.position.y - self.radius,
            self.position.x + self.radius, self.position.y + self.radius,
            fill="gray", outline="black", width=2
        )

        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')

    def move(self):
        # Pantulan dari tepi
        if self.position.x >= WIDTH - self.radius or self.position.x <= self.radius:
            self.velocity.x *= -1

        self.position.add(self.velocity)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )

        self.canvas.itemconfig(self.text_pos, text=f"Posisi: {self.position} , dx: {self.velocity.x:.1f}")

class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.bola = Ball(self.canvas, x, y, radius)
        self.animate()

    def animate(self):
        self.bola.move()
        self.bola.display()
        self.root.after(30, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("BOLA BERGERAK DENGAN TEXT DAN BOLAK BALIK SUMBU X")
    app = App(root)
    root.mainloop()
````
