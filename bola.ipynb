```python
import tkinter as tk
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas):
        self.canvas = canvas
        self.position = Vector(WIDTH / 2, 50)
        self.radius = 15
        self.id = canvas.create_oval(
            self.position.x - self.radius, self.position.y - self.radius,
            self.position.x + self.radius, self.position.y + self.radius,
            fill="gray", outline="black", width=2
        )

        # Info teks
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )

         # Tampilkan info
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f})")

class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()
        self.mover = Mover(self.canvas)
        self.animate()

    def animate(self):
        self.mover.display()
        self.root.after(30, self.animate)


if __name__ == "__main__":
    root = tk.Tk()
    root.title(" Ball (Statis)")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas):
        self.canvas = canvas
        self.mass = 1
        self.position = Vector(WIDTH / 2, 30)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.radius = 24  # 48 diameter
        self.id = canvas.create_oval(
            self.position.x - self.radius, self.position.y - self.radius,
            self.position.x + self.radius, self.position.y + self.radius,
            fill="gray", outline="black", width=2
        )
        # Info teks
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_acc = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')

    def apply_force(self, force):
        f = Vector.dived(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.last_acceleration = self.acceleration.copy()
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )
         # Tampilkan info
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: {self.velocity}")
        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: {self.last_acceleration}")


    def check_edges(self):
        damping = 0.9  # redam kecepatan saat pantulan (0.0 - 1.0)
        # redaman energi (gesekan udara, gesekan lantai dll), dan di simulasi diwakili dengan damping < 1.
        if self.position.x > WIDTH:
            self.position.x = WIDTH
            self.velocity.x *= -1 * damping
        elif self.position.x < 0:
            self.position.x = 0
            self.velocity.x *= -1 * damping

        if self.position.y > HEIGHT:
            self.position.y = HEIGHT
            self.velocity.y *= -1 * damping


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.mover = Mover(self.canvas)

        self.animate()

    def animate(self):
        gravity = Vector(0, 0.05) # >> hanya menyebabkan bola mengarah ke bawah
        self.mover.apply_force(gravity)
        self.mover.update()
        self.mover.check_edges()
        self.mover.display()

        self.root.after(16, self.animate)  # ~60 FPS


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Forces Simulation - Nature of Code")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas):
        self.canvas = canvas
        self.mass = 1
        self.position = Vector(WIDTH / 2, 30)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.radius = 24  # 48 diameter
        self.id = canvas.create_oval(
            self.position.x - self.radius, self.position.y - self.radius,
            self.position.x + self.radius, self.position.y + self.radius,
            fill="gray", outline="black", width=2
        )
        # Info teks
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_acc = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')
        
    def apply_force(self, force):
        f = Vector.dived(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.last_acceleration = self.acceleration.copy()
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )

         # Tampilkan info
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: {self.velocity}")
        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: {self.last_acceleration}")


    def check_edges(self):
        if self.position.x > WIDTH:
            self.position.x = WIDTH
            self.velocity.x *= -1
        elif self.position.x < 0:
            self.position.x = 0
            self.velocity.x *= -1

        if self.position.y > HEIGHT:
            self.position.y = HEIGHT
            self.velocity.y *= -1


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.mover = Mover(self.canvas)

        self.wind_active = False
        self.canvas.bind("<ButtonPress-1>", self.on_mouse_press)
        self.canvas.bind("<ButtonRelease-1>", self.on_mouse_release)

        self.animate()

    def on_mouse_press(self, event):
        self.wind_active = True

    def on_mouse_release(self, event):
        self.wind_active = False

    def animate(self):
        gravity = Vector(0, 0.1)
        self.mover.apply_force(gravity)

        if self.wind_active:
            wind = Vector(0.1, 0)
            self.mover.apply_force(wind)

        self.mover.update()
        self.mover.check_edges()
        self.mover.display()

        self.root.after(16, self.animate)  # ~60 FPS


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Forces Simulation - Nature of Code")
    app = App(root)
    root.mainloop()
```
