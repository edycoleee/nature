```python
import tkinter as tk
import random
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas, x, y, color="gray"):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(0, random.uniform(2, 5))  # Kecepatan acak ke bawah
        self.radius = 15
        self.color = color

        self.id = canvas.create_oval(
            self.position.x - self.radius, self.position.y - self.radius,
            self.position.x + self.radius, self.position.y + self.radius,
            fill=color, outline="black", width=2
        )

        self.text_pos = canvas.create_text(x + 20, y - 10, anchor='nw', text="", font=('Arial', 8), fill='black')
        self.text_vel = canvas.create_text(x + 20, y + 10, anchor='nw', text="", font=('Arial', 8), fill='blue')

    def update(self):
        self.position.add(self.velocity)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )
        self.canvas.coords(self.text_pos, x + 20, y - 10)
        self.canvas.coords(self.text_vel, x + 20, y + 10)

        self.canvas.itemconfig(self.text_pos, text=f"Pos: ({x:.1f},{y:.1f})")
        self.canvas.itemconfig(self.text_vel, text=f"Vel: {self.velocity}")

    def check_edges(self):
        if self.position.y + self.radius >= HEIGHT or self.position.y - self.radius <= 0:
            self.velocity.y *= -1  # Pantulan


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        # Buat dua mover dengan posisi acak
        self.movers = [
            Mover(self.canvas, random.randint(50, WIDTH-50), random.randint(30, HEIGHT-60), color="gray"),
            Mover(self.canvas, random.randint(50, WIDTH-50), random.randint(30, HEIGHT-60), color="orange")
        ]

        self.animate()

    def animate(self):
        for mover in self.movers:
            mover.update()
            mover.check_edges()
            mover.display()

        self.root.after(30, self.animate)


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Multiple Bouncing Balls")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas):
        self.canvas = canvas
        self.mass = 1
        self.position = Vector(WIDTH / 2, 30)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.radius = 24  # 48 diameter
        self.id = canvas.create_oval(
            self.position.x - self.radius, self.position.y - self.radius,
            self.position.x + self.radius, self.position.y + self.radius,
            fill="gray", outline="black", width=2
        )
        # Info teks
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_acc = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')

    def apply_force(self, force):
        f = Vector.dived(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.last_acceleration = self.acceleration.copy()
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )
         # Tampilkan info
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: {self.velocity}")
        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: {self.last_acceleration}")


    def check_edges(self):
        damping = 0.9  # redam kecepatan saat pantulan (0.0 - 1.0)
        # redaman energi (gesekan udara, gesekan lantai dll), dan di simulasi diwakili dengan damping < 1.
        if self.position.x > WIDTH:
            self.position.x = WIDTH
            self.velocity.x *= -1 * damping
        elif self.position.x < 0:
            self.position.x = 0
            self.velocity.x *= -1 * damping

        if self.position.y > HEIGHT:
            self.position.y = HEIGHT
            self.velocity.y *= -1 * damping


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.mover = Mover(self.canvas)

        self.animate()

    def animate(self):
        gravity = Vector(0, 0.05) # >> hanya menyebabkan bola mengarah ke bawah
        self.mover.apply_force(gravity)
        self.mover.update()
        self.mover.check_edges()
        self.mover.display()

        self.root.after(16, self.animate)  # ~60 FPS


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Forces Simulation - Nature of Code")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas, x, y, mass, color="gray"):
        self.canvas = canvas
        self.mass = mass
        self.radius = mass * 8
        self.position = Vector(x, y)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.color = color

        self.id = canvas.create_oval(
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius,
            fill=color, outline="black", width=2
        )

    def apply_force(self, force):
        f = Vector.dived(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )

    def check_edges(self):
        if self.position.x > WIDTH - self.radius:
            self.position.x = WIDTH - self.radius
            self.velocity.x *= -1
        elif self.position.x < self.radius:
            self.position.x = self.radius
            self.velocity.x *= -1

        if self.position.y > HEIGHT - self.radius:
            self.position.y = HEIGHT - self.radius
            self.velocity.y *= -1


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.moverA = Mover(self.canvas, 200, 30, 10, "skyblue")
        self.moverB = Mover(self.canvas, 440, 30, 2, "tomato")

        self.wind_active = False
        self.canvas.bind("<ButtonPress-1>", self.on_mouse_press)
        self.canvas.bind("<ButtonRelease-1>", self.on_mouse_release)

        self.animate()

    def on_mouse_press(self, event):
        self.wind_active = True

    def on_mouse_release(self, event):
        self.wind_active = False

    def animate(self):
        gravity = Vector(0, 0.1)

        # Gravity scaled by mass
        gravityA = gravity.multed(self.moverA.mass)
        gravityB = gravity.multed(self.moverB.mass)
        self.moverA.apply_force(gravityA)
        self.moverB.apply_force(gravityB)

        if self.wind_active:
            wind = Vector(0.1, 0)
            self.moverA.apply_force(wind)
            self.moverB.apply_force(wind)

        for mover in [self.moverA, self.moverB]:
            mover.update()
            mover.check_edges()
            mover.display()

        self.root.after(16, self.animate)  # ~60 FPS


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Example 2.3 - Gravity Scaled by Mass")
    app = App(root)
    root.mainloop()

    #gravity * mass → total gaya gravitasi meningkat dengan massa, sehingga semua benda jatuh dengan percepatan yang sama.
```

```python
import tkinter as tk
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas, x, y, mass):
        self.canvas = canvas
        self.mass = mass
        self.radius = mass * 8
        self.position = Vector(x, y)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)

        self.id = canvas.create_oval(
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius,
            fill="gray", outline="black", width=2
        )
        # Info teks
        self.text_pos = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_acc = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')

    def apply_force(self, force):
        f = Vector.dived(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.last_acceleration = self.acceleration.copy()
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )
         # Tampilkan info
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({x:.1f}, {y:.1f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: {self.velocity}")
        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: {self.last_acceleration}")

    def contact_edge(self):
        return self.position.y > HEIGHT - self.radius - 1

    def bounce_edges(self):
        bounce = -0.9
        if self.position.x > WIDTH - self.radius:
            self.position.x = WIDTH - self.radius
            self.velocity.x *= bounce
        elif self.position.x < self.radius:
            self.position.x = self.radius
            self.velocity.x *= bounce

        if self.position.y > HEIGHT - self.radius:
            self.position.y = HEIGHT - self.radius
            self.velocity.y *= bounce


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.mover = Mover(self.canvas, WIDTH / 2, 30, 5)
        self.wind_active = False

        self.canvas.bind("<ButtonPress-1>", self.on_mouse_press)
        self.canvas.bind("<ButtonRelease-1>", self.on_mouse_release)

        self.animate()

    def on_mouse_press(self, event):
        self.wind_active = True

    def on_mouse_release(self, event):
        self.wind_active = False

    def animate(self):
        gravity = Vector(0, 1)
        self.mover.apply_force(gravity)

        if self.wind_active:
            wind = Vector(0.5, 0)
            self.mover.apply_force(wind)

        if self.mover.contact_edge():
            c = 0.1  # coefficient of friction
            friction = self.mover.velocity.copy()
            friction.mult(-1)
            friction.set_mag(c)
            self.mover.apply_force(friction)

        self.mover.bounce_edges()
        self.mover.update()
        self.mover.display()

        self.root.after(16, self.animate)


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Example 2.4 - Including Friction")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
import random
from vector import Vector

WIDTH = 640
HEIGHT = 240

class Liquid:
    def __init__(self, canvas, x, y, w, h, c):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.w = w
        self.h = h
        self.c = c
        self.id = canvas.create_rectangle(x, y, x + w, y + h, fill="#cceeff", width=0)

    def contains(self, mover):
        pos = mover.position
        return self.x < pos.x < self.x + self.w and self.y < pos.y < self.y + self.h

    def calculate_drag(self, mover):
        speed = mover.velocity.mag()
        drag_magnitude = self.c * speed * speed

        drag = mover.velocity.copy()
        drag.mult(-1)
        drag.set_mag(drag_magnitude)
        return drag

class Mover:
    def __init__(self, canvas, x, y, mass):
        self.canvas = canvas
        self.mass = mass
        self.radius = self.mass * 8
        self.position = Vector(x, y)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.id = canvas.create_oval(
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius,
            fill="gray", outline="black", width=2
        )

    def apply_force(self, force):
        f = Vector.div(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        r = self.radius
        self.canvas.coords(self.id, x - r, y - r, x + r, y + r)

    def check_edges(self):
        if self.position.y > HEIGHT - self.radius:
            self.velocity.y *= -0.9
            self.position.y = HEIGHT - self.radius

class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.liquid = Liquid(self.canvas, 0, HEIGHT/2, WIDTH, HEIGHT/2, 0.1)
        self.movers = []
        self.reset()

        self.canvas.bind("<Button-1>", lambda event: self.reset())
        self.animate()

    def reset(self):
        self.movers.clear()
        for i in range(9):
            mass = random.uniform(0.5, 3)
            x = 40 + i * 65
            mover = Mover(self.canvas, x, 0, mass)
            self.movers.append(mover)

    def animate(self):
        for mover in self.movers:
            if self.liquid.contains(mover):
                drag = self.liquid.calculate_drag(mover)
                mover.apply_force(drag)

            gravity = Vector(0, 0.1 * mover.mass)
            mover.apply_force(gravity)

            mover.update()
            mover.check_edges()
            mover.display()

        self.root.after(16, self.animate)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Fluid Resistance - Nature of Code")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector
import random

WIDTH = 640
HEIGHT = 240

class Mover:
    def __init__(self, canvas, x, y, mass):
        self.canvas = canvas
        self.mass = mass
        self.radius = mass * 8
        self.position = Vector(x, y)
        self.velocity = Vector(1, 0)
        self.acceleration = Vector(0, 0)

        self.id = canvas.create_oval(
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius,
            fill="gray", outline="black"
        )

    def apply_force(self, force):
        f = Vector.dived(force, self.mass)
        self.acceleration.add(f)

    def update(self):
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.mult(0)

    def display(self):
        x, y = self.position.x, self.position.y
        self.canvas.coords(
            self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )


class Attractor:
    def __init__(self, canvas):
        self.canvas = canvas
        self.mass = 20
        self.G = 1
        self.position = Vector(WIDTH/2, HEIGHT/2)
        self.dragOffset = Vector(0, 0)
        self.dragging = False
        self.rollover = False

        self.id = canvas.create_oval(
            self.position.x - self.mass, self.position.y - self.mass,
            self.position.x + self.mass, self.position.y + self.mass,
            fill="lightgray", outline="black", width=2
        )

    def attract(self, mover):
        force = Vector.subbed(self.position, mover.position)
        distance = force.mag()
        distance = max(5, min(distance, 25))
        strength = (self.G * self.mass * mover.mass) / (distance * distance)
        force.set_mag(strength)
        return force

    def display(self):
        color = "black" if self.dragging else ("gray" if self.rollover else "lightgray")
        self.canvas.coords(
            self.id,
            self.position.x - self.mass, self.position.y - self.mass,
            self.position.x + self.mass, self.position.y + self.mass
        )
        self.canvas.itemconfig(self.id, fill=color)

    def handle_hover(self, mx, my):
        d = ((mx - self.position.x) ** 2 + (my - self.position.y) ** 2) ** 0.5
        self.rollover = d < self.mass

    def handle_press(self, mx, my):
        d = ((mx - self.position.x) ** 2 + (my - self.position.y) ** 2) ** 0.5
        if d < self.mass:
            self.dragging = True
            self.dragOffset = Vector(self.position.x - mx, self.position.y - my)

    def handle_drag(self, mx, my):
        if self.dragging:
            self.position.x = mx + self.dragOffset.x
            self.position.y = my + self.dragOffset.y

    def stop_dragging(self):
        self.dragging = False


class App:
    def __init__(self, root):
        self.root = root
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
        self.canvas.pack()

        self.movers = [Mover(self.canvas, random.uniform(0, WIDTH), random.uniform(0, HEIGHT), random.uniform(0.5, 3))]
        self.attractor = Attractor(self.canvas)

        self.canvas.bind("<Motion>", self.mouse_moved)
        self.canvas.bind("<ButtonPress-1>", self.mouse_pressed)
        self.canvas.bind("<B1-Motion>", self.mouse_dragged)
        self.canvas.bind("<ButtonRelease-1>", self.mouse_released)

        self.animate()

    def animate(self):
        self.canvas.delete("info")
        self.attractor.display()

        for mover in self.movers:
            force = self.attractor.attract(mover)
            mover.apply_force(force)
            mover.update()
            mover.display()

        self.root.after(16, self.animate)

    def mouse_moved(self, event):
        self.attractor.handle_hover(event.x, event.y)

    def mouse_pressed(self, event):
        self.attractor.handle_press(event.x, event.y)

    def mouse_dragged(self, event):
        self.attractor.handle_hover(event.x, event.y)
        self.attractor.handle_drag(event.x, event.y)

    def mouse_released(self, event):
        self.attractor.stop_dragging()


if __name__ == "__main__":
    root = tk.Tk()
    root.title("Attraction with Movers - Nature of Code")
    app = App(root)
    root.mainloop()
```

```python
import tkinter as tk
from vector import Vector
import math

WIDTH, HEIGHT = 640, 240

class Body:
    def __init__(self, canvas, x, y, mass, vx, vy, color="gray"):
        self.canvas = canvas
        self.position = Vector(x, y)
        self.velocity = Vector(vx, vy)
        self.acceleration = Vector(0, 0)
        self.mass = mass
        self.radius = math.sqrt(self.mass) * 4
        self.id = canvas.create_oval(x - self.radius, y - self.radius,
                                     x + self.radius, y + self.radius,
                                     fill=color, outline="black", width=2)

    def apply_force(self, force):
        f = force.copy()
        f.div(self.mass)
        self.acceleration.add(f)

    def attract(self, other):
        force = Vector.subbed(self.position, other.position)
        d = max(5, min(force.mag(), 25))
        G = 1
        strength = (G * self.mass * other.mass) / (d * d)
        force.set_mag(strength)
        other.apply_force(force)

    def update(self):
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.set(0, 0)

    def show(self):
        x, y = self.position.x, self.position.y
        r = self.radius
        self.canvas.coords(self.id, x - r, y - r, x + r, y + r)

def animate():
    bodyA.attract(bodyB)
    bodyB.attract(bodyA)

    bodyA.update()
    bodyB.update()

    bodyA.show()
    bodyB.show()

    root.after(20, animate)

root = tk.Tk()
root.title("Two-Body Attraction")
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

bodyA = Body(canvas, 320, 40, mass=8, vx=1, vy=0, color="red")
bodyB = Body(canvas, 320, 200, mass=8, vx=-1, vy=0, color="blue")

animate()
root.mainloop()
```

```python
import tkinter as tk
import random
from vector import Vector

# Body class
class Body:
    def __init__(self, canvas, x, y, m):
        self.canvas = canvas
        self.mass = m
        self.position = Vector(x, y)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        r = self.mass * 8
        self.r = r
        self.id = canvas.create_oval(x - r, y - r, x + r, y + r, fill="gray", outline="black", width=2)

    def apply_force(self, force):
        f = force.copy()
        f.div(self.mass)
        self.acceleration.add(f)

    def update(self):
        self.velocity.add(self.acceleration)
        self.position.add(self.velocity)
        self.acceleration.set(0, 0)

    def show(self):
        x = self.position.x
        y = self.position.y
        r = self.r
        self.canvas.coords(self.id, x - r, y - r, x + r, y + r)

    def attract(self, other):
        force = Vector.subbed(self.position, other.position)
        distance = force.mag()
        distance = max(5, min(distance, 25))
        strength = (G * self.mass * other.mass) / (distance * distance)
        force.set_mag(strength)
        return force

# Global constants
WIDTH, HEIGHT = 640, 240
G = 1

# Tkinter setup
root = tk.Tk()
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

# Create bodies
bodies = []
for _ in range(10):
    x = random.uniform(0, WIDTH)
    y = random.uniform(0, HEIGHT)
    m = random.uniform(0.1, 2)
    bodies.append(Body(canvas, x, y, m))

# Animation loop
def animate():
    for i in range(len(bodies)):
        for j in range(len(bodies)):
            if i != j:
                force = bodies[j].attract(bodies[i])
                bodies[i].apply_force(force)
        bodies[i].update()
        bodies[i].show()
    root.after(20, animate)

animate()
root.mainloop()
```

```python
import tkinter as tk
import math
import random
from vector import Vector

WIDTH, HEIGHT = 600, 600
G = 20

class Attractor:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.pos = Vector(x, y)
        self.radius = 32
        self.id = canvas.create_oval(x - self.radius, y - self.radius,
                                     x + self.radius, y + self.radius, fill='black')

    def attract(self, mover):
        force = Vector.subbed(self.pos, mover.pos)
        distance = max(5, min(force.mag(), 25))
        strength = (G * 1 * mover.mass) / (distance * distance)
        force.set_mag(strength)
        return force

    def show(self):
        self.canvas.coords(self.id,
            self.pos.x - self.radius, self.pos.y - self.radius,
            self.pos.x + self.radius, self.pos.y + self.radius
        )

class Mover:
    def __init__(self, canvas, x, y, r):
        self.canvas = canvas
        self.radius = r
        self.mass = r
        self.pos = Vector(x, y)
        angle = random.uniform(0, math.pi * 2)
        self.vel = Vector(2 * math.cos(angle), 2 * math.sin(angle))
        self.acc = Vector(0, 0)
        self.id = canvas.create_oval(0, 0, 0, 0, fill='gray', outline='black')
        self.line_id = canvas.create_line(0, 0, 0, 0, fill='black', width=2)

    def apply_force(self, force):
        f = force.copy()
        f.div(self.mass)
        self.acc.add(f)

    def update(self):
        self.vel.add(self.acc)
        self.pos.add(self.vel)
        self.acc = Vector(0, 0)

    def show(self):
        x, y = self.pos.x, self.pos.y
        self.canvas.coords(self.id,
            x - self.radius, y - self.radius,
            x + self.radius, y + self.radius
        )
        heading = self.vel.heading()
        line_x = x + self.radius * math.cos(heading)
        line_y = y + self.radius * math.sin(heading)
        self.canvas.coords(self.line_id, x, y, line_x, line_y)

# Setup tkinter window
root = tk.Tk()
canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg='white')
canvas.pack()

attractor = Attractor(canvas, WIDTH // 2, HEIGHT // 2)
mover = Mover(canvas, random.randint(0, WIDTH), random.randint(0, HEIGHT), 8)

def draw():
    force = attractor.attract(mover)
    mover.apply_force(force)
    mover.update()
    mover.show()
    attractor.show()
    root.after(16, draw)

draw()
root.mainloop()
```
