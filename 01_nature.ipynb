```python
# SCRIPT 2 - Simulasi Bola Memantul (Tanpa Vektor Kecepatan)
import tkinter as tk

# 1. Kelas Vector2D untuk merepresentasikan posisi 2D ##############

# 2. Setup Tkinter window ###########################################
# titik awal bola
x = 100
y = 100
# kecepatan bola
xspeed = 2
yspeed = 3

window = tk.Tk()
# Judul dan ukuran window 
window.title("Bola Memantul (Tanpa Vektor Kecepatan)")
canvas = tk.Canvas(window, width=500, height=300, bg="white")
canvas.pack()

# 3. Fungsi untuk update posisi mouse dan menggambar garis ##########
def update():
    # Global variabel untuk akses di dalam fungsi
    global x, y, xspeed, yspeed
    # Hapus gambar sebelumnya
    canvas.delete("all")
    # Update posisi bola
    # Penjelasan: posisi bola = posisi sebelumnya + kecepatan
    x += xspeed
    y += yspeed

    # Memantul di pinggir
    # Penjelasan: jika bola mencapai pinggir, kecepatan dibalik
    # Jika bola mencapai pinggir kiri/kanan atau atas/bawah, kecepatan dibalik
    if x <= 0 or x >= 500:
        xspeed *= -1
    if y <= 0 or y >= 300:
        yspeed *= -1

    # Gambar bola
    canvas.create_oval(x-10, y-10, x+10, y+10, fill="blue")

    # Tampilkan nilai kecepatan di layar
    canvas.create_text(10, 10, anchor="nw",
                       text=f"xspeed: {xspeed}, yspeed: {yspeed}",
                       font=("Arial", 12), fill="black")
    
    # Tampilkan nilai koordinat di layar
    canvas.create_text(10, 40, anchor="nw",
                       text=f"x: {x}, y: {y}",
                       font=("Arial", 12), fill="black")
    # Panggil fungsi update lagi setelah 20ms
    window.after(20, update)

# 4. Jalankan #######################################################
update()
window.mainloop()
```

```python
# SCRIPT 3 - Simulasi Pergerakan Titik 2D dengan Kelas Vektor
# --------------------------------------------
# Simulasi pergerakan titik 2D menggunakan kelas Vector2D
# Titik bergerak berdasarkan vektor kecepatan
# --------------------------------------------

# Kelas Vector2D untuk merepresentasikan posisi atau kecepatan 2D
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Fungsi untuk menambahkan vektor lain ke vektor ini (penjumlahan vektor)
    def add(self, other):
        self.x += other.x
        self.y += other.y

    # Fungsi untuk mengembalikan posisi sebagai tuple
    def get_position(self):
        return (self.x, self.y)

# Inisialisasi posisi awal dan kecepatan sebagai vektor
position = Vector2D(100, 100)
velocity = Vector2D(2, 3)

# Cetak informasi awal
print(f"Posisi awal: ({position.x}, {position.y})")
print(f"Kecepatan per langkah: ({velocity.x}, {velocity.y})\n")

# Simpan semua koordinat setelah tiap langkah
positions = []

# Lakukan 10 kali update posisi
for i in range(10):
    position.add(velocity)  # tambahkan kecepatan ke posisi
    positions.append(position.get_position())  # simpan hasil posisi

# Cetak hasil
print("10 Koordinat hasil gerakan:")
for i, (px, py) in enumerate(positions):
    print(f"Langkah {i+1}: posisi = ({px}, {py})")
```

```python
# SCRIPT 4 - Bola Memantul (Dengan Kelas Vektor)
import tkinter as tk

# Kelas Vector2D sederhana
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self, other):
        self.x += other.x
        self.y += other.y

# Inisialisasi posisi dan kecepatan
position = Vector2D(100, 100)
velocity = Vector2D(2, 3)

# Ukuran kanvas dan bola
canvas_width = 500
canvas_height = 300
ball_radius = 10

# Setup Tkinter
window = tk.Tk()
window.title("Bola Memantul (Dengan coords())")
canvas = tk.Canvas(window, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# Buat objek bola dan teks 1x
ball = canvas.create_oval(position.x - ball_radius, position.y - ball_radius,
                          position.x + ball_radius, position.y + ball_radius,
                          fill="green")
info_text = canvas.create_text(10, 10, anchor="nw", font=("Arial", 12), text="")

# Fungsi update animasi
def update():
    global position, velocity

    # Update posisi
    position.add(velocity)

    # Memantul di tepi
    if position.x - ball_radius <= 0 or position.x + ball_radius >= canvas_width:
        velocity.x *= -1
    if position.y - ball_radius <= 0 or position.y + ball_radius >= canvas_height:
        velocity.y *= -1

    # Update posisi bola (pakai coords)
    canvas.coords(ball,
                  position.x - ball_radius, position.y - ball_radius,
                  position.x + ball_radius, position.y + ball_radius)

    # Update info kecepatan dan posisi
    canvas.itemconfig(info_text, text=f"Position: ({position.x:.1f}, {position.y:.1f})\n"
                                      f"Velocity: ({velocity.x:.1f}, {velocity.y:.1f})")

    window.after(20, update)

update()
window.mainloop()

```

```python
# SCRIPT 6 - operasi-operasi vektor ini digunakan dalam berbagai konteks fisika, animasi, atau robotika
import math

class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, v):
        self.x += v.x
        self.y += v.y

    def sub(self, v):
        self.x -= v.x
        self.y -= v.y

    def mult(self, scalar):
        self.x *= scalar
        self.y *= scalar

    def div(self, scalar):
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar

    def mag(self):
        return math.sqrt(self.x**2 + self.y**2)

    def normalize(self):
        m = self.mag()
        if m != 0:
            self.div(m)

    def limit(self, max_val):
        if self.mag() > max_val:
            self.normalize()
            self.mult(max_val)

    def heading(self):
        return math.atan2(self.y, self.x)

    def copy(self):
        return Vector2D(self.x, self.y)

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"


# 1. add(), sub(), mult(), div() — Operasi Dasar Vektor
a = Vector2D(3, 4)
b = Vector2D(1, 2)
c = 3

print("# Operasi Dasar")
a.add(b)   # a = (3+1, 4+2) → (4, 6)
print("(3, 4) + (1, 2) : ", a) #(4.00, 6.00)

a.sub(b)   # a = (4-1, 6-2) → (3, 4)
print("(4, 6) - (1, 2) : ", a) # (3.00, 4.00)

a.mult(c)  # a = (3*3, 4*3) → (9, 12)
print("(3, 4) * 3 : ", a) # (9.00, 12.00)

a.div(c)   # a = (9/3, 12/3) → (3, 4)
print("(9, 12) / 3 : ", a) # (3.00, 4.00)

# 2. mag(), normalize(), limit() — Magnitudo, Normalisasi, dan Pembatasan
print("\n# Magnitudo dan Normalisasi")
#magnitudo gunanya untuk menghitung panjang vektor
a_mag = a.mag()  # Magnitudo dari (3, 4) → 5.0  
print("Magnitudo dari (3, 4) : ", round(a_mag, 2)) # 5.0
#normalisasi gunanya untuk mengubah vektor menjadi panjang 1
a.normalize()  # Normalisasi (3, 4) → (0.6, 0.8) 
print("Normalisasi dari (3, 4) : ", a) # (0.60, 0.80)
#limit() gunanya untuk membatasi panjang vektor
a.limit(1)  # Batasi vektor ke panjang 1 → (0.6, 0.8)
print("Batasi ke panjang 1 : ", a)

# 3. heading() — Arah Vektor
print("\n# Arah Vektor")
#heading() mengembalikan arah vektor dalam radian
a_heading = a.heading()  # Arah dari (0.6, 0.8) → sekitar 0.93 radian
print("Arah dari (0.6, 0.8) : ", round(a_heading, 2), "radian") # 0.93 radian

a_heading_deg = math.degrees(a_heading)
print("Arah dari (0.6, 0.8) : ", round(a_heading, 2), "radian /", round(a_heading_deg, 2), "derajat") #Arah dari (0.6, 0.8) :  0.93 radian / 53.13 derajat

# 4. copy() — Membuat Salinan Vektor
print("\n# Salinan Vektor")
a_copy = a.copy()  # Membuat salinan dari (0.6, 0.8)
print("Salinan dari (0.6, 0.8) : ", a_copy)

# 5. __str__() — Representasi String Vektor
print("\n# Representasi String")
print("Representasi string dari (0.6, 0.8) : ", str(a))

 
```

```python
# SCRIPT 6 - # Simulasi Penjumlahan Titik 2D dengan Mouse
import tkinter as tk

# 1. Kelas Vector2D untuk merepresentasikan posisi 2D ##############
class Vector2D:
    # Inisialisasi dengan koordinat x dan y
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    # Fungsi untuk menambahkan vektor lain ke vektor ini
    def add(self, v):
        self.x += v.x
        self.y += v.y
    # Fungsi untuk mengembalikan posisi sebagai tuple
    def copy(self):
        return Vector2D(self.x, self.y)
    # Untuk representasi string yang mudah dibaca
    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# 2. Setup Tkinter window ###########################################

# Ukuran window
WIDTH = 600
HEIGHT = 400

window = tk.Tk()
# Judul window
window.title("Vector Addition: Fixed Point + Mouse")
# Buat canvas untuk menggambar
canvas = tk.Canvas(window, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

# Titik tetap (fixed point) untuk penjumlahan
fixed_point = Vector2D(100, 100)
# Gambar titik tetap ke hasil penjumlahan
line_id1 = canvas.create_line(fixed_point.x, fixed_point.y, fixed_point.x, fixed_point.y,
                             arrow=tk.LAST, fill="blue", width=2)
# Gambar garis dari 0,0 ke mouse
line_id2 = canvas.create_line(fixed_point.x, fixed_point.y, fixed_point.x, fixed_point.y,
                             arrow=tk.LAST, fill="red", width=2)
# Gmbar titik mouse
mouse_dot = canvas.create_oval(0, 0, 0, 0, fill="red")
# Label untuk menampilkan informasi
info = tk.Label(window, text="", font=("Arial", 12), justify="left")
info.pack()

# 3. Fungsi untuk update posisi mouse dan menggambar garis #############
def update(event):
    # Ambil posisi mouse
    mouse = Vector2D(event.x, event.y)
    # Copy titik tetap
    result = fixed_point.copy()
    # Tambahkan posisi mouse ke titik tetap
    result.add(mouse)

    # Update titik mouse
    r = 5
    # Gambar titik mouse sebagai lingkaran kecil
    canvas.coords(mouse_dot,mouse.x - r, mouse.y - r,mouse.x + r, mouse.y + r)
    
    # Update garis dari fixed_point ke hasil penjumlahan
    canvas.coords(line_id1, fixed_point.x, fixed_point.y, result.x, result.y)

    # Update garis dari mouse ke hasil penjumlahan
    canvas.coords(line_id2, 0,0, mouse.x, mouse.y)

    # Update informasi di label
    info.config(text=(
        f"Fixed Point: {fixed_point}\n"
        f"Mouse Position: {mouse}\n"
        f"Result (Fixed + Mouse): {result}"
    ))

# 4. Jalankan ############################################################
# Bind pergerakan mouse ke fungsi update
canvas.bind("<Motion>", update)
# Jalankan aplikasi Tkinter
window.mainloop()
```

```python
import tkinter as tk
import math

# Class Vector2D untuk operasi vektor dasar
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def sub(self, v):
        self.x -= v.x
        self.y -= v.y

    def copy(self):
        return Vector2D(self.x, self.y)

    def mag(self):
        return math.sqrt(self.x**2 + self.y**2)

    def normalize(self):
        m = self.mag()
        if m != 0:
            self.x /= m
            self.y /= m

    def mult(self, scalar):
        self.x *= scalar
        self.y *= scalar

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# Setup jendela Tkinter
WIDTH = 600
HEIGHT = 400
window = tk.Tk()
window.title("Vector Subtraction: Mouse - Center")
canvas = tk.Canvas(window, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

info = tk.Label(window, text="", justify="left", font=("Arial", 12))
info.pack()

center_x, center_y = WIDTH // 2, HEIGHT // 2
max_length = 100  # panjang maksimal panah vektor (untuk normalisasi visual)

# Gambar line awal dari center ke center
line_id = canvas.create_line(center_x, center_y, center_x, center_y,
                             arrow=tk.LAST, fill="blue", width=2)

# Gambar titik tengah dan titik mouse
center_dot = canvas.create_oval(center_x-5, center_y-5, center_x+5, center_y+5,
                                fill="black")
mouse_dot = canvas.create_oval(0, 0, 0, 0, fill="red")

# Text untuk vector info di canvas
text_id = canvas.create_text(10, 10, anchor='nw', text="", fill="black", font=("Arial", 12))

# Fungsi untuk update posisi mouse dan vektor
def update(event):
    # Posisi mouse dan titik tengah
    mouse = Vector2D(event.x, event.y)
    center = Vector2D(center_x, center_y)

    # Hitung vektor arah (arah = mouse - center)
    direction = mouse.copy()
    direction.sub(center)
    raw_direction = direction.copy()  # untuk tampilan info

    # Normalisasi panjang vektor agar tidak terlalu panjang di layar
    display_vector = direction.copy()
    if display_vector.mag() > 0:
        display_vector.normalize()
        display_vector.mult(min(direction.mag(), max_length))

    # Update garis panah
    canvas.coords(line_id, center_x, center_y,
                  center_x + display_vector.x,
                  center_y + display_vector.y)

    # Update titik mouse
    r = 5
    canvas.coords(mouse_dot,
                  mouse.x - r, mouse.y - r,
                  mouse.x + r, mouse.y + r)

    # Update teks vektor di canvas
    canvas.itemconfig(text_id, text=f"Vector: {raw_direction}")

    # Info teks detail
    info_text = (
        f"Mouse position : ({mouse.x:.2f}, {mouse.y:.2f})\n"
        f"Center position: ({center_x:.2f}, {center_y:.2f})\n"
        f"Vector = Mouse - Center\n"
        f"       = ({mouse.x:.2f}, {mouse.y:.2f}) - ({center_x:.2f}, {center_y:.2f})\n"
        f"       = {raw_direction}\n"
        f"Magnitude: {raw_direction.mag():.2f}"
    )
    info.config(text=info_text)

# Event mouse digerakkan
canvas.bind("<Motion>", update)

# Jalankan aplikasi
window.mainloop()
```

```python

# sub(): digunakan untuk menghitung vektor dari mouse - fixed_point
# copy(): digunakan untuk membuat salinan dari vektor mouse sebelum melakukan operasi subtraksi
# mag(): digunakan untuk menghitung panjang (magnitudo) dari vektor hasil subtraksi

import tkinter as tk
import math

class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def sub(self, v):
        self.x -= v.x
        self.y -= v.y

    def copy(self):
        return Vector2D(self.x, self.y)

    def mag(self):
        return math.sqrt(self.x**2 + self.y**2)

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# Setup Tkinter window
window = tk.Tk()
window.title("Vector Magnitude: Fixed Point to Mouse")
canvas = tk.Canvas(window, width=600, height=400, bg="white")
canvas.pack()

fixed_point = Vector2D(50, 50)
line_id = canvas.create_line(fixed_point.x, fixed_point.y, fixed_point.x, fixed_point.y,
                             arrow=tk.LAST, fill="green", width=2)

info = tk.Label(window, text="", font=("Arial", 12), justify="left")
info.pack()

def update(event):
    mouse = Vector2D(event.x, event.y)

    # Hitung vektor dari titik tetap ke mouse
    result = mouse.copy()
    result.sub(fixed_point)
    magnitude = result.mag()

    # Gambar garis dari titik tetap ke posisi mouse
    canvas.coords(line_id, fixed_point.x, fixed_point.y, mouse.x, mouse.y)

    info.config(text=(
        f"Fixed Point: {fixed_point}\n"
        f"Mouse Position: {mouse}\n"
        f"Vector (Mouse - Fixed): {result}\n"
        f"Magnitude (Length): {magnitude:.2f} px"
    ))

canvas.bind("<Motion>", update)
window.mainloop()
```

```python
# SCRIPT 17 - Multiplying Vector by Scalar
import tkinter as tk
import math

# Class Vector2D
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, v): self.x += v.x; self.y += v.y
    def sub(self, v): self.x -= v.x; self.y -= v.y
    def mult(self, scalar): self.x *= scalar; self.y *= scalar
    def div(self, scalar): 
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar
    def mag(self): return math.sqrt(self.x**2 + self.y**2)
    def normalize(self):
        m = self.mag()
        if m != 0: self.div(m)
    def copy(self): return Vector2D(self.x, self.y)
    def __str__(self): return f"({self.x:.2f}, {self.y:.2f})"

# Fungsi gambar panah
def draw_vector(canvas, v, origin_x, origin_y, scale=1, color="green"):
    canvas.create_line(origin_x, origin_y,
                       origin_x + v.x * scale,
                       origin_y + v.y * scale,
                       arrow=tk.LAST, fill=color, width=2)

# Setup GUI
window = tk.Tk()
window.title("Multiplying Vector by Scalar")
canvas = tk.Canvas(window, width=600, height=400, bg="white")
canvas.pack()

# Label info perhitungan
info = tk.Label(window, text="", justify="left", font=("Arial", 12))
info.pack()

# Update saat mouse bergerak
def update(event=None):
    canvas.delete("all")
    mouse = Vector2D(event.x, event.y)
    center = Vector2D(300, 200)

    # Vektor dari tengah ke mouse
    direction = mouse.copy()
    direction.sub(center)

    # Gambar vektor asli
    draw_vector(canvas, direction, 300, 200, color="green")
    
    # Gambar setengahnya
    half_vector = direction.copy()
    half_vector.mult(0.5)
    draw_vector(canvas, half_vector, 300, 200, color="blue")

    # Tampilkan perhitungan di bawah
    info_text = (
        f"Mouse position: ({mouse.x:.2f}, {mouse.y:.2f})\n"
        f"Center: (300.00, 200.00)\n"
        f"Direction = Mouse - Center = ({mouse.x:.2f}, {mouse.y:.2f}) - (300.00, 200.00)\n"
        f"          = {direction}\n"
        f"Setengah arah (0.5x) = {half_vector}"
    )
    info.config(text=info_text)

canvas.bind("<Motion>", update)
window.mainloop()
```
