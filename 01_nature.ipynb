```python
# SCRIPT 2 - Simulasi Bola Memantul (Tanpa Vektor Kecepatan)
import tkinter as tk

# 1. Kelas Vector2D untuk merepresentasikan posisi 2D ##############

# 2. Setup Tkinter window ###########################################
# titik awal bola
x = 100
y = 100
# kecepatan bola
xspeed = 2
yspeed = 3

window = tk.Tk()
# Judul dan ukuran window 
window.title("Bola Memantul (Tanpa Vektor Kecepatan)")
canvas = tk.Canvas(window, width=500, height=300, bg="white")
canvas.pack()

# 3. Fungsi untuk update posisi mouse dan menggambar garis ##########
def update():
    # Global variabel untuk akses di dalam fungsi
    global x, y, xspeed, yspeed
    # Hapus gambar sebelumnya
    canvas.delete("all")
    # Update posisi bola
    # Penjelasan: posisi bola = posisi sebelumnya + kecepatan
    x += xspeed
    y += yspeed

    # Memantul di pinggir
    # Penjelasan: jika bola mencapai pinggir, kecepatan dibalik
    # Jika bola mencapai pinggir kiri/kanan atau atas/bawah, kecepatan dibalik
    if x <= 0 or x >= 500:
        xspeed *= -1
    if y <= 0 or y >= 300:
        yspeed *= -1

    # Gambar bola
    canvas.create_oval(x-10, y-10, x+10, y+10, fill="blue")

    # Tampilkan nilai kecepatan di layar
    canvas.create_text(10, 10, anchor="nw",
                       text=f"xspeed: {xspeed}, yspeed: {yspeed}",
                       font=("Arial", 12), fill="black")
    
    # Tampilkan nilai koordinat di layar
    canvas.create_text(10, 40, anchor="nw",
                       text=f"x: {x}, y: {y}",
                       font=("Arial", 12), fill="black")
    # Panggil fungsi update lagi setelah 20ms
    window.after(20, update)

# 4. Jalankan #######################################################
update()
window.mainloop()
```

```python
# SCRIPT 3 - Simulasi Pergerakan Titik 2D dengan Kelas Vektor
# --------------------------------------------
# Simulasi pergerakan titik 2D menggunakan kelas Vector2D
# Titik bergerak berdasarkan vektor kecepatan
# --------------------------------------------

# 1. Kelas Vector2D untuk merepresentasikan posisi atau kecepatan 2D
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Fungsi untuk menambahkan vektor lain ke vektor ini (penjumlahan vektor)
    def add(self, other):
        self.x += other.x
        self.y += other.y

    # Fungsi untuk mengembalikan posisi sebagai tuple
    def get_position(self):
        return (self.x, self.y)

# 2. Inisialisasi posisi awal dan kecepatan sebagai vektor
position = Vector2D(100, 100)
velocity = Vector2D(2, 3)

# Cetak informasi awal
print(f"Posisi awal: ({position.x}, {position.y})")
print(f"Kecepatan per langkah: ({velocity.x}, {velocity.y})\n")

# Simpan semua koordinat setelah tiap langkah
positions = []

# Lakukan 10 kali update posisi
for i in range(10):
    position.add(velocity)  # tambahkan kecepatan ke posisi
    positions.append(position.get_position())  # simpan hasil posisi

# Cetak hasil
print("10 Koordinat hasil gerakan:")
for i, (px, py) in enumerate(positions):
    print(f"Langkah {i+1}: posisi = ({px}, {py})")
```

```python
# SCRIPT 4 - Bola Memantul (Dengan Kelas Vektor)
import tkinter as tk

# 1. Kelas Vector2D sederhana
class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self, other):
        self.x += other.x
        self.y += other.y

# 2. Inisialisasi posisi dan kecepatan
position = Vector2D(100, 100)
velocity = Vector2D(2, 3)

# Ukuran kanvas dan bola
canvas_width = 500
canvas_height = 300
ball_radius = 10

# Setup Tkinter
window = tk.Tk()
window.title("Bola Memantul (Dengan coords())")
canvas = tk.Canvas(window, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# Buat objek bola dan teks 1x
ball = canvas.create_oval(position.x - ball_radius, position.y - ball_radius,
                          position.x + ball_radius, position.y + ball_radius,
                          fill="green")
# Buat teks untuk menampilkan informasi posisi dan kecepatan
info_text = canvas.create_text(10, 10, anchor="nw", font=("Arial", 12), text="")

# 3. Fungsi update animasi
def update():
    # Global variabel untuk akses di dalam fungsi
    global position, velocity

    # Update posisi
    position.add(velocity)

    # Memantul di tepi
    # Jika bola mencapai pinggir kiri/kanan atau atas/bawah, kecepatan dibalik
    if position.x - ball_radius <= 0 or position.x + ball_radius >= canvas_width:
        velocity.x *= -1
    if position.y - ball_radius <= 0 or position.y + ball_radius >= canvas_height:
        velocity.y *= -1

    # Update posisi bola (pakai coords)
    canvas.coords(ball,
                  position.x - ball_radius, position.y - ball_radius,
                  position.x + ball_radius, position.y + ball_radius)

    # Update info kecepatan dan posisi
    canvas.itemconfig(info_text, text=f"Position: ({position.x:.1f}, {position.y:.1f})\n"
                                      f"Velocity: ({velocity.x:.1f}, {velocity.y:.1f})")
    # Panggil update lagi setelah 20ms
    window.after(20, update)

#4. Jalankan animasi
update()
window.mainloop()

```

```python
# SCRIPT 6 - # Simulasi Penjumlahan Titik 2D center (0,0) dengan Mouse 
import tkinter as tk
import math

# ===== Vector2D Class =====
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, v):
        return Vector2D(self.x + v.x, self.y + v.y)

    def mag(self):
        return math.sqrt(self.x**2 + self.y**2)

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# ===== Setup Tkinter =====
WIDTH, HEIGHT = 600, 400
window = tk.Tk()
window.title("Simulasi Penjumlahan Vektor")
canvas = tk.Canvas(window, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()
info_label = tk.Label(window, text="", font=("Arial", 12), justify="left")
info_label.pack()

# Titik pusat (0,0) di tengah layar
center_x, center_y = WIDTH // 2, HEIGHT // 2

# ===== Update Fungsi Utama =====
def update(event):
    canvas.delete("all")

    # Titik Pusat = A mulai dari (0,0) di tengah layar
    origin = Vector2D(center_x, center_y)

    # Vektor A = (mouse.x - center_x, 0)
    A = Vector2D(event.x - center_x, 0)

    # Vektor B = (0, mouse.y - center_y)
    B = Vector2D(0, event.y - center_y)

    # Vektor C = A + B
    C = A.add(B)

    # ===== Gambar Panah =====
    # A: Horizontal dari center
    canvas.create_line(center_x, center_y,
                       center_x + A.x, center_y + A.y,
                       arrow=tk.LAST, fill="blue", width=2)

    # B: Vertikal dari ujung A
    canvas.create_line(center_x + A.x, center_y + A.y,
                       center_x + A.x + B.x, center_y + A.y + B.y,
                       arrow=tk.LAST, fill="green", width=2)

    # C: langsung dari pusat ke mouse (A+B)
    canvas.create_line(center_x, center_y,
                       center_x + C.x, center_y + C.y,
                       arrow=tk.LAST, fill="red", width=3)

    # ===== Tampilkan Info =====
    info = (
        f"Vektor A (horizontal): {A}\n"
        f"Vektor B (vertikal)  : {B}\n"
        f"Hasil A + B = C      : {C}\n"
        f"Magnitude C (|C|)    : {C.mag():.2f} px"
    )
    info_label.config(text=info)

    # Titik mouse
    r = 4
    canvas.create_oval(center_x + C.x - r, center_y + C.y - r,
                       center_x + C.x + r, center_y + C.y + r, fill="black")

# ===== Bind Mouse =====
canvas.bind("<Motion>", update)
window.mainloop()
```

```python
# SCRIPT 7 - # Simulasi Pengurangan Vektor 2D: A - B = C
import tkinter as tk
import math

# Class Vector2D
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def sub(self, v):
        self.x -= v.x
        self.y -= v.y

    def copy(self):
        return Vector2D(self.x, self.y)

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# Setup window
WIDTH = 600
HEIGHT = 400
window = tk.Tk()
window.title("Simulasi Pengurangan Vektor: A - B = C")
canvas = tk.Canvas(window, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

info = tk.Label(window, text="", justify="left", font=("Arial", 12))
info.pack()

# Titik origin di tengah
#origin_x, origin_y = WIDTH // 2, HEIGHT // 2
origin_x, origin_y = 0, 0

# Panah vektor
arrow_a = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="red", width=2)
arrow_b = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="green", width=2)
arrow_c = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="blue", width=2)

# Fungsi update saat mouse digerakkan
def update(event):
    # Vektor A: dari origin ke mouse
    a = Vector2D(event.x - origin_x, event.y - origin_y)

    # Vektor B: tetap, misalnya arah kanan atas
    b = Vector2D(100, 100)

    # Vektor C = A - B
    c = a.copy()
    c.sub(b)

    # Koordinat ujung A
    ax = origin_x + a.x
    ay = origin_y + a.y

    # Koordinat ujung B (dari origin)
    bx = origin_x + b.x
    by = origin_y + b.y

    # Koordinat ujung C = A - B
    cx = origin_x + c.x
    cy = origin_y + c.y

    # Gambar panah A
    canvas.coords(arrow_a, origin_x, origin_y, ax, ay)
    # Gambar panah B
    canvas.coords(arrow_b, origin_x, origin_y, bx, by)
    # Gambar panah C
    canvas.coords(arrow_c, origin_x, origin_y, cx, cy)

    # Update teks info
    info_text = (
        f"Vektor A (merah) : {a}\n"
        f"Vektor B (hijau) : {b}\n"
        f"Vektor C = A - B : {c}"
    )
    info.config(text=info_text)

# Bind mouse movement
canvas.bind("<Motion>", update)

# Jalankan GUI
window.mainloop()
```

```python
# SCRIPT 8 - operasi-operasi vektor ini digunakan dalam berbagai konteks fisika, animasi, atau robotika
import math

# 1. Kelas Vector2D: merepresentasikan vektor 2D dengan operasi dasar
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, v):
        self.x += v.x
        self.y += v.y

    def sub(self, v):
        self.x -= v.x
        self.y -= v.y

    def mult(self, scalar):
        self.x *= scalar
        self.y *= scalar

    def div(self, scalar):
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar

    def mag(self):
        return math.sqrt(self.x**2 + self.y**2)

    def normalize(self):
        m = self.mag()
        if m != 0:
            self.div(m)

    def limit(self, max_val):
        if self.mag() > max_val:
            self.normalize()
            self.mult(max_val)

    def heading(self):
        return math.atan2(self.y, self.x)

    def copy(self):
        return Vector2D(self.x, self.y)

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"


# 1. add(), sub(), mult(), div() — Operasi Dasar Vektor
a = Vector2D(3, 4)
b = Vector2D(1, 2)
c = 3

# Penjumlahan vektor a dan b hasilnya disimpan di a
print("# Operasi Dasar")
a.add(b)   # a = (3+1, 4+2) → (4, 6)
print("(3, 4) + (1, 2) : ", a) #(4.00, 6.00)

# Pengurangan vektor a dan b hasilnya disimpan di a
a.sub(b)   # a = (4-1, 6-2) → (3, 4)
print("(4, 6) - (1, 2) : ", a) # (3.00, 4.00)

# Perkalian vektor a dengan skalar c hasilnya disimpan di a
a.mult(c)  # a = (3*3, 4*3) → (9, 12)
print("(3, 4) * 3 : ", a) # (9.00, 12.00)

# Pembagian vektor a dengan skalar c hasilnya disimpan di a
a.div(c)   # a = (9/3, 12/3) → (3, 4)
print("(9, 12) / 3 : ", a) # (3.00, 4.00)

# 2. mag(), normalize(), limit() — Magnitudo, Normalisasi, dan Pembatasan
print("\n# Magnitudo dan Normalisasi")

# magnitudo gunanya untuk menghitung panjang vektor a
a_mag = a.mag()  # Magnitudo dari (3, 4) → 5.0  
print("Magnitudo dari (3, 4) : ", round(a_mag, 2)) # 5.0

# normalisasi vektor a untuk mengubahnya menjadi panjang 1
a.normalize()  # Normalisasi (3, 4) → (0.6, 0.8) 
print("Normalisasi dari (3, 4) : ", a) # (0.60, 0.80)

#limit() gunanya untuk membatasi panjang vektor
a.limit(1)  # Batasi vektor ke panjang 1 → (0.6, 0.8)
print("Batasi ke panjang 1 : ", a)

# 3. heading() — Arah Vektor
print("\n# Arah Vektor")
#heading() mengembalikan arah vektor dalam radian
a_heading = a.heading()  # Arah dari (0.6, 0.8) → sekitar 0.93 radian
print("Arah dari (0.6, 0.8) : ", round(a_heading, 2), "radian") # 0.93 radian

a_heading_deg = math.degrees(a_heading)
print("Arah dari (0.6, 0.8) : ", round(a_heading, 2), "radian /", round(a_heading_deg, 2), "derajat") #Arah dari (0.6, 0.8) :  0.93 radian / 53.13 derajat

# 4. copy() — Membuat Salinan Vektor
print("\n# Salinan Vektor")
a_copy = a.copy()  # Membuat salinan dari (0.6, 0.8)
print("Salinan dari (0.6, 0.8) : ", a_copy)

# 5. __str__() — Representasi String Vektor
print("\n# Representasi String")
print("Representasi string dari (0.6, 0.8) : ", str(a))
```

```python
# SCRIPT 9 - Perkalian Vektor dengan Skalar
import tkinter as tk
import math

# Kelas Vector2D dengan operasi dasar vektor
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def add(self, v):  # Menambahkan vektor
        self.x += v.x
        self.y += v.y

    def sub(self, v):  # Mengurangi vektor
        self.x -= v.x
        self.y -= v.y

    def mult(self, scalar):  # Perkalian dengan skalar
        self.x *= scalar
        self.y *= scalar

    def div(self, scalar):  # Pembagian dengan skalar
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar

    def mag(self):  # Menghitung panjang (magnitudo)
        return math.sqrt(self.x**2 + self.y**2)

    def normalize(self):  # Menormalisasi vektor (panjang jadi 1)
        m = self.mag()
        if m != 0:
            self.div(m)

    def copy(self):  # Menyalin vektor (tidak merusak asli)
        return Vector2D(self.x, self.y)

    def __str__(self):  # Format string untuk ditampilkan
        return f"({self.x:.2f}, {self.y:.2f})"

# Setup GUI Tkinter
window = tk.Tk()
window.title("Multiplying Vector by Scalar")
canvas_width = 600
canvas_height = 400
canvas = tk.Canvas(window, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# Label untuk menampilkan perhitungan
info = tk.Label(window, text="", justify="left", font=("Arial", 12))
info.pack()

# Titik pusat
center = Vector2D(canvas_width / 2, canvas_height / 2)

# Buat dua panah garis sekali saja (tidak dihapus tiap frame)
original_vector_id = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="green", width=2)  #: Garis vektor asli
scaled_vector_id = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="blue", width=2)    #: Garis vektor hasil skalar

# Fungsi update saat mouse bergerak
def update(event=None):
    # Ambil posisi mouse
    mouse = Vector2D(event.x, event.y)

    # Hitung vektor dari titik tengah ke posisi mouse
    direction = mouse.copy()
    direction.sub(center)  #: Tahap 1: Hitung vektor arah (mouse - center)

    # Gambar vektor asli (hijau)
    canvas.coords(original_vector_id,
                  center.x, center.y,
                  center.x + direction.x,
                  center.y + direction.y)  #: Tahap 2: Gambar arah asli

    # Skala vektor (misal dikali 1.5)
    scaled = direction.copy()
    scaled.mult(1.5)  #: Tahap 3: Kalikan vektor dengan skalar

    # Gambar vektor hasil skalar (biru)
    canvas.coords(scaled_vector_id,
                  center.x, center.y,
                  center.x + scaled.x,
                  center.y + scaled.y)  #: Tahap 4: Gambar arah hasil skalar

    # Tampilkan informasi perhitungan di bawah
    info_text = (
        f"Mouse position: ({mouse.x:.2f}, {mouse.y:.2f})\n"
        f"Center: ({center.x:.2f}, {center.y:.2f})\n"
        f"Direction = Mouse - Center = {direction}\n"
        f"Scaled Vector (1.5x) = {scaled}"
    )
    info.config(text=info_text)  #: Tahap 5: Tampilkan teks perhitungan

# Jalankan update setiap mouse bergerak
canvas.bind("<Motion>", update)

window.mainloop()
```

```python
# SCRIPT 10 - Pembagian Vektor dengan Skalar
import tkinter as tk
import math

# Kelas Vector2D untuk operasi dasar vektor
class Vector2D:
    def __init__(self, x=0, y=0):  #: Tahap 0: Inisialisasi vektor
        self.x = x
        self.y = y

    def add(self, v): self.x += v.x; self.y += v.y  #: Menambahkan vektor
    def sub(self, v): self.x -= v.x; self.y -= v.y  #: Mengurangi vektor
    def mult(self, scalar): self.x *= scalar; self.y *= scalar  #: Perkalian skalar
    def div(self, scalar):  #: Pembagian skalar
        if scalar != 0:
            self.x /= scalar
            self.y /= scalar

    def mag(self): return math.sqrt(self.x**2 + self.y**2)  #: Magnitudo (panjang)
    def normalize(self):
        m = self.mag()
        if m != 0: self.div(m)  #: Normalisasi (panjang 1)

    def copy(self): return Vector2D(self.x, self.y)  #: Salin vektor
    def __str__(self): return f"({self.x:.2f}, {self.y:.2f})"  #: Tampilkan teks

# Setup GUI Tkinter
window = tk.Tk()
window.title("Division Vector by Scalar")
canvas_width = 600
canvas_height = 400
canvas = tk.Canvas(window, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# Label untuk menampilkan info perhitungan
info = tk.Label(window, text="", justify="left", font=("Arial", 12))
info.pack()

# Titik tengah layar sebagai pusat vektor
center = Vector2D(canvas_width / 2, canvas_height / 2)

# Buat dua panah garis hanya sekali (pakai coords nanti)
original_vector_id = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="green", width=2)  #: Tahap 1: Garis asli
divided_vector_id = canvas.create_line(0, 0, 0, 0, arrow=tk.LAST, fill="blue", width=2)    #: Tahap 2: Garis hasil pembagian

# Fungsi update saat mouse bergerak
def update(event=None):
    # Ambil posisi mouse
    mouse = Vector2D(event.x, event.y)

    # Hitung arah dari center ke mouse
    direction = mouse.copy()
    direction.sub(center)  #: Tahap 3: Hitung vektor arah = mouse - center

    # Gambar vektor asli (hijau)
    canvas.coords(original_vector_id,
                  center.x, center.y,
                  center.x + direction.x,
                  center.y + direction.y)  #: Tahap 4: Perbarui garis asli

    # Bagi vektor dengan skalar (misal 2)
    divided = direction.copy()
    divided.div(2)  #: Tahap 5: Bagi arah dengan 2

    # Gambar hasil pembagian (biru)
    canvas.coords(divided_vector_id,
                  center.x, center.y,
                  center.x + divided.x,
                  center.y + divided.y)  #: Tahap 6: Perbarui garis hasil pembagian

    # Tampilkan informasi hasil perhitungan
    info_text = (
        f"Mouse position: ({mouse.x:.2f}, {mouse.y:.2f})\n"
        f"Center: ({center.x:.2f}, {center.y:.2f})\n"
        f"Direction = Mouse - Center = {direction}\n"
        f"Half Direction (÷2) = {divided}"
    )
    info.config(text=info_text)  #: Tahap 7: Tampilkan teks hasil

# Jalankan update saat mouse bergerak
canvas.bind("<Motion>", update)

# Jalankan aplikasi
window.mainloop()
```

```python
# SCRIPT 13 - Simulasi Limit dan Normalisasi Vektor
import tkinter as tk
import math

# ====== Vector2D class ======
class Vector2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def mag(self):
        return math.sqrt(self.x**2 + self.y**2)

    def normalize(self):
        m = self.mag()
        if m != 0:
            return Vector2D(self.x / m, self.y / m)
        return Vector2D(0, 0)

    def limit(self, max_length):
        if self.mag() > max_length:
            return self.normalize().mult(max_length)
        return self.copy()

    def copy(self):
        return Vector2D(self.x, self.y)

    def mult(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)

    def __str__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

# ====== Setup Tkinter ======
WIDTH = 600
HEIGHT = 400
window = tk.Tk()
window.title("Simulasi Limit dan Normalisasi Vektor")
canvas = tk.Canvas(window, width=WIDTH, height=HEIGHT, bg="white")
canvas.pack()

info_label = tk.Label(window, text="", justify="left", font=("Arial", 12))
info_label.pack()

origin_x = WIDTH // 2
origin_y = HEIGHT // 2
arrow_length = 50
max_vector_length = 100  # batas maksimum vektor

# ====== Update Visual ======
def update(event):
    canvas.delete("all")

    # Vektor dari pusat ke mouse
    vec_ab = Vector2D(event.x - origin_x, event.y - origin_y)
    vec_norm = vec_ab.normalize()
    vec_limited = vec_ab.limit(max_vector_length)

    mag_ab = vec_ab.mag()
    mag_norm = vec_norm.mag()
    mag_limited = vec_limited.mag()

    # ====== Garis koordinat ======
    canvas.create_line(0, origin_y, WIDTH, origin_y, fill="gray", dash=(2, 2))
    canvas.create_line(origin_x, 0, origin_x, HEIGHT, fill="gray", dash=(2, 2))

    # ====== Vektor asli ======
    canvas.create_line(origin_x, origin_y,
                       origin_x + vec_ab.x, origin_y + vec_ab.y,
                       arrow=tk.LAST, fill="green", width=3)

    # ====== Vektor dinormalisasi (panjang tetap) ======
    canvas.create_line(origin_x, origin_y,
                       origin_x + vec_norm.x * arrow_length,
                       origin_y + vec_norm.y * arrow_length,
                       arrow=tk.LAST, fill="blue", width=2)

    # ====== Vektor dibatasi (max 100 px) ======
    canvas.create_line(origin_x, origin_y,
                       origin_x + vec_limited.x, origin_y + vec_limited.y,
                       arrow=tk.LAST, fill="red", width=2)

    # Titik mouse
    canvas.create_oval(origin_x + vec_ab.x - 4, origin_y + vec_ab.y - 4,
                       origin_x + vec_ab.x + 4, origin_y + vec_ab.y + 4,
                       fill="red")

    # ====== Bar Graph Magnitude ======
    bar_x = 100
    bar_y = HEIGHT - 200
    canvas.create_text(bar_x, bar_y - 130, text="|AB|", font=("Arial", 12))
    canvas.create_rectangle(bar_x, bar_y - 110,
                             bar_x + min(mag_ab, 200), bar_y - 85, fill="green")
    canvas.create_text(bar_x + min(mag_ab, 200) + 40, bar_y - 98,
                       text=f"{mag_ab:.2f}", font=("Arial", 12))

    canvas.create_text(bar_x, bar_y - 75, text="|Normalized|", font=("Arial", 12))
    canvas.create_rectangle(bar_x, bar_y - 55,
                             bar_x + mag_norm * 50, bar_y - 30, fill="blue")
    canvas.create_text(bar_x + 50 + 40, bar_y - 43,
                       text=f"{mag_norm:.2f}", font=("Arial", 12))

    canvas.create_text(bar_x, bar_y - 20, text="|Limited|", font=("Arial", 12))
    canvas.create_rectangle(bar_x, bar_y,
                             bar_x + mag_limited, bar_y + 25, fill="red")
    canvas.create_text(bar_x + mag_limited + 40, bar_y + 13,
                       text=f"{mag_limited:.2f}", font=("Arial", 12))

    # ====== Info Text ======
    info = (
        f"Vector AB: {vec_ab}\n"
        f"Magnitude AB: {mag_ab:.2f}\n\n"
        f"Normalized AB: {vec_norm}\n"
        f"Magnitude Normalized: {mag_norm:.2f}\n\n"
        f"Limited Vector AB: {vec_limited}\n"
        f"Magnitude Limited: {mag_limited:.2f}"
    )
    info_label.config(text=info)

canvas.bind("<Motion>", update)
window.mainloop()
```

```python
#SCRIPT 14 - Simulasi Gerakan satu Bola dengan Vector Velocity
import tkinter as tk
import random

# Kelas Vector 2D sederhana
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self, other):
        self.x += other.x
        self.y += other.y

    def copy(self):
        return Vector(self.x, self.y)

# Kelas Mover (bola yang bergerak)
class Mover:
    def __init__(self, canvas, width, height, info_label):
        self.canvas = canvas
        self.width = width
        self.height = height
        self.radius = 10
        self.info_label = info_label

        # Posisi dan velocity random
        self.location = Vector(random.randint(100, width-100), random.randint(100, height-100))
        self.velocity = Vector(random.choice([-2, -1, 1, 2]), random.choice([-2, -1, 1, 2]))

        self.id = canvas.create_oval(self.location.x - self.radius, self.location.y - self.radius,
                                     self.location.x + self.radius, self.location.y + self.radius,
                                     fill="blue")

    def update(self):
        # Salin posisi sebelum bergerak
        prev_location = self.location.copy()

        # Tambah lokasi dengan velocity
        self.location.add(self.velocity)
        self.check_edges()

        # Update posisi bola di canvas
        self.canvas.coords(self.id,
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius)

        # Tampilkan info perhitungan
        info_text = (
            f"Velocity: (vx = {self.velocity.x}, vy = {self.velocity.y})\n"
            f"Lokasi Sebelumnya: ({prev_location.x}, {prev_location.y})\n"
            f"Perhitungan:\n"
            f"  lokasi_x = {prev_location.x} + {self.velocity.x} = {self.location.x}\n"
            f"  lokasi_y = {prev_location.y} + {self.velocity.y} = {self.location.y}"
        )
        self.info_label.config(text=info_text)

    def check_edges(self):
        # Jika keluar layar, lompat ke sisi lain
        if self.location.x > self.width:
            self.location.x = 0
        elif self.location.x < 0:
            self.location.x = self.width
        if self.location.y > self.height:
            self.location.y = 0
        elif self.location.y < 0:
            self.location.y = self.height

# Fungsi utama
def main():
    WIDTH = 640
    HEIGHT = 360

    root = tk.Tk()
    root.title("Simulasi Gerakan dengan Vector Velocity")

    canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
    canvas.pack()

    info_label = tk.Label(root, text="", justify="left", font=("Arial", 11))
    info_label.pack()

    mover = Mover(canvas, WIDTH, HEIGHT, info_label)

    def animate():
        mover.update()
        root.after(100, animate)  # Update tiap 100 ms agar mudah dilihat

    animate()
    root.mainloop()

main()
```

```python
# SCRIPT 19 - Simulasi Pergerakan Bola dengan Perlin Noise
import tkinter as tk
import math
import random
from perlin_noise import PerlinNoise  # Pakai modul perlin_noise

# -----------------------------
# Kelas Vector: untuk posisi, kecepatan, percepatan
# -----------------------------
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self, v):
        self.x += v.x
        self.y += v.y

    def mult(self, scalar):
        self.x *= scalar
        self.y *= scalar

    def limit(self, max_val):
        mag = math.sqrt(self.x**2 + self.y**2)
        if mag > max_val:
            self.x = (self.x / mag) * max_val
            self.y = (self.y / mag) * max_val

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2)

# -----------------------------
# Kelas Mover: Bola biru yang bergerak
# -----------------------------
class Mover:
    def __init__(self, canvas, width, height, noise_gen):
        self.canvas = canvas
        self.noise = noise_gen
        self.width = width
        self.height = height
        self.radius = 15

        self.location = Vector(width // 2, height // 2)
        self.velocity = Vector(0, 0)
        self.acceleration = Vector(0, 0)
        self.topspeed = 5

        self.noise_offset = random.uniform(0, 1000)

        self.id = canvas.create_oval(
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius,
            fill="blue"
        )

        # Teks perhitungan di layar
        self.text_acc = canvas.create_text(10, 10, anchor="nw", text="", font=("Arial", 10), fill="black")
        self.text_vel = canvas.create_text(10, 30, anchor="nw", text="", font=("Arial", 10), fill="blue")
        self.text_pos = canvas.create_text(10, 50, anchor="nw", text="", font=("Arial", 10), fill="red")

    def update(self, t):
        # 1. Dapatkan angka halus dari Perlin noise (hasil 0.0 - 1.0)
        noise_val = self.noise(t + self.noise_offset)
        angle = noise_val * 2 * math.pi  # ubah jadi sudut (0 - 2π)

        # 2. Buat vektor percepatan dari sudut
        self.acceleration = Vector(math.cos(angle), math.sin(angle))
        self.acceleration.mult(0.5)

        # 3. Tambahkan percepatan ke kecepatan
        self.velocity.add(self.acceleration)
        self.velocity.limit(self.topspeed)

        # 4. Tambahkan kecepatan ke posisi
        self.location.add(self.velocity)
        self.check_edges()

        # 5. Gambar ulang bola
        self.canvas.coords(self.id,
            self.location.x - self.radius, self.location.y - self.radius,
            self.location.x + self.radius, self.location.y + self.radius)

        # 6. Tampilkan nilai-nilai di layar
        acc_mag = self.acceleration.magnitude()
        vel_mag = self.velocity.magnitude()
        pos_x, pos_y = self.location.x, self.location.y

        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: {acc_mag:.2f}")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: {vel_mag:.2f}")
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({pos_x:.1f}, {pos_y:.1f})")

    def check_edges(self):
        # Jika bola keluar layar, masuk dari sisi sebaliknya
        if self.location.x > self.width:
            self.location.x = 0
        elif self.location.x < 0:
            self.location.x = self.width
        if self.location.y > self.height:
            self.location.y = 0
        elif self.location.y < 0:
            self.location.y = self.height

# -----------------------------
# Fungsi utama
# -----------------------------
def main():
    WIDTH, HEIGHT = 600, 400
    root = tk.Tk()
    root.title("Simulasi Perlin Noise - Bola Bergerak")
    canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="white")
    canvas.pack()

    noise = PerlinNoise(octaves=1)  # Buat generator Perlin noise
    mover = Mover(canvas, WIDTH, HEIGHT, noise)
    t = [0.0]  # Waktu sebagai input noise

    def animate():
        mover.update(t[0])
        t[0] += 0.01  # Tambah waktu perlahan agar perubahan halus
        root.after(30, animate)

    animate()
    root.mainloop()

main()
```

```python
# SCRIPT 24 - Simulasi Bola Mengejar Mouse dengan Percepatan
import tkinter as tk
import math

class Bola:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.location = [x, y]          # Posisi bola (x, y)
        self.velocity = [0, 0]          # Kecepatan awal (vx, vy)
        self.acceleration = [0, 0]      # Percepatan awal (ax, ay)
        self.radius = 10
        self.topspeed = 10              # Kecepatan maksimum

        # Gambar bola di layar
        self.id = canvas.create_oval(x - self.radius, y - self.radius,
                                     x + self.radius, y + self.radius,
                                     fill='blue')

        # Tambahkan teks untuk menampilkan hasil perhitungan
        self.text_acc = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_pos = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')

    def update(self, mouse_x, mouse_y):
        # 🧭 Step 1: Hitung arah dari bola ke mouse
        dx = mouse_x - self.location[0]
        dy = mouse_y - self.location[1]

        # 📏 Step 2: Hitung jarak dan normalisasi arah
        dist = math.sqrt(dx**2 + dy**2)
        if dist != 0:
            dx /= dist
            dy /= dist

        # 🌀 Step 3: Skala percepatan (agar kecil)
        dx *= 0.5
        dy *= 0.5

        # 💨 Step 4: Simpan sebagai percepatan
        self.acceleration = [dx, dy]

        # ➕ Tambahkan percepatan ke kecepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        # 🚫 Batasi kecepatan maksimum
        speed = math.sqrt(self.velocity[0]**2 + self.velocity[1]**2)
        if speed > self.topspeed:
            self.velocity[0] = (self.velocity[0] / speed) * self.topspeed
            self.velocity[1] = (self.velocity[1] / speed) * self.topspeed

        # 📍 Update posisi
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        # 🔁 Gambar ulang posisi bola
        x, y = self.location
        self.canvas.coords(self.id, x - self.radius, y - self.radius, x + self.radius, y + self.radius)

        # 🧾 Tampilkan perhitungan di layar
        ax, ay = self.acceleration
        vx, vy = self.velocity
        px, py = self.location

        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: ({ax:.2f}, {ay:.2f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: ({vx:.2f}, {vy:.2f})")
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({px:.1f}, {py:.1f})")

# 🌟 Setup Tkinter
root = tk.Tk()
root.title("Simulasi Bola Mengejar Mouse")

canvas = tk.Canvas(root, width=600, height=400, bg="white")
canvas.pack()

bola = Bola(canvas, 300, 200)

# 🔁 Fungsi animasi: jalan terus setiap 16 milidetik
def animasi():
    # Dapatkan posisi mouse relatif terhadap canvas
    x = canvas.winfo_pointerx() - canvas.winfo_rootx()
    y = canvas.winfo_pointery() - canvas.winfo_rooty()
    bola.update(x, y)
    root.after(16, animasi)  # sekitar 60 frame per detik

animasi()
root.mainloop()
```

```python
# SCRIPT 26 - Simulasi Bola dengan Percepatan Variabel dan Jejak
import tkinter as tk
import math

class Bola:
    def __init__(self, canvas, x, y):
        self.canvas = canvas
        self.location = [x, y]
        self.velocity = [0, 0]
        self.acceleration = [0, 0]
        self.radius = 10
        self.topspeed = 10
        self.id = canvas.create_oval(x - self.radius, y - self.radius,
                                     x + self.radius, y + self.radius,
                                     fill='green')

        # 🌀 Trail effect
        self.trails = []  # menyimpan jejak
        self.trail_ids = []  # menyimpan id jejak yang digambar

        # Info teks
        self.text_acc = canvas.create_text(10, 10, anchor='nw', text="", font=('Arial', 10), fill='black')
        self.text_vel = canvas.create_text(10, 30, anchor='nw', text="", font=('Arial', 10), fill='blue')
        self.text_pos = canvas.create_text(10, 50, anchor='nw', text="", font=('Arial', 10), fill='red')
        self.text_dist = canvas.create_text(10, 70, anchor='nw', text="", font=('Arial', 10), fill='purple')

    def update(self, mouse_x, mouse_y):
        # Hitung arah
        dx = mouse_x - self.location[0]
        dy = mouse_y - self.location[1]

        # Hitung jarak
        distance = math.sqrt(dx**2 + dy**2)
        if distance == 0:
            return

        # Normalisasi
        dx /= distance
        dy /= distance

        # Percepatan bervariasi
        magnitude = distance * 0.01
        self.acceleration = [dx * magnitude, dy * magnitude]

        # Update kecepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        # Batasi kecepatan
        speed = math.sqrt(self.velocity[0]**2 + self.velocity[1]**2)
        if speed > self.topspeed:
            self.velocity[0] = (self.velocity[0] / speed) * self.topspeed
            self.velocity[1] = (self.velocity[1] / speed) * self.topspeed

        # Update posisi
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        x, y = self.location
        self.canvas.coords(self.id, x - self.radius, y - self.radius,
                                      x + self.radius, y + self.radius)

        # ✨ Tambahkan titik ke trail
        self.trails.append((x, y))
        if len(self.trails) > 30:
            self.trails.pop(0)  # hapus jejak paling lama

        # 🧹 Hapus jejak lama yang digambar
        for trail_id in self.trail_ids:
            self.canvas.delete(trail_id)
        self.trail_ids.clear()

        # 🎨 Gambar ulang trail
        for i, (tx, ty) in enumerate(self.trails):
            size = max(2, 6 - (30 - len(self.trails) + i) * 0.1)  # makin lama makin kecil
            color = f'#00aa00'  # warna hijau
            trail = self.canvas.create_oval(tx - size, ty - size, tx + size, ty + size, fill=color, outline="")
            self.trail_ids.append(trail)

        # Tampilkan info
        ax, ay = self.acceleration
        vx, vy = self.velocity
        px, py = self.location

        self.canvas.itemconfig(self.text_acc, text=f"Percepatan: ({ax:.2f}, {ay:.2f})")
        self.canvas.itemconfig(self.text_vel, text=f"Kecepatan: ({vx:.2f}, {vy:.2f})")
        self.canvas.itemconfig(self.text_pos, text=f"Posisi: ({px:.1f}, {py:.1f})")
        self.canvas.itemconfig(self.text_dist, text=f"Jarak ke mouse: {distance:.1f}")

# Setup tkinter
root = tk.Tk()
root.title("Bola dengan Percepatan Variabel + Trail")

canvas = tk.Canvas(root, width=600, height=400, bg="white")
canvas.pack()

bola = Bola(canvas, 300, 200)

def animasi():
    x = canvas.winfo_pointerx() - canvas.winfo_rootx()
    y = canvas.winfo_pointery() - canvas.winfo_rooty()
    bola.update(x, y)
    root.after(16, animasi)

animasi()
root.mainloop()
```

```python
# SCRIPT 27 - Simulasi Array of Movers yang Bergerak Menuju Mouse
import tkinter as tk       # Modul untuk membuat GUI jendela dan gambar
import math                # Untuk menghitung jarak dan kecepatan
import random              # Untuk memilih posisi awal secara acak

# 🎂 Cetakan untuk setiap bola
class Mover:
    def __init__(self, canvas, width, height):
        self.canvas = canvas             # Tempat menggambar (kanvas)
        self.width = width              # Lebar area layar
        self.height = height            # Tinggi area layar
        # 🎯 Posisi awal acak di layar
        self.location = [random.randint(0, width), random.randint(0, height)]
        self.velocity = [0, 0]          # Kecepatan awal nol
        self.acceleration = [0, 0]      # Percepatan awal nol
        self.radius = 8                 # Ukuran bola
        self.topspeed = 4               # Batas kecepatan maksimum
        # 🔵 Gambar bola di canvas (awal kosong, nanti di-update)
        self.id = canvas.create_oval(0, 0, 0, 0, fill='skyblue', outline='black')

    # 🚀 Fungsi untuk memperbarui posisi bola agar mendekati mouse
    def update(self, mouse_x, mouse_y):
        dx = mouse_x - self.location[0]     # Selisih posisi horizontal ke mouse
        dy = mouse_y - self.location[1]     # Selisih posisi vertikal ke mouse

        dist = math.sqrt(dx**2 + dy**2)     # Jarak ke mouse (pakai rumus Pythagoras)
        if dist == 0:
            return                          # Kalau sudah tepat di mouse, tidak perlu gerak

        dx /= dist                          # Normalisasi arah agar hanya menunjukkan arah
        dy /= dist

        magnitude = 0.5                     # Besarnya percepatan (selalu tetap)
        self.acceleration = [dx * magnitude, dy * magnitude]  # 🚗 Arah percepatan ke mouse

        # 💨 Update kecepatan berdasarkan percepatan
        self.velocity[0] += self.acceleration[0]
        self.velocity[1] += self.acceleration[1]

        speed = math.sqrt(self.velocity[0]**2 + self.velocity[1]**2)  # Hitung total kecepatan

        # 🛑 Batasi kecepatan agar tidak terlalu cepat
        if speed > self.topspeed:
            self.velocity[0] = (self.velocity[0] / speed) * self.topspeed
            self.velocity[1] = (self.velocity[1] / speed) * self.topspeed

        # 🧭 Update posisi berdasarkan kecepatan
        self.location[0] += self.velocity[0]
        self.location[1] += self.velocity[1]

        self.check_edges()  # 🌍 Jika keluar layar, muncul dari sisi lain

        # 🎨 Gambar ulang bola di posisi barunya
        x, y = self.location
        self.canvas.coords(self.id, x - self.radius, y - self.radius,
                                      x + self.radius, y + self.radius)

    # 🔄 Fungsi jika bola keluar dari layar, muncul kembali dari sisi berlawanan
    def check_edges(self):
        if self.location[0] > self.width:
            self.location[0] = 0
        elif self.location[0] < 0:
            self.location[0] = self.width

        if self.location[1] > self.height:
            self.location[1] = 0
        elif self.location[1] < 0:
            self.location[1] = self.height

# ==== Bagian Utama Program ====
root = tk.Tk()  # Buat jendela utama
root.title("Array of Movers - Menuju Mouse")

canvas_width = 640
canvas_height = 360
canvas = tk.Canvas(root, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# 🟢 Buat 20 bola yang disimpan dalam daftar movers
movers = [Mover(canvas, canvas_width, canvas_height) for _ in range(20)]

# 🎬 Fungsi animasi berulang (frame per frame)
def animate():
    # Ambil posisi mouse saat ini
    x = canvas.winfo_pointerx() - canvas.winfo_rootx()
    y = canvas.winfo_pointery() - canvas.winfo_rooty()

    # Update semua bola untuk bergerak mendekati mouse
    for mover in movers:
        mover.update(x, y)

    root.after(16, animate)  # Panggil fungsi ini lagi setelah 16ms (~60fps)

animate()  # Jalankan animasi pertama
root.mainloop()  # Tampilkan jendela GUI
```

```python
# SCRIPT 28 - Simulasi Ikan Gugup yang Bergerak Acak
import tkinter as tk
import random
import math

# -----------------------------
# Bagian 1: Membuat Kelas Ikan
# -----------------------------
class Fish:
    def __init__(self, canvas, lebar, tinggi):
        # Set awal posisi dan kecepatan ikan
        self.canvas = canvas  # Tempat gambar ikan
        self.lebar = lebar    # Lebar layar
        self.tinggi = tinggi  # Tinggi layar
        
        # Posisi awal acak
        self.posisi = [random.randint(0, lebar), random.randint(0, tinggi)]
        
        # Kecepatan awal acak antara -2 sampai 2
        self.kecepatan = [random.uniform(-2, 2), random.uniform(-2, 2)]
        
        # Awalnya tidak ada percepatan
        self.percepatan = [0, 0]  
        
        # Kecepatan maksimum ikan
        self.kecepatan_max = 3  
        
        # Bentuk ikan (lingkaran orange)
        self.bentuk = canvas.create_oval(0, 0, 0, 0, fill='orange')
        self.ukuran = 12  # Ukuran ikan
# -----------------------------
# Bagian 2: Cara Ikan Bergerak
# -----------------------------
    def update(self):
        # 1. Kadang ikan ganti arah tiba-tiba (5% kemungkinan)
        if random.random() < 0.05:
            sudut = random.uniform(0, 2 * math.pi)  # Arah acak
            kekuatan = 0.5
            self.percepatan = [math.cos(sudut) * kekuatan, math.sin(sudut) * kekuatan]
        
        # 2. Hitung kecepatan baru
        self.kecepatan[0] += self.percepatan[0]  # X
        self.kecepatan[1] += self.percepatan[1]  # Y
        
        # 3. Pastikan tidak terlalu cepat
        total_cepat = math.sqrt(self.kecepatan[0]**2 + self.kecepatan[1]**2)
        if total_cepat > self.kecepatan_max:
            self.kecepatan[0] = (self.kecepatan[0] / total_cepat) * self.kecepatan_max
            self.kecepatan[1] = (self.kecepatan[1] / total_cepat) * self.kecepatan_max
        
        # 4. Pindahkan ikan
        self.posisi[0] += self.kecepatan[0]
        self.posisi[1] += self.kecepatan[1]
        
        # 5. Cek tepi layar
        self.cek_tepi()
        
        # 6. Gambar ikan di posisi baru
        x, y = self.posisi
        self.canvas.coords(self.bentuk, 
                          x - self.ukuran, y - self.ukuran,
                          x + self.ukuran, y + self.ukuran)
# -----------------------------
# Bagian 3: Ikan Takut Tepi Layar
# -----------------------------
    def cek_tepi(self):
        jarak_aman = 20  # Jarak dari tepi saat mulai takut
        kekuatan_balik = 1.5  # Seberapa kuat menghindar
        
        # Jika terlalu kiri
        if self.posisi[0] < jarak_aman:
            self.percepatan[0] = kekuatan_balik  # Dorong ke kanan
            
        # Jika terlalu kanan
        elif self.posisi[0] > self.lebar - jarak_aman:
            self.percepatan[0] = -kekuatan_balik  # Dorong ke kiri
            
        # Jika terlalu atas
        if self.posisi[1] < jarak_aman:
            self.percepatan[1] = kekuatan_balik  # Dorong ke bawah
            
        # Jika terlalu bawah
        elif self.posisi[1] > self.tinggi - jarak_aman:
            self.percepatan[1] = -kekuatan_balik  # Dorong ke atas
# -----------------------------
# Bagian 4: Membuat Layar dan Animasi
# -----------------------------
# Buat jendela utama
root = tk.Tk()
root.title("Ikan Gugup")

# Ukuran kolam (canvas)
lebar_kolam = 640
tinggi_kolam = 360
kolam = tk.Canvas(root, width=lebar_kolam, height=tinggi_kolam, bg="lightblue")
kolam.pack()

# Buat satu ikan
ikan = Fish(kolam, lebar_kolam, tinggi_kolam)

# Fungsi animasi
def animasi():
    ikan.update()  # Update posisi ikan
    root.after(20, animasi)  # Ulangi setiap 20ms

# Mulai animasi
animasi()
root.mainloop()
```

```python
# SCRIPT 29 - Simulasi Kelinci Melompat
# -----------------------------
# 🐇 Bagian 1: Membuat Kelas Kelinci
# -----------------------------
import tkinter as tk
import random
import math

class Kelinci:
    def __init__(self, kanvas, lebar, tinggi):
        self.kanvas = kanvas  # Tempat menggambar
        self.lebar = lebar    # Lebar layar
        self.tinggi = tinggi  # Tinggi layar
        
        # Posisi awal acak
        self.posisi = [random.randint(0, lebar), random.randint(0, tinggi)]
        
        # Kecepatan awal (diam)
        self.kecepatan = [0, 0]
        
        # Percepatan awal (tidak ada)
        self.percepatan = [0, 0]
        
        # Ukuran kelinci
        self.ukuran = 14
        
        # Kecepatan maksimum
        self.kecepatan_max = 7
        
        # Waktu tunggu sebelum lompat
        self.waktu_tunggu = 0
        
        # Bentuk kelinci (lingkaran pink)
        self.bentuk = kanvas.create_oval(0, 0, 0, 0, fill='pink')

# -----------------------------
# 📊 Bagian 2: Logika Melompat (Update)
# -----------------------------
    def update(self):
        # Tampilkan informasi perhitungan
        print(f"\n--- Perhitungan Frame Baru ---")
        print(f"Posisi awal: ({self.posisi[0]:.1f}, {self.posisi[1]:.1f})")
        print(f"Kecepatan awal: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 1. Cek apakah sudah waktunya lompat
        if self.waktu_tunggu <= 0:
            print("Waktunya lompat!")
            # Buat lompatan acak
            sudut = random.uniform(0, 2 * math.pi)  # Arah acak
            kekuatan = random.uniform(3, 6)         # Kekuatan acak
            self.percepatan = [
                math.cos(sudut) * kekuatan, 
                math.sin(sudut) * kekuatan
            ]
            print(f"Percepatan baru: ({self.percepatan[0]:.1f}, {self.percepatan[1]:.1f})")
            
            # Set waktu tunggu acak (20-40 frame)
            self.waktu_tunggu = random.randint(20, 40)
            print(f"Set waktu tunggu: {self.waktu_tunggu} frame")
        else:
            # Tidak lompat, kurangi waktu tunggu
            self.waktu_tunggu -= 1
            self.percepatan = [0, 0]
            print(f"Menunggu... sisa {self.waktu_tunggu} frame")
        
        # 2. Update kecepatan
        self.kecepatan[0] += self.percepatan[0]
        self.kecepatan[1] += self.percepatan[1]
        print(f"Kecepatan setelah percepatan: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 3. Perlambatan (efek gesekan)
        self.kecepatan[0] *= 0.85
        self.kecepatan[1] *= 0.85
        print(f"Kecepatan setelah perlambatan: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 4. Batasi kecepatan maksimum
        total_cepat = math.sqrt(self.kecepatan[0]**2 + self.kecepatan[1]**2)
        if total_cepat > self.kecepatan_max:
            self.kecepatan[0] = (self.kecepatan[0]/total_cepat) * self.kecepatan_max
            self.kecepatan[1] = (self.kecepatan[1]/total_cepat) * self.kecepatan_max
            print(f"Kecepatan dibatasi: ({self.kecepatan[0]:.1f}, {self.kecepatan[1]:.1f})")
        
        # 5. Update posisi
        self.posisi[0] += self.kecepatan[0]
        self.posisi[1] += self.kecepatan[1]
        print(f"Posisi baru: ({self.posisi[0]:.1f}, {self.posisi[1]:.1f})")
        
        # 6. Cek tepi layar
        self.cek_tepi()
        
        # 7. Gambar ulang kelinci
        x, y = self.posisi
        self.kanvas.coords(self.bentuk, 
                          x - self.ukuran, y - self.ukuran,
                          x + self.ukuran, y + self.ukuran)

# -----------------------------
# 🌍 Bagian 3: Cek Tepi Layar
# -----------------------------
    def cek_tepi(self):
        # Jika kelinci keluar layar, muncul di sisi seberang
        if self.posisi[0] < 0:
            self.posisi[0] = self.lebar
            print("Kelinci keluar kiri, muncul di kanan")
        elif self.posisi[0] > self.lebar:
            self.posisi[0] = 0
            print("Kelinci keluar kanan, muncul di kiri")
            
        if self.posisi[1] < 0:
            self.posisi[1] = self.tinggi
            print("Kelinci keluar atas, muncul di bawah")
        elif self.posisi[1] > self.tinggi:
            self.posisi[1] = 0
            print("Kelinci keluar bawah, muncul di atas")

# -----------------------------
# 🖥️ Bagian 4: Program Utama
# -----------------------------
# Buat jendela utama
root = tk.Tk()
root.title("Kelinci Melompat")

# Ukuran layar
lebar_layar = 640
tinggi_layar = 360

# Buat kanvas (area gambar)
kanvas = tk.Canvas(root, width=lebar_layar, height=tinggi_layar, bg="lightgreen")
kanvas.pack()

# Buat satu kelinci
kelinci = Kelinci(kanvas, lebar_layar, tinggi_layar)

# Fungsi animasi
def animasi():
    kelinci.update()
    root.after(30, animasi)  # Update setiap 30ms

# Mulai animasi
print("=== Program Dimulai ===")
animasi()
root.mainloop()
```
